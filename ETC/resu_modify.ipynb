(1) 매매자료 공동위탁관리 관련
    - 금융위 금융혁신과와 가상자산사업자의 시스템구축관련 애로사항등을 논의하였고, 공동위탁관리의 법률적 이슈 해소를 위한 유권해석이 필요하다는데
      공감대를 형성하였으며, 이 문제에 대해 신속히 처리될 수 있도록 할 것입니다. 하지만, 그 결과는 아직 장담할 수 없습니다. 

(2) 매매자료 축적 양식 및 시스템 구축 
    - 첨부와 같이 모범사례 양식을 확정하고자 합니다. 이 양식을 참고하시어 시스템 구축을 시작해주시기 바라며, 
      구축하시면서 발생할 수 있는 문제점, 문의사항은 언제든지 연락부탁드립니다. 

(3) 공동위탁관리 여부와 관계 없이 진행
    - 코스콤과 공동위탁관리 사업모델에 대해 논의한 결과, 설령 공동위탁관리를 진행한다 하더라도 각 거래소가 정해진 매매데이터 양식에 따라 
      매매자료를 코스콤에 전송해야 위탁관리가 가능하다고 합니다. 즉, 위 매매자료 축적 양식에 맞게 정보가 생성될 수 있도록 각 거래소가 
      기본적인 시스템을 갖추어야 하는 것이죠. 또한, 공동위탁관리에 따른 비용 문제, 위탁-수탁자간 법류적 책임 문제 등 고려할 사항이 많기 때문에
      공동위탁관리가 만능의 해결책은 아닐 수 있습니다. 법시행까지 시간은 계속 흘러가고 있으니 준비할 수 있는 부분은 
      계속해나가는 것이 필요할 것입니다.
