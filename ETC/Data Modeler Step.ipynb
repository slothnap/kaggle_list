# 데이터 모델링 역량 성숙도 

- 1단계 : 일반인
    - 비기술자
    - 데이터모델을 거의 본적없고, 어떤 용도인지도 모름
- 2단계 : 초보 모델러
    - 엔터티, 속성, 관계 등 데이터모델의 구성요소와 표기법을 이해함
    - 데이터모델의 용도를 알고 있음(데이터모델은 DB테이블로 구현)
    - 데이터요구 파악이 어렵기는 하지만 시행착오를 거쳐 테이블 설계까지 진행할 수는 있으나, 그 결과물에 대해 본인 스스로도 의구심을 가짐
- 3단계 : 중급 모델러
    - 개발 문서 또는 인터뷰를 통해 데이터 요구사항을 도출하고, 데이터모델로 형상화 할 수 있음
    - 익숙한 업무 영역에 대해서 나름 모델링에 자신감을 가지기 시작하지만, 막상 다른 이해관계자의 검토 결과 또는 지적 사항에는 세련되게 대응하지는 못함
- 4단계 : 고급 모델러
    - 익숙하지 않은 새로운 업무 영역에서도 데이터 요구사항을 신속하게 도출 할 수 있음
    - 이해 상충되는 여러 요구 및 상황에 맞는 다양한 대안 모델과 설명 내용을 제시할 수 있음
    - 복잡한 데이터 요구에 대해서도 모델로 잘 형상화할 수 있고, 다양한 이해관계자를 설득할 수도 있음
- 5단계 : 특급 모델러
    - 복잡한 데이터 개념에 대해서 실제 업무 또는 비유 등을 통해 잘 설명할 수 있음
    - 조직에 적합한 모델 관리 체계를 제시할 수 있음 (단순 관리 기능 뿐만 아니라, 모델링 능력이 조직내에 체화 될 수 있게 함)

# DA 직무 역량

2023년 IT분야 역량체계 ITSQF직무기술서를 참고


한국 소프트웨어산업협회와 정보기술산업 인적자원개발위원회가 발간한 2023년 ITSQF 직무기술서에서 정의한 "데이터 아키텍트(이하 DA)"의 레벨 별 직무정의 내용을 참고하였습니다.
ITSQF에서 정의한 직무기술서에는 DA 직무가 포함하는 역할을 "데이터 모델러, 데이터 표준화 담당자, 데이터 품질관리자"로 한정하고 있습니다.

DA는 데이터 아키텍처를 구축하고 운영하는 직무입니다.

DA가 가지고 있어야 할 능력과 지식(직무)의 내용은
. 데이터 거버넌스 이해 - 데이터 품질, 보안, 규정, 표준준수를 보장하기 위한 데이터 거버넌스 정책 및 절차에 대한 이해
. 데이터 전략, 데이터베이스 관리,데이터 보안,데이터 품질관리,데이터 표준,데이터 아키텍처 운영을 위한 직무 지식
. 데이터 모델링 기술
. 데이터 품질관리 도구 활용 기술
. 데이터 거버넌스를 위한 데이터 관리정책 수립 및 시행 프레임워크 지식
. 전사 아키텍처와 데이터 품질관리에 대한 지식을 바탕으로 전사에서 보유한 정형데이터와 비정형데이터를 체계적, 구조적으로 정의하고 검증, 관리

로 정의합니다.

ITSQF의 직무는 레벨(4)에서 레벨(8)로 직무능력을 등급화하였으며 DA는 다른 직무에 비해 출발 등급이 높은 레벨 6부터 레벨 8까지 있는데,

레벨 6 :
데이터 아키텍처 및 데이터 표준관리 정책을 수립하고, 데이터 품질을 검증 및 개선하는 활동을 원활하게 수행
- 데이터 모델링 수행
- 데이터 표준화 정책 수립
- 데이터 품질검증 활동
- 데이터베이스 검증
- 데이터 아키텍처 수행관리

레벨 7 :
데이터 거버넌스 - 데이터 분야 전문 지식을 바탕으로 전사데이터의 가용성,유용성,통합성, 보안성을 관리할 수 있는 조직구성, 데이터관리정책과 프로세스 수립 수행
- 데이터 거버넌스 수립
- 데이터 아키텍처 요구사항 분석
- 데이터 아키텍처 설계
- 데이터 모델 검증
- 데이터베이스 검증
- 데이터 품질관리 계획 수립
- 데이터 아키텍처 수행관리 : 일정, 조직관리

레벨 8 :
데이터 거버넌스 - 데이터 분야 최고의 전문 지식을 바탕으로 전사데이터의 가용성,유용성,통합성, 보안성을 관리할 수 있는 조직구성, 데이터관리정책과 프로세스 수립 수행
- 데이터 거버넌스 수립
- 데이터아키텍처 구축계획 수립

로 분류하고 있습니다.

데이터 모델링을 잘 한다는 의미는 DA 직무 레벨 6 정도는 된다는 의미로 볼 수 있습니다. 모델러에게 DA로서의 역할을 감당해달라는 요구를 현장에서 많이 합니다만, 구체적인 요구사항을 제대로 제시하지 못한 채 DA라는 단어만 붙이는 경향도 있습니다.
해야할 일거리가 참 많긴 합니다...:)

### 레벨 7
- 정의
    - 전사아키텍처에 대한 이해와 데이터에 대한 고도의 전문 지식을 바탕으로 데이터아키텍처 요구사항 분석, 데이터아키텍처 설계 및 품질관리 계획을 수립할 수 있는 수준
- 요구역량정의
    - 데이터 아키텍처 구축 계획 수립
      - 데이터 아키텍처 구축 계획 수립이란 전사 아키텍처 프레임워크 및 참조 모델을 기반 으로 데이터 아키텍처의 방향을 수립하고, 구축 계획을 수립하는 능력이다.    
    - 데이터 아키텍처 요구사항 분석
      - 데이터 아키텍처 요구사항 분석이란 정형 및 비정형 데이터의 구조와 흐름을 설계하기 위한 전사적 데이터 아키텍처 요구사항을 수집, 분석, 정의, 검증 및 확인하는 능력이다.   
    - 데이터 아키텍처 설계
      - 데이터 아키텍처 설계란 데이터 아키텍처를 상세화하고, 데이터 아키텍처 요구사항에 부합하는지 검증 및 확정하는 능력이다.    
    - 데이터 모델 검증
      - 데이터 모델 검증이란 비즈니스 데이터요건을 반영한 개념, 논리, 물리 데이터 모델을 관리하고 검증하는 능력이다.    
    - 데이터베이스 검증

# DBA 직무 역량

한글자 차이인 DA와 DBA를 마구 섞어 이야기하고 역할을 요구하는 사업장을 종종 발견합니다.
아무 이유없이 이들이 그렇게 요구하는 것일까? 돈이 없어서? 싼 맛에 둘 다 하게 하려고? 아니면 무슨 근거라도?

정답은 아닙니다만 한국 소프트웨어산업협회와 정보기술산업 인적자원개발위원회가 발간한 2023년 ITSQF 직무기술서에서 그렇게 직무를 정의하는 배경을 발견할 수 있습니다.

DBA의 직무를 아래와 같이 정의합니다.
"데이터베이스를 데이터 요구사항에 맞게 설계, 구축, 전환하고 최적의 성능과 품질을 확보할 수 있도록 데이터베이스를 수정, 개선, 백업을 수행하는 일"

레벨 별(참고로 DBA직무는 레벨 5부터 시작합니다) 상세 직무내용을 보면

레벨5
물리설계/구현 - 데이터베이스에 대한 전문적인 이해를 바탕으로 물리 DB 설계, 구현 및 운용
- 물리 데이터베이스 운영
- 백업 및 복구 전략에 따른 운용
- 데이터 표준화

레벨6
논리설계/성능분석 - 데이터베이스에 대한 개념/논리/물리 DB 설계, 데이터 전환 설계와 성능이슈 진단 및 해결 수준
- 데이터베이스 요구사항 분석
- 논리/물리 데이터베이스 아키텍처 설계
- 데이터 전환설계
- 데이터베이스 성능 확보전략 수립
- 데이터베이스 운영 아키텍처 설계
- 논리 데이터 구조 설계

레벨7
요구분석/개념설계 - 비즈니스에 대한 이해를 바탕으로 고도화된 DB개념 설계, 데이터 전환에 따른 핵심 요소를 개발하여 설계에 반영
- 데이터베이스 요구사항 분석
- 논리 데이터베이스 아키텍처 설계
- 데이터 전환설계
- 데이터베이스 운영 아키텍처 설계
- 개념 데이터 구조 설계

라고 정의하고 있습니다

데이터 모델링 직무와 같은 수준과 같은 역할을 DBA에게 정의하고 있습니다. 심지어 레벨 7은 DA에게 요구하는 7수준의 개념 수준의 상위 데이터 구조 설계 능력을 보유하길 원합니다.

현실은 똑똑한 DBA 한분이면 사업을 충분히 꾸려나갈 수 있다 신념이 단순히 한 사람의 신념을 넘어서 근거를 제시할 수 있는 "요구"가 됩니다.
