### 블록체인 입출금 시스템 고도화 


* 목표
	- 신규 상장 시 거래소 내 소스코드 및 DB에 대한 변경 작업 발생으로 인해 기존 거래소 시스템의 영향도 문제 발생
	- 거래소 분리 프로젝트 시행 전 역량 개발 및 실행 계획 논의 

* 효과	
	- 거래소 분리를 통해 서로 간의 장애 및 변경 영향도를 최소화
	- 관리 주체를 명확하게 구분 가능
	- 관리 페이지를 구현하여 업무 효율화 및 운영팀, 회계감사팀 등 입/출금 시스템 사용 운영자에 대한 편의성 개선


* 아키텍처
	- 전체 아키텍처  
		- https://bithumbcorp.atlassian.net/wiki/spaces/~6299b107cd636500697dafa4/pages/293602487/06-03.
	- 블록체인 데이터 수집 서비스 고도화 
		- 개요
			- 블록체인 데이터 서비스는 블록체인(코인) 노드와 연결하여 블록체인 네트워크 상의 블록, 
			  트랜잭션 정보를 수집하고 수집된 데이터를 비즈니스 주소를 기반으로 필터링하여 입출금 시스템에서 
			  자산 처리를 위한 데이터를 제공한다.
			- 또한, 입출금 시스템으로 부터 요청 받은 자산 이동 트랜잭션을 블록체인 네트워크에 요청할 수 있다.
			- 처리된 데이터를 API를 통해 외부 서비스로 제공할 수 있으며, 현재 사용 중인 블록체인 노드의 실시간, 
			  로그 모니터링 정보를 제공한다.
			- 블록체인 데이터 서비스 Prototype 은 실제 업무 적용에 앞서 블록체인 데이터 수집 및 처리를 위한 
			  기능 및 기술 검증을 위한 단계이다.
		- 핵심 기능
			- 다양한 블록체인 네트워크의 데이터를 비즈니스 요구사항에 맞게 수집하고 처리한다.
			- 다양한 블록체인 네트워크에 트랜잭션을 전송 할 수 있다.
			- 블록체인 네트워크 관련 기준정보를 관리(코인, 회원 주소, 코인 서버)한다.
			- OpenAPI Spec 기반 블록체인 데이터 API 제공 
			- 블록체인 데이터 로그 모니터링
			- 블록체인 노드 모니터링 및 실시간 지표 제공
		- 주요 Components 
			- 카프카 클러스터
				- 블록체인 데이터를 수집
				- 카프카 스트림즈을 활용하여 카프카에 저장된 데이터를 입/출금 정책에 맞게 처리
				- 트랜잭션 전송 요청 건 처리
			- Open API: 블록체인 데이터 및 네트워크 정보 제공을 위한 API 서비스
			- 기분 정보 관리: 블록체인 네트워크 및 서버, 코인, 주소 정보 관리
			- 모니터링 도구: 블록체인 노드 실시간 지표 및 알림 서비스 제공
			- 로그 모니터링: 블록체인 데이터 로그 모니터링
		- 프로젝트 정의 
			- https://bithumbcorp.atlassian.net/wiki/spaces/~6299b107cd636500697dafa4/pages/293602528/06-04.+Prototype
		- AS-IS 아키텍처 
			- https://bithumbcorp.atlassian.net/wiki/spaces/~6299b107cd636500697dafa4/pages/293604301/AS-IS
		- ERD
			- https://bithumbcorp.atlassian.net/wiki/spaces/~6299b107cd636500697dafa4/pages/293602915
			- 영역
				- 데몬서버
					- 서버, 데몬추가서버 
				- 트랜잭션
					- 트랜잭션수량로그, 트랜잭션탐색서버, 트랜잭션탐색관리, 트랜잭션탐색실패
				- 블록체인 네트워크
					- 토큰 카테고리
					- 블록체인네트워크수수료, 용도별수수료 
				- 코인
					- 상장코인, 코인수량, 코인수량스냅샷
				- 주소
					- 주소용도, 변환 주소, 침해사고주소 
	- 입출금 시스템 
		- 핵심 기능
			- 코인 입/출금 관련된 처리 기능을 제공
			- 블록체인 네트워크에 서명된 트랜잭션 요청 처리 
			- 코인 상세, 회원 주소, 블록체인 노드 정보 등의 기준정보 관리
			- 거래소와의 입/출금 관련 데이터 통신을 위한 API 제공
			- 입/출금 업무 처리를 위한 관리 페이지
		- 주요 Components
			- 입금 처리: 입금 관련 프로세스를 처리하는 서비스(블록체인 데이터로부터 발생)
			- 출금 처리: 출금 관련 프로세스를 처리하는 서비스(외부 API로 부터 발생)
			- 집금 처리: 코인 내부 이동(집금)을 위한 기능 제공(내부 작업자에 의해 발생)
			- 트랜잭션 관리: 서명된 트랜잭션 데이터를 관리 및 블록체인 노드로 요청
			- 기준 정보 관리: 코인, 주소, 회원, 블록체인 노드 등의 입출금 시스템 관리를 위한 모든 기준 정보 관리
			- 노티 서비스: 슬랙, 텔레그램, 문자 등의 알람 서비스
			- 카프카 클러스터: 입출금 시스템 내 콘솔망을 통한 이벤트 데이터 처리
			- 외부 API: 거래소 연동을 위한 API 서비스
				- 이벤트 기반: 거래소측에서 입출금 시스템 이벤트에 대한 실시간 데이터 처리가 필요한 경우
				- REST API: 거래소에서 조회가 필요한 경우
			- 관리 페이지: 기존 B Manage 화면과 유사한 백오피스 기능 제공
