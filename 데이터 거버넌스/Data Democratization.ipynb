### 데이터 민주화란
- 데이터 민주화는 기술적인 노하우와 관계없이 조직의 모든 사람이 편안하게 데이터로 작업 하고 데이터 에 대해 자신있게 이야기할 수 있도록 하여 결과적으로 데이터 정보에 입각한 의사 결정 을 내리고 고객 경험 을 구축 할 수 있도록 하는 지속적인 프로세스 입니다. 

### 데이터 민주화 3대 요소
- 직원들이 편안하게 데이터 관련 질문을 할 수 있도록 지원
- 모든 사람이 데이터 작업을 할 수 있도록 올바른 도구 제공
- 데이터의 민주화를 조직 전체의 문화적 변화가 필요할 수도 있는 진행 중인 프로세스로 인식

## 데이터 민주화 핵심
### 가장 일반적인 데이터 문제
- 필요한 데이터에 액세스할 수 없습니다.
- 데이터를 믿을 수 없어
- 데이터에 접근할 수 있지만 질문에 대한 답변을 찾는 기술이 부족합니다.
- 우리 회사에서 제공하는 분석 도구는 제품 팀을 위해 설계되지 않았습니다.
- 우리 회사의 데이터 전문가는 너무 바빠서 나를 도울 수 없습니다.


### 팀별 데이터 사용 방식
- 마케팅 은 데이터와 협력하여 매력적이고 전환율이 높은 콘텐츠와 캠페인을 만듭니다.
- Growth 는 데이터와 함께 실험을 실행하고 개인화된 경험을 제공합니다.
- 제품 및 엔지니어링 은 데이터를 사용하여 고객 가치를 창출하는 기능을 구축하고 그렇지 않은 기능은 폐기합니다.
- 지원 은 데이터와 함께 작동하여 더 빠른 해결을 제공합니다(사용자가 제품 내에서 수행한 작업과 수행하지 않은 작업을 확인하여).
- Customer Success 는 데이터를 사용하여 더 나은 고객 경험을 제공합니다(사용 패턴을 기반으로 고객에게 올바른 질문을 함으로써).
- Sales 는 데이터를 사용하여 전환할 가능성이 있는 잠재 고객을 식별합니다(무료 평가판 동안 수행한 작업을 확인하여).
- 경영진 은 데이터를 사용하여 비즈니스가 어떻게 수행되고 있는지, 향후 어디에 투자해야 하는지 이해합니다.-
