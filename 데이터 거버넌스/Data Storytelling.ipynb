### 데이터 기반 스토리텔링의 요소
- **데이터**   
  데이터 기반 스토리텔링은 (가능한 한) 깨끗하고 완전한 데이터를 기반으로 해야 합니다. 당연해 보일 수 있지만 데이터가 여러 국가, 비즈니스 단위 및 부서에 걸쳐 존재하기 때문에 어렵습니다. IIoT 와 같은 새로운 데이터 소스의 출현은  데이터 볼륨을 증가시킬 뿐입니다. 데이터에 빠져 있는 회사(대부분)의 경우 데이터 관리 솔루션은 모든 소스에서 데이터를 수집하고 신뢰할 수 있는 실행 가능한 통찰력을 제공하여 이러한 문제를 해결하는 데 도움이 됩니다. 여기에서 시작합니다. 

- **내러티브**   
  역사를 통틀어 인간은 스토리텔링을 통해 효과적으로 정보를 전달해 왔습니다. 데이터 기반 스토리텔링 역시 시작, 중간, 끝이 있는 전통적인 내러티브 스토리라인(일명 "스토리 아크")을 따릅니다. 내러티브는 데이터가 드러내는 이야기를 전달하고, 그 맥락을 강조하며, 잠재적인 조치를 제안해야 합니다. 데이터 스토리텔링 소프트웨어는 ERP 플랫폼과 함께 작동하여 여러 유형의 데이터 분석 (설명, 진단, 예측, 규범)을 통합하여 어떤 데이터가 스토리와 가장 관련성이 있거나 설득력이 있는지 밝혀내는 데 도움을 줍니다.  

- **시각화**   
  좋은 시각화는 독자가 빠르게 이해할 수 있는 방식으로 데이터 연결을 설명하고 잠재적인 결과를 고려하는 데 사용할 수 있습니다. 기존 스프레드시트 및 데이터 시각화 소프트웨어는 차트, 지도, 그래프 및 다이어그램을 생성할 수 있지만 그래픽과 내러티브를 결합하여 가장 중요한 컨텍스트와 의미를 제공합니다. 그림은 천 단어보다 가치가 있습니다. 수천 개의 Excel 행보다 가치가 있습니다. 

### 데이터 기반 스토리텔링이 효과적인 이유
- 의사 결정을 방해하는 엄청난 양의 데이터를 단순화합니다. Gartner 에 따르면 기존 분석 통찰력의 80%는 2022년까지 비즈니스 결과를 제공하지 못할 것입니다. 
- 순수한 데이터보다 두뇌의 더 많은 부분을 활용하여 이해력과 기억력을 높입니다.  
- 어떤 일이 일어나는  이유 를 설명할 수 있음
- 증거로 뒷받침된다  
- 데이터를 사용하여 새로운 패턴 발견 
- 독자의 감정과 지성을 활용할 수 있습니다. 
- 실행 가능한 통찰력 생성 
- 항상 최신 분석을 위해 라이브 데이터에 연결할 수 있습니다. 
- 대규모 개인화 가능 
- 청중에게 가치 제공 
- 클릭률 및 전환과 같은 지표에서 좋은 성과를 거두십시오.

### 데이터 스토리 및 증강 분석의 성장 추세
- 데이터 스토리텔링은 기업에 새로운 것이 아닙니다. 하지만 최근까지만 해도 대부분 시간이 많이 걸리는 수동 프로세스였습니다. 직원이 데이터를 가져와 시각화를 설계 및 개발하고 내러티브 초안을 작성하고 인사이트를 생성해야 했습니다. 회사는 일반적으로 대량의 데이터를 처리하는 데이터 과학자 또는 IT에 의존하여 분석 대시보드를 만듭니다. 이 직원들은 종종 명확하고 설득력 있는 이야기를 전개할 수 있는 소프트 스킬이 부족했습니다. 종종 결과 대시보드는 경영진(또는 다른 사람)이 쉽게 이해할 수 있는 방식으로 가장 관련성이 높은 정보를 전달할 수 없었습니다. 


- 오늘날 ERP 기술 시스템과 연결된 스토리텔링 플랫폼에는 데이터가 민주화되어 있으므로 비즈니스 분석가, 마케터, 아티스트 및 편집 직원은 데이터 과학자와 협력하여 매력적인 스토리를 만들 수 있습니다. 다른 부서의 직원을 데려오면 소프트 스킬이 추가될 뿐만 아니라 여러 관점에서 통찰력을 얻을 수 있습니다.  


- 올바른 도구를 사용하면 조직 내 모든 사람이 잠재적으로 데이터 스토리를 만들 수 있습니다. 인공 지능 , 머신 러닝 및 자연어 처리를 통합 한 노 코드 및 로우 코드 소프트웨어 및 증강 분석 도구 는 이전에는 많은 시간이 소요되었던 많은 기능을 자동화할 수 있습니다. 전체 프로세스가 더 빠르고 효율적이며 효과적입니다.  


- 따라서 Gartner 에 따르면 데이터 스토리(대시보드와 반대)는 2025년까지 분석을 소비하는 가장 광범위한 방법이 될 것이며 증강 분석 기술은 이러한 스토리의 75%를 자동으로 생성할 것입니다.
