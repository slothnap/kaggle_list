### 데이터 메시 정의

- 데이터 메시(Data mesh)는 분산된 아키텍처 프레임워크를 사용하는 데이터 관리에 대한 접근 방식입니다. 즉, 이는 비즈니스 전반에 걸쳐 특정 데이터 집합에 대한 소유권과 책임을 해당 데이터의 의미와 최상의 활용 방법을 이해하기 위해 전문 지식을 보유한 사용자에게 분배합니다.

 

 

- 데이터 메시 아키텍처는 데이터 레이크와 웨어하우스와 같은 다양한 소스에서 데이터를 연결 및 가져오고 관련 데이터 세트를 비즈니스 전반의 적절한 인적 전문가 및 도메인 팀에 분배합니다. 기본적으로 중앙 데이터 레이크에 있는 엄청난 양의 데이터가 관리 가능한 청크로 정렬되어 이를 이해하고 활용하는 데 가장 적합한 청크로 분산됩니다.

### 데이터 레이크 과제를 위한 데이터 메시 원칙

- 데이터 레이크와 데이터 메시(data mesh)에 대해 이야기할 때 기본적으로 빅 데이터를 살펴보겠습니다. 데이터를 "크게"하는 것은 단순히 엄청난 양이 아닙니다. 다른 기준들 중에서도 빅데이터는 복잡함, 가변성, 급하게 생성되는 것, 구조화되지 않은 것으로도 정의된다.

선형 데이터베이스는 스프레드시트와 같습니다
. 여기에는 모든 데이터 구성 요소가 적합해야 하는 열과 행 및 변경 불가능한 범주가 있습니다. 기계, 센서, 산업 소스에서 생성된 일부 데이터가 구조화되어 선형 데이터베이스에 적합합니다. 처리해야 하는 데이터 볼륨이 아무리 많더라도 100% 구조화되면 빅데이터 기준을 충족하지 않고 선형 데이터베이스에 보관할 수 있으므로 필터링과 추출이 비교적 간단합니다.   

 

그러나 갈수록 최신 빅 데이터는 구조화되지 않고 시각적 구성요소, 개방형 텍스트, 심지어 비디오 및 리치 미디어로 구성됩니다. 이 중요한 데이터는 많은 회사를 위해 수천 테라바이트의 정보로 구성될 수 있으며 단순히 표준 선형 데이터베이스에 저장될 수는 없습니다.

 

데이터 레이크를 입력하십시오. 빅데이터 볼륨이 증가하기 시작하면서 데이터 레이크는 복잡한 데이터를 원시 형식으로 중앙 저장소에 저장하고 액세스할 수 있는 곳으로 개발되었습니다. 데이터 레이크는 빅데이터 문제에 대한 탁월한 해결책을 제시하지만, 그럼에도 약점이 있습니다. 데이터 레이크에는 특정 분석 기능이 없어 검색, 인덱싱, 변환, 쿼리 및 분석 기능을 위한 다른 서비스에 종속됩니다. 또한 비즈니스 관리 관점에서 볼 때 데이터 레이크는 다음과 같은 세 가지 추가 과제를 제시합니다.

 

1.      데이터 레이크에 대한 소유권은 너무 많은 플레이어가 데이터를 생성하고 액세스하는 시기를 정의하기 위해 복잡합니다. 명확하게 정의된 역할 및 책임이 없는 경우, 동일한 데이터 세트를 서로 다른 관계자에 의해 다르게 관리할 수 있으므로 사용이 어려워지는 불일치가 발생할 수 있습니다. 마찬가지로 다른 데이터는 궁극적으로 이를 사용할 사용자가 적극적으로 관리하지 않을 때 무시되는 것으로 끝납니다. 데이터 메시 아키텍처는 데이터 거버넌스를 도메인별로 명확하게 분배하여 각 팀이나 도메인 전문가가 자신이 생성하고 사용하는 데이터를 관리할 수 있도록 합니다. 이를 뒷받침하기 위해 데이터 메시는 또한 연합 거버넌스 구조를 사용하여 데이터 모델링, 보안 정책 및 규제준수를 중앙에서 제어할 수 있도록 합니다.

 

2.      데이터 볼륨이 너무 크거나 중앙 데이터 관리자 자신이 이해하지 못하는 경우 데이터 레이크가 데이터 품질을 보장하지 못할 수 있습니다. 데이터 메시 아키텍처는 근본적으로 데이터를 중요한 제품으로 취급하여 데이터의 품질과 완전성을 데이터 관리의 최전선에 둡니다. 아마도 각 팀은 수집 중인 데이터에서 추정하고자 하는 가장 중요한 기준과 이슈를 알고 있습니다. 데이터 메시는 이러한 기준과 우선순위를 아키텍처에 통합함으로써 더 큰 규모의 데이터세트가 관련되어 있는 경우에도 명확하고 신선하며 완전한 데이터를 지속적이고 우선적으로 제공하는 데 도움이 될 수 있습니다. 물론 기계 학습 알고리즘이 적용되면 이러한 기준과 결과 데이터 세트가 시간이 지남에 따라 점점 정확하고 유용해집니다.

 

3.      데이터 레이크는 중앙 집중식 아키텍처와 전통적으로 어려운 데이터 검색 프로세스 및 프로토콜로 인해 병목 현상을 일으킬 수 있습니다. 즉, 일반적으로 많은 양의 연결된 데이터를 단일 IT 또는 데이터 관리 팀에서 관리할 수 있습니다. 그리고  데이터의 양(및 검색 요구)이 증가함에 따라 IT 팀은 과다 과세됩니다. 

 

또한 데이터 거버넌스 원칙을 준수하고 규제를 준수하도록 데이터를 검토하고 적절히 구성해야 합니다. 부당한 압력에 직면할 때 이러한 규제 준수 단계를 거쳐서 잠재적인 리스크와 손실이 발생하는 경향이 있을 수 있습니다. 반면에 데이터 메시 아키텍처는 데이터에 대한 충분한 관심을 갖는 인증된 특수 사용자에게 액세스 및 제어 권한을 부여하고 엄격한 제빵인 보안 프로토콜을 사용합니다.

 

 

이렇게 증가하는 데이터 레이크 과제에 직접 대응하여 데이터 메시(Data mesh) 원칙이 등장했습니다. 분산되고 민주화된 데이터 관리 아키텍처는 언제 어디서나 적합한 데이터를 원하는 사람에게 즉시 사용할 수 있도록 함으로써 비즈니스를 더 현명하고, 더 민첩하고, 더 정확하게 만들었습니다. 데이터 메시(Data mesh)는 데이터-제품(data-as-a-product)을 실제 현실로 만들어 장벽을 줄이고 정보의 가치를 우선시함으로써 팀이 필수 데이터에 빠르고 무방비하게 액세스할 수 있도록 합니다.

### 데이터 메시 아키텍처 설명
- 데이터 메시(Data mesh)가 데이터를 필수 비즈니스 관리 툴로 처리하는 분산된 형태의 데이터 아키텍처인 방법에 대해 알아보았습니다. 그리고 중요한 점은 독립 팀은 작업과 전문 지식 영역 내에서 데이터를 처리하는 동시에 중앙 집중식 데이터 관리 관행을 준수할 책임이 있습니다. 이러한 사고방식의 변화는 데이터 메시의 핵심입니다. 이를 더 잘 이해하기 위해 다음과 같은 세 가지 주요 컴포넌트가 있는 데이터 메시 아키텍처를 볼 수 있습니다.

 

1.      **데이터 소스**는 기본 원시 데이터가 입력되는 저장소(예: 데이터 레이크)를 나타냅니다. 클라우드 IIoT 네트워크, 고객 피드백 양식 또는 스크랩된 웹 데이터에서 수집되든, 네트워크 전반의 사용자가 필요에 따라 참조하고 처리할 원시 입력 데이터입니다. 데이터 레이크 접근법은 이러한 모든 데이터를 하나의 중앙 위치에 배치하는 반면, 데이터 메시 방법론은 대신 일련의 책임 영역 내에서 이 원시 데이터의 수집, 저장, 처리 및 추출에 대한 책임을 분배합니다.

 

2.      **데이터 메시 인프라**는 이 정보가 개별 부서 영역 내에서만 격리되는 것이 아니라 확립된 데이터 거버넌스 가이드라인을 준수하면서 조직 운영 네트워크 전반에 걸쳐 공유될 수 있음을 의미합니다. 이는 데이터 메시의 두 가지 핵심 요소인 셀프 서비스 데이터 플랫폼과 통합 거버넌스의 직접적인 결과입니다. 셀프 서비스 데이터 플랫폼은 데이터를 보편적으로 수집, 변환, 처리 및 서비스하기 위해 각 도메인에 필요한 툴과 인프라를 제공합니다. 한편, 연합 거버넌스 원칙은 조직 전반의 표준화를 보장하여 모든 도메인 팀 간의 간편한 데이터 상호 운용성을 가능하게 합니다.

 

3.      **데이터 소유자**는 데이터 메시의 최종 구성요소이며 부서 데이터에 대한 규정 준수, 거버넌스 및 범주화 프로토콜을 적용하는 일을 담당합니다. 예를 들어, HR 파일은 특정 보안 프로토콜을 사용하여 저장해야 하며, 이 프로토콜에는 사용하지 않아야 하며, 그러한 사용자에게는 릴리스되어야 합니다. 물론 각 부서에는 해당 부서 또는 목적에 고유한 범주와 유형의 데이터가 있습니다. 데이터 레이크 시스템에서 IT 팀은 호수에 덤프된 물건이 있는 모든 다른 데이터 소유자에 대해 이러한 모든 프로토콜과 범주를 잡아야 합니다. 반면에 데이터 메시 아키텍처는 영역 전문가가 자신의 데이터를 관리하고 품질 기준을 충족시킬 수 있는 영역 전문가보다 더 뛰어나기 때문에 도메인 소유자에게 이러한 문제에 대한 완전한 권한과 제어권을 제공합니다. 


### 실제 데이터 메시: 사용 주체 및 이유

- 데이터 관리 솔루션이 발전하고 더 성공하기 위해서는 다양한 어플리케이션과 운영에 적합하고 사용 가능해야 합니다. 데이터 메시 아키텍처와 사용자 친화성이 향상됨에 따라, 제품과 툴로서의 데이터에 대한 안전하고 분산된 접근 방식으로 비즈니스 기능을 확장할 수 있는 비즈니스 기능의 범위가 증가하고 있습니다.

 

- **영업**: 영업팀의 경우, 인수, 육성, 마감 리드로 이어집니다. 영업 팀 멤버가 책상에서 관리 태스크를 수행하는 데 더 많은 시간을 투자할수록 신규 고객과의 관계를 구축하는 데 드는 시간이 줄어듭니다. 데이터 메시 아키텍처를 사용하는 영업 팀 사용자는 가장 강력하고 관련성 있는 데이터 세트와 조합을 손쉽게 이용할 수 있도록 데이터 관리 및 검색 전문가일 필요가 없습니다. 영업 부서가 분석할 수 있는 적절한 모든 데이터를 확보하면 보다 실행 가능한 인사이트와 전략으로 변모합니다.



- **공급망 및 물류**: 현대의 공급망은 엄청난 범위의 혼란에 취약합니다. 경쟁 우위는 기업이 신속하게 전환하고 동일한 민첩성으로 위협과 기회에 대응할 수 있는 경우에 발생합니다. 오늘날의 글로벌 공급망 데이터는 고객 피드백, IIoT 네트워크, 디지털 트윈에 이르기까지 굵고 빠르게 유입되고 있습니다. 경험이 풍부하고 정통한 공급망 관리자가 실시간으로 이러한 모든 데이터 세트를 큐레이션하고 분석할 수 있는 경우, 기업은 강력한 인사이트 소스를 확보하여 감동을 얻을 수 있습니다.



- **제조**: 공급망의 일부로, 기업의 제조 작업은 시장의 급격한 변화와 급변하는 고객 수요에 똑같이 취약합니다. 과거에는 디자인과 R&D 팀이 과거 고객 데이터에 의존하여 다른 부서로부터 데이터를 제공해야 했습니다. 오늘날 데이터 메시(data mesh)는 드래프트 표 뒤의 R&D 및 테스팅 팀, 그리고 제조 층까지 모두 라이브 데이터 액세스를 제공합니다. 실시간 고객 피드백은 IIoT 네트워크와 디지털 시뮬레이션을 통해 즉각적이고 최신 정보로 제품 개발에 정보를 제공할 수 있으므로 공장의 운영 속도가 더 안전하고 신속하며 효율적으로 진행될 수 있습니다.



- **마케팅**: 오늘날 고객 수요와 기대는 미래를 형성하고 있으며 변화하며 전례 없는 속도로 성장하고 있습니다. 단일 브랜드는 일반적으로 소셜 미디어, 타겟 디지털 광고, 온라인 및 옴니채널 쇼핑 포털 전반에 걸쳐 무수한 소비자 접점을 가지고 있습니다. 현재 시장은 빠른 커스터마이징, 제품 수명 주기 단축, 엄청난 수준의 선택과 경쟁에 대한 욕구가 커지고 있다고 보고 있습니다. 이러한 추세를 이해하고 활용하려면 최신 마케팅 담당자가 다양한 데이터 세트에 실시간으로 동시에 액세스할 수 있어야 합니다. 과거에는 이 데이터가 다른 부서에 요청(및 대기)되었습니다. 그러나 데이터 메시 설정을 통해 마케팅 담당자는 자신의 관점에서 이 데이터를 즉시 큐레이션하고 액세스할 수 있습니다.



- **인사 관리**: HR 팀은 매우 복잡하고 민감한 대량의 데이터를 관리해야 합니다. 또한 원격 및 하이브리드 직장에 대한 추세가 증가함에 따라 데이터는 매일 더욱 복잡하고 지리적으로 다양해지고 있습니다. HR 팀이 급하게 해결해야 하는 규제 준수 및 법적 문제는 항상 언급하지 마십시오. 채용부터 퇴직까지 HR 리더는 모든 조직에서 가장 광범위하게 서로 다른 데이터 세트를 검증, 평가 및 분석할 수 있어야 합니다. 데이터 메시 아키텍처는 적절한 보안 프로토콜과 엄격한 제한적인 액세스를 가능하게 하며 동시에 권한이 있는 HR 사용자가 복잡한 내부 프로토콜과 다중 부서 관료에 의존하지 않고도 데이터와 정보에 빠르게 액세스할 수 있게 합니다.



- **재무**: HR과 마찬가지로 재무 및 회계 팀도 매우 중요하고 중요한 데이터에 대한 책임을 집니다. 최신 ERP 시스템은 인메모리 데이터베이스 기술을 사용하여 최신 리포트, 분석 및 예측을 사용자 정의하여 재무를 혁신하고 있습니다. 재무 팀이 최고의 데이터베이스와 ERP를 사용하고 있는 경우에도 오랜 세상과 경직된 문화, 심한 사일로, 관료적이고 낡은 학교 프로세스로 인해 장애물에 직면하는 경우가 많습니다. 데이터 메시 아키텍처는 재무 데이터의 조회 및 관리 방식에 근본적인 변화를 가져왔으며, 팀이 자신만의 노후화된 데이터 프로세스를 소유하고 수정할 기회가 있을 때 발생할 수 있는 정체된 사고를 흔들 수도 있습니다.

### 데이터 메시의 이점
- 과거에는 레거시 데이터베이스와 제한된 데이터 관리 인프라가 데이터를 단일 저장소에 보관하고 몇몇 데이터 관리자의 재량에 따라 측정한다는 사실에 기여했습니다. 이제 데이터는 비즈니스를 이끄는 연료이며, 경쟁적인 시기에 수익을 창출하고 이를 실현하는 방법을 가장 잘 아는 대상 전문가에게 자유롭게 제공해야 합니다. 데이터 메시 아키텍처의 주요 장점은 다음과 같습니다.


- **데이터 접근성 증가.** 데이터 메시(Data mesh)는 조직 내 모든 적임자가 자신의 직무에서 최고의 능력을 발휘하기 위해 필요한 데이터에 액세스할 수 있도록 합니다.



- **분석 기능 향상.** 데이터가 매일 사용되는 제품으로 간주되면 팀에서 계획 및 전략에 대한 데이터 우선 접근 방식을 취하기 시작합니다. 이로 인해 오류가 감소하고 비즈니스 개발에 대한 의견 중심의 접근법이 보다 객관적입니다.



- **커스터마이즈 가능한 데이터 파이프라인 및 프로세스** 성공을 달성하는 데 필요한 고유하고 맞춤화된 데이터 집합을 엄선하는 데 엄청난 번거로움 때문에 가장 수익성이 높은 프로젝트가 대부분일 수 있습니다. 데이터 메시(data mesh)를 통해 팀은 기존의 시간 또는 리소스 손실 없이 새로운 프로젝트 모델에 빠르게 액세스하고 테스트할 수 있습니다.


- **병목 현상을 줄일 수 있습니다.** 이는 IT 팀과 데이터 소유자 모두에게 확실한 성공/획득입니다. 뿐만 아니라 좌절과 자극의 원천을 줄임으로써 건전한 비즈니스 개발 방식을 고수하는 사일로를 없애는 데 도움이 될 수 있습니다.



- **중앙 데이터 관리 팀의 업무 부담 감소.** 이는 백로그와 실망감을 줄일 뿐만 아니라 재능 있는 IT 팀이 보다 전문적이고 흥미롭고 수익성 있는 추구에 전념할 수 있도록 많은 시간을 할애하는 것을 의미합니다.
