* 표준화 정의서 
	- 공통 원칙
		- 관용화된 용어를 우선하여 사용한다. 
			- 관용화된 용어를 사용함으로써 이해 당사자들 간 의사소통이 명확해진다. 
		- 한글명에 대해서는 복수의 영문명을 허용하지 않는다. (동음이의어 불가)
			- DB에 반영할 때 물리명을 어떤 것을 선택해야 할지 판단이 애매모호함
	- 표준 용어 
		- '~일자', '~일' 등 날짜를 의미하는 용어는 '~일자'로 통일한다. 
			- 동일한 의미의 날짜에 대하여 일자(년월일, YYYYMMDD)로 통일함으로써 의미전달이 명확해짐
		- 용어의 길이는 한글 12자,영문 24자 이내로 제한한다. 
			- DB에 반영할 때 길이에 대한 제약사항이 존재한다. 
	- 표준 코드
		- 코드성 속성은 맨 뒤에 '코드'를 붙인다. 
			- 명칭만으로도 코드 속성을 인지할 수 있어 가독성이 좋아짐
		- 코드는 전체 모델 내에서 유일하게 정의한다.  
			- 중복하여 사용할 경우 데이터 중복 및 데이터 불일치가 발생될 수 있음
	- 표준 도메인
		- 표준 도메인은 기본적으로 Number, String, Datetime으로 정의한다. 



* 엔터티 정의서 
	- 고객 엔터티
		- 개요: 중고차 거래(매입 및 판매)를 위한 고객을 관리함
		- 성격: 키 엔터티이며, 개인이나 법인 고객이라는 실체(개체집합)을 관리 
		- 구성상의 특징: 식별자 고객번호로 구분하고, 매매 엔터티의 부모엔터티임
		- 특이사항: 개인과 법인이라는 서브타입으로 구성

	- 차량 엔터티
		- 개요: 중고차 거래의 대상인 차량을 관리함
		- 성격: 키 엔터티이며, 차량이라는 실체(개체집합)을 관리 
		- 구성상의 특징: 식별자 차량번호로 구분하고, 매매차량과 사고현황의 자식엔터티임
		- 특이사항: 승합차, 화물차 등 차량종류와 국산차, 수입차 등 국산수입 구분이라는 서브타입으로 구성

* 모델링 지문 팁
    - 엔터티        == 네모 
    - 속성          == 밑줄 
    - 식별자        == 밑줄 두개 
    - 서브타입      == 동그라미 
    - 서브타입 속성 == 동그라미 밑줄 


* 모델링 체크사항
    - 엔터티
    - 서브타입
    - 속성 null 타입 
    - 관계 (관계명, 식별관계 등)
    - 지문에서 해당 교차엔터티에 "~ID" 또는 "~번호" 부여하면 비식별관계, 그런 언급없으면 



* 모델링 문제 유형 
    - 재귀관계 반드시 나온다. 
    - 제약사항: 시스템속성(입력일시, 입력자, 수정일시, 수정자) 작성은 논리모델에서는 생략한다. 물리모델에서 명시
    - 지문에서 "~별"은 교차 엔터티를 뜻함





    



# ERD 그릴때 주의
1. 엔터티의 끝은 둥글게 그린다. 
2. 관계명을 적는다
    - 관계의 식별 및 선택 제대로 적는다
3. 상속의 FK를 기입한다.

# 표준화 정의서 

**공통원칙**  
	1. 관용화된 용어를 우선하여 사용한다.    
	2. 한글명에 대해서 복수개의 영문명을 허용하지 않는다.  

**표준용어**   
	1. "일자", "일" 등 날짜를 의미하는 용어는 '일자'로 통일한다.   
    2. 용어의 길이는 한글12자, 영문 24자이내로 제한한다.   
    
**표준코드**  
	1. 코드 속성 명칭에 대하여 맨 끝에 '코드'를 붙여 사용한다.   
	2. 코드는 전사모델에서 유일하게 정의한다.   
		
**표준도메인**  
	1. 표준 도메인은 기본적으로 Number, String, Datetime으로 한다.   

## 표준코드
- 의뢰구분코드
    - 1: 긴급의뢰
    - 2: 일반의뢰
- 작업구분코드
    - 1: 사내 자체작업
    - 2: 설비업체 외주작업
## 표준도메인
- 문자
    - 명: varchar(100)
    - 개요: varchar(200)
- 일자
    - 년도: char(4)
    - 일자: Date
- 금액
    - 원화금액: number(18,0)
- 번호
    - 주민등록번호: varchar(13)

# 엔터티 정의서

**고객 엔터티**  
	- 개요: 중고차 거래(매입 및 판매)를 위한 고객을 관리함  
	- 성격: 키 엔터티이며, 개인이나 법인 고객이라는 실체(개체집합)을 관리  
	- 구성상의 특징: 식별자 고객번호로 구분하고, 매매 엔터티의 부모 엔터티임  
	- 특이사항: 개인과 법인이라는 서브타입으로 구성  
  
**차량 엔터티**  
	- 개요: 중고차 거래의 대상인 차량을 관리함  
	- 성격: 키엔터티이며, 차량이라는 실체(개체집합)을 관리  
	- 구성상의 특징: 식별자 차량번호로 구분하고, 매매차량과 사고현황의 자식엔터티임  
	- 특이사항: 승합차, 화물차 등 차량종류와 국산차, 수입차 등 국산수입 구분이라는 서브타입으로 구성되며,     
	            선택 속성으로 사고차량여부가 존재(사고차량여부가 Optional이라 가정 시)  
    
**보유도서 엔터티**    
	- 개요: 도서관에서 보유하고 있는 도서를 관리  
	- 성격: 개체 엔터티로 키엔터티로 분류되며 개념이 아닌 도서라는 실체를 관리  
	- 구성상의 특징: 도서 종류로 구분될 수 있으며, 대출도서 엔터티의 부모가 된다.  
	- 특이사항: 도서상태는 예약, 대출중, 대출기능의 Sub Type으로 구성  

**시장 엔터티**    
	- 개요: 도서관에서 보유하고 있는 도서를 관리  
	- 성격: 개체 엔터티로 키엔터티로 분류되며 개념이 아닌 도서라는 실체를 관리  
	- 구성상의 특징: 도서 종류로 구분될 수 있으며, 대출도서 엔터티의 부모가 된다.  
	- 특이사항: 도서상태는 예약, 대출중, 대출기능의 Sub Type으로 구성   

## 공통 원칙
1. 관용화된 용어를 우선하여 사용한다.
2. 영문명(물리명) 전환 시 발음식은 지양한다.
3. 일반적인 명명규칙 시 띄어쓰기는 하지 않는다.
4. 한글명에 대해서는 복수의 영문명을 허용하지 않는다.(동음이의어 불가)
5. 영문명에 대해서는 복수의 한글명을 허용한다.(이음동의어 허용)


## 표준 용어
1. '일자', '일' 등 날짜를 의미하는 용어는 '~일자'로 통일하여 사용한다.
2. 적용일자, 유효일자 등의 내용은 유효일자로 통합하여 사용한다.
3. 용어는 띄어쓰기를 허용하지 않는다.
4. 용어의 길이는 한글의 경우 12자 이내, 영문의 경우 24자 이내로 제한한다.
5. 단독 인조식별자로서의 일련번호, ID, SEQ는 ID로 통합한다.
6. 영문 약어의 경우 5자 이내로 제한한다.
7. 필요시 단어와 단어의 구분은 _(언더바)로 한다.


## 표준 코드
1. 코드성 속성은 맨 뒤에 '코드'를 붙인다. 예) 상태코드, 결과코드
2. 코드는 알파벳과 문자열을 조합하여 일정한 길이로 구성한다.
3. 코드 속성에는 기본적으로 3자리 문열인 코드 도메인을 지정한다.
4. 코드는 전체 모델 내에서 유일하게 정의한다.


## 표준 도메인
1. 표준 도메인은 기본적으로 Number, String, Datetime으로 정의한다.
2. 원화금액 도메인은 (18,0)로 정의한다.
3. 외화금액 도메인은 (18,2)로 정의한다.
