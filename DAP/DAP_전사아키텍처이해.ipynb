# 전사아키텍처 이해(p40)
## 전사아키텍처와 데이터아키텍처(p43)
### 아키텍처 개요(p45)
    1. 아키텍처 정의
        - 가. 아키텍처 개념
            - 규칙(Rule) 관점
                - 원칙을 수립 및 적용 == 아키텍처의 연속성 유지
            - 모델(Model) 관점
                - 사전에 약속된 표기법에 의하여 표현 == 신속한 의사결정 가능 
            - 계획(Plan) 관점
                - 목표 아키텍처로의 이행 전략과 이행 계획
        - 나. 아키텍처의 중요성
            - 아키텍처는 기업이나 조직의 경영 전략 내지 비즈니스 목표를 달성하기 위한 방향을 제시
            - 정보기술체계의 나아갈 방향, 의사소통의 매개체
            - 정보시스템 품질 속성을 정의
            - 의사결정과 대안 제시, 복잡도 감소에 도움
        - 다. 아키텍처 활용과 기대효과
            - 정보관리 역량 강화
                - 의사소통 체계 수립
                - 정보 관리의 일관성 및 품질 수준 유지관리
            - 의사결정의 정확성과 신속성 제고
                - 정확한 의사소통을 가능하게 함
                - 새로운 서비스나 기능 도입에 대해 최적의 대응을 가능하게 함
            - 변화에 대한 유연한 대응과 비용 절감
                - 시스템 내,외부 요건 변경등 기술변화에 유연하게 대처
                - 중복적 요소 제거, 목표 아키텍처 이행 전략 수립
                - IT자원 효율적인 배치와 사용을 가능하게함 == 비용을 절감
    2. 전사아키텍처 개요
        - 가. 전사아키텍처 도입 배경 및 추진 현황
            - 정보시스템의 상호운용성 및 통합성 결여
            - 정보시스템의 비전과 구축 전략 수립의 필요성 증대
            - IT 대응 체계 미흡
            - 체계적 아키텍처 구축 미흡
        - 나. 전사아키텍처 정의
            - 전사아키텍처는 기업의 목표와 요구를 잘 지원하기 위해 IT 인프라의 각 부분들이
              어떻게 구성되고 작동되어야 하는가를 체계적으로 기술하는 것
            - 조직 및 업무 활동과 정보기술 간의 관계를 현재 모습과 향후 추구할 모습을 별도로 정의한 청사진
            - 다양한 정보기술의 혁신 활동과 관리 통제를 포함
            - 시스템의 도입과 구축뿐만 아니라 운영과 평가까지 통합적으로 관리하는 것
        - 다. 전사 개념
            - 전사(Enterprise)란 일반적으로 기업 또는 기관을 지칭
            - 공동의 목표를 추구하기 위해 고객과 상품 또는 서비스가 존재하고, 이를 지원하기 위한
              조직,자원,기술을 보유하며 필요한 업무 프로세스를 수행하는 조직의 집합체
            - 비즈니스의 변화 동인을 확인하고 전략을 정의, 비즈니스 활동 정보를 공유하는 조직 단위
        - 라. 아키텍처 도메인 구성
            - 1) 비즈니스 아키텍처
                - 기업이나 조직의 경영 목표를 달성하기 위한 업무 구조를 정의
                - 타 아키텍처(데이터, 애플리케이션, 기술)의 방향을 정의하고 검증하는 시발점
            - 2) 애플리케이션 아키텍처
                - 기업의 업무를 지원하는 전체 애플리케이션을 식별하고 연관성을 정의하며,
                  업무와 IT 특성을 고려하여 그룹화하고 범주화함으로써 전체 애플리케이션 구조를 체계화 하는 것
                - 애플리케이션이 지원하는 업무와 데이터의 특성을 고려하여 정의
                - 서비스 간의 상호 연관관계를 분석, 정의한다. 
            - 3) 데이터 아키텍처
                - 기업의 업무 수행에 필요한 데이터의 구조를 체계적으로 정의하는 것 
                - 데이터 구조, 데이터 표준, 데이터 관리 체계
            - 4) 기술 아키텍처
                - 비즈니스, 데이터, 애플리케이션 아키텍처에서 정의된 요건을 지원하는 전사의 기술 인프라 체계
                - 하드웨어, 시스템 소프트웨어, 통신 네트워크, 시스템 개발도구, 시스템 관리도구
    3. 데이터아키텍처 개요
        - 가. 데이터아키텍처 개념
            - 장기간에 걸쳐 출적된 데이터는 다양한 분석과 함께 업무적 영감과 새로운 비즈니스를 창출하는데 도움
            - 기업이나 조직의 업무 수행에 필요한 데이터의 구조를 체계적으로 정의하여 구축하고 관리하기 위한 방법
            - 효과적인 데이터아키텍처의 주요 특징
                - 비즈니스를 효과적으로 지원하는 데이터 전략(Aligned)
                    - 광범위한 데이터 전략의 일부를 볼 수 있다. 
                    - 현행 비즈니스를 충분히 지원하면서 미래의 비즈니스 변화를 유연하게 수용할 수 있는
                      유연성과 확장성을 갖도록 해야 한다. 
                - 필수, 핵심 데이터에 집중(Targeted)
                    - 비즈니스에 대해 우선순위를 부여하고 이에 따른 필수,핵심 데이터를 선정하여 이를 중심으로
                      체계적인 데이터아키텍처를 구축하는 것이 성공 가능성을 높이는데 유리 
                - 명확한 활동과 마일스톤(Actionable)
                    - 데이터아키텍처를 유지, 관리하기 위해 필요한 활동들을 정의하고, 수행 과정을 단계별로 체크
                - 변화하는 비즈니스 요구와 새로운 기술을 충족하기 위한 유연성(Evolutionary)
                    - 지속적으로 변화하는 비즈니스 요구 사항을 최소한의 변경으로 용이하게 수용할 수 있도록
                      유연성 및 확장성을 확보해야 한다.
            - 데이터아키텍처의 실패 요인
                - 기업이나 조직의 모든 데이터 및 정보 포함 시도
                - 전문성 결여와 내제 기술력에 대한 오해 
                - IT 부서의 독점적 소유 및 조치
                - 외부 구매에 의존하는 기술적 해결 시도
                - 이해 당사자에 의한 독자적 개발 진행 및 통제 부재
                - 학술적 접근과 과도한 복잡성
                - 이상에 치우친 원칙과 정책
                - 단계적 성과를 고려하지 않은 장기 수행 전략
                - 변화 수용에 대한 배척
        - 나. 데이터아키텍처의 중요성
            - 전사의 데이터 구성을 용이하게 파악 가능
            - 데이터 품질 확보 용이 
            - 명확한 의사소통 향상
            - 신속하고 적절한 의사결정 가능
            - 데이터 연계 및 상호운용성 향상에 따른 업무 정확성, 효율성 향상
        - 다. 데이터아키텍처 구성
            - 전사 데이터 영역 모델(계획자 관점)
                - 상위 수준의 전사 데이터 영역을 분류하여 표현한 것
                - 주제 영역은 기능과 대응되는 개념으로 그 목적과 용도가 유사하거나 공통된 사안을 내포하고
                  있는 데이터를 그룹화 한것으로, 전사의 범위 또는 규모에 따라 계층적으로 구성하기도한다. 
            - 개념 데이터 모델(책임자 관점)
                - 전사 수준의 데이터 모델로 단위 주제 영역 또는 핵심 엔터티 정도를 표현한 데이터 모델
            - 논리 데이터 모델(설계자 관점)
                - 개념 데이터 모델에서 정의한 주제 영역과 핵심 엔터티를 기본 정보로 하여 업무 요건을 
                  충족시키기 위한 데이터의 상세한 구조를 논리적으로 구체화한 것
                - 엔터티와 속성에 대한 명칭, 정의, 형식, 규칙, 코드 등 전사적인 차원의 표준을 정의
            - 물리 데이터 모델(개발자 관점)
                - 기술적 환경과 특성을 고려하여 특정 데이터베이스 관리 시스템이나 분산 서버 환경 등에
                  적합한 물리적 데이터 구조를 설계하고, 데이터베이스 객체를 정의한 것
### 아키텍처 프레임워크(p63)
    1. 아키텍처 프레임워크 개요
        - 가. 전사아키텍처 프레임워크
            - 전사아키텍처 정책과 전사아키텍처 정보, 전사아키텍처 관리 체계
        - 나. 데이터아키텍처 프레임워크
            - 데이터아키텍처 정책과 데이터아키텍처 정보, 데이터아키텍처 관리 체계로 구성 
    2. 전사아키텍처 프레임워크 구성
        - 가. 전사아키텍처 정책
            - 아키텍처 매트릭스
                - 전사아키텍처의 정보를 체계적으로 분류한 틀
                - 전사아키텍처 정보의 수준과 활용 계층을 결정하는 수단
                - 뷰와 관점이 교차하는 각 셀에는 전사아키텍처 정보의 실체가 되는 산출물을 정의하는 구조
                - 뷰: 비즈니스, 애플리케이션, 데이터, 기술
                - 관점: 계획자, 책임자, 설계자, 개발자
            - 전사아키텍처 비전
                - 전사아키텍처 구축 목표와 그 목표를 효과적으로 달성하기 위한 전략 및 방향
            - 전사아키텍처 원칙
                - 조직 구성원이 공유해야 할 규범
                - 전사아키텍처 대원칙, 아키텍처 원칙, 표준 등
        - 나. 전사아키텍처 정보
            - 현행 아키텍처
                - 아키텍처 도메인별로 정의된 산출물에 대하여 기업이나 조직의 현재 상태를 아키텍처 정보로 정의
            - 목표 아키텍처
                - 아키텍처 도메인별로 정의된 산출물에 대하여 기업이나 조직의 궁극적으로 달성하고자 하는 목표
                  아키텍처 
            - 전사아키텍처 이행 계획
                - 아키텍처 도메인별로 현재 모습에서 바람직한 목표 모습으로 이행하기 위한 이행 전략과 이행 계획
                  을 정의
        - 다. 전사아키텍처 관리
            - 전사아키텍처 관리 체계
                - 전사아키텍처 거버넌스
                - 구축된 전사아키텍처를 유지하고 개선하기 위한 제도적 기반을 수립하는 것
                - 정의된 전사아키텍처 원칙을 준수하도록 확인하고 통제하기 위한 조직과 프로세스를 정의한 것
            - 전사아키텍처 관리 시스템
                - 전사아키텍처의 정보 관리 효율성을 제고하고 전사아키텍처 정보의 공유를 활성화하기 위해
                  구축하는 정보시스템
                - 모델링 도구, 전사아키텍처 리포지터리, 전사아키텍처 포털
            - 전사아키텍처 평가
                - 전사아키텍처의 수준을 객관적이고 정확하게 평가할 수 있는 전사 아키텍처 성숙 모형이 필요
    3. 전사아키텍처 프레임워크 사례
        - ZEAF, FEAF, TEAF, DoDAF, TOGAF, 공공부문EAF
    4. 데이터아키텍처 프레임워크 구성
        - 가. 데이터아키텍처 정책
            - 정의
                - 기업이나 조직의 데이터 아키텍처 구축 목적과 방향을 정의
                - 전사아키텍처의 목적 및 방향과 얼라인먼트(Alignment)를 유지
            - 데이터아키텍처 매트릭스
                - 데이터아키텍처의 정보를 체계적으로 분류한 틀
                - 기업이나 조직이 관리하려는 데이터아키텍처 정보의 수준과 활용 계층을 결정하는 수단
                - C(Chief)DA(개괄정 관점), 주제영역, 데이터 통합 구조도, 데이터 관리 정책
                - DA(개념적 관점), 개념모델, 데이터 흐름도, 데이터 표준 관리
                - 모델러(논리적 관점), 논리모델, , 데이터 모델 관리
                - DBA(물리적 관점), 물리모델 데이터베이스, , 데이터베이스 관리
                - 유저(운용적 관점), 사용자뷰
            - 데이터아키텍처 비전
                - 데이터아키텍처 수립을 통하여 기업이나 조직이 궁극적으로 달성하고자 하는 모습
            - 데이터아키텍처 원칙
                - 데이터아키텍처 정보를 효율적으로 구축하고, 기업이나 조직의 목적에 맞게 데이터 아키텍처 정보를
                  효과적으로 활용하기 위해서 조직 구성원이 공유해야 할 규범을 말한다. 
                - 데이터아키텍처 대원칙, 데이터아키텍처 원칙, 표준
        - 나. 데이터아키텍처 정보
            - 현행 데이터아키텍처
                - 현행 주제영역 모델, 현행 개념 모델, 현행 논리 모델, 현행 물리 모델
            - 목표 데이터아키텍처
                - 목표 주제영역 모델, 목표 개념 모델, 목표 논리 모델, 목표 물리 모델
