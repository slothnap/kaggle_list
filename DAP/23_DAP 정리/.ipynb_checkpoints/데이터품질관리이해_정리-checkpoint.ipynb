1. 데이터 품질 관리 이해 (p639)
    - 데이터 품질 관리 프레임 워크 
        - 데이터 품질 관리 프레임워크
            - 데이터 품질 관리 원칙 
            - 데이터 품질 관리 전문 조직
            - 데이터 품질 관리 프로세스 
            - 인프라 
        - 데이터 품질 관리 프레임워크 구성 
            - 데이터 표준 
                - 표준 단어 
                - 표준 용어 
                - 표준 도메인 
                - 표준 코드 
            - 데이터 모델 
                - 개념 모델 
                - 논리 모델
                - 물리 모델 
                - DB Catalog 
            - 데이터 값 
                - 데이터 품질 지표 
                - 핵심 정보
                - 데이터 프로파일링 
                - 업무규칙 
            - 데이터 활용 
                - 데이터 매핑 정보 
                - 데이터 연관 관계 
                - 데이터 흐름
    - 표준 관점 데이터 품질 관리 
        - 세부 관리 대상
            - 표준 단어 사전 
                - 표준성
                    - 일반적으로 사용되는 사전적 의미 단어 추출
                    - 지나치게 업무에 의존적이거나 방언을 사용해서는 안되며, 약어의 사용도 최소화 
                - 참조 가능성
                    - 기업이나 기관에서 새로운 업무를 정의할 때 참조 
                - 일반성
                    - 일반인도 해당 단어의 의미를 이해할 수 있어야 한다. 
                - 대표성 
                    - 동의어를 가질 수 있으나 표준 단어는 비슷한 의미의 동의어들을 대표할 수 있어야 한다. 
            - 표준 도메인 사전 
                - 표준성
                    - 전사 차원에서 공통적으로 사용되는 속성을 대상으로 정의 
                - 유일성
                    - 동일한 내용의 중복 도메인이 서로 다른 이름으로 선언되지 않도록 관리 
                - 업무 지향성
                    - 지나치게 일반화하여 정의하기보다는 업무의 특성을 충분히 반영할 수 있도록 선언하여 관리 
            - 표준 용어 사전
                - 표준성
                    - 용어의 표준화를 통해 용어 사용의 차이에 따라 발생되는 전사 차원의 혼란을 최소화 
                - 일반성 
                    - 용어가 지나치게 업무 관점에서 정의되어 이해하기 힘들거나 의미상 혼란을 초래해서는 안된다. 
                    - 일반적인 의미와 전혀 다르게 사용된 용어는 적절한 다른 용어로 대체, 새로운 용어 개발을 자제 
                - 업무 지향성 
                    - 용어는 기업의 업무 범위 내에서 약어를 사용하거나 내부에서 별도로 정의하여 사용할 수 있어야 하낟. 
                    - 지나친 약어의 사용은 업무에 대한 이해도를 떨어뜨릴 수 있으므로 주의 
            - 표준 코드
                - 재사용성
                    - 기업에서 자체적으로 정의하여 사용하는 것보다 표준화 기구나, 정부, 공공 기관에서 정의한 코드를 재사용
                - 일관성
                    - 업무 범위 내에서 가능한 한 유일하게 정의 
                - 정보 분석성
                    - 가능한 범위의 데이터는 모두 코드화하여 관리 
                    - 사용자가 텍스트로 직접 입력하는 값을 최소화 
            - 데이터 표준 요소 
                - 통합성
                    - 데이터 표준 요소의 각 요소는 전사적으로 통합하여 관리 및 적용 
                - 일관성
                    - 정의된 표준 데이터가 데이터 모델 및 데이터베이스 스키마의 전 영역에 걸쳐 일관되게 적용되고 있는지 
                      정기적으로 검토 확인한다. 
    - 모델 관점 데이터 품질 관리 
        - 정의 및 관리 목적
            - 데이터 구조 최신 정보 유지, 데이터 모델 공유와 재사용성, 모델 변경 관리 
        - 세부 관리 대상 
            - 데이터 참조모델, 개념 데이터 모델, 논리 데이터 모델, 물리 데이터 모델 
            - 완전성
                - 모델 데이터는 개념-논리-물리 모델에 대한 모든 메타 데이터를 포함해야 한다. 
            - 일관성
                - 모델 데이터는 단어, 용어, 도메인 및 데이터 관련 요소 표준을 준수해 정의 
            - 추적성
                - 모델 데이터는 데이터 모델의 변경 이력에 대한 추적이 용이
            - 상호 연계성
                - 모델 데이터는 데이터 구조를 입체적, 체계적으로 관리 
                - 데이터 모델 간의 상호 연관 관계를 표현 
            - 최신성
                - 모델 데이터는 데이터 구조의 단계별 데이터 모델과 업무 규칙은 물론 물리 데이터도 논리적으로 일치해야 한다. 
            - 호환성 
                - 모델 데이터는 다른 종류의 관리 데이터와도 상호 호환이 가능해야 한다. 
    - 값(Value) 관점 데이터 품질 관리 
        - 정의 및 관리 목적 
            - 오류 데이터를 찾고 개선하는 일련의 품질 관리 활동 
        - 세부 관리 대상 
            - 품질 기준 
                - 품질 관리 대상
                    - 품질 측정 대상 시스템 및 테이블 관리 
                - 데이터 품질 지표 
                    - 완전성, 유효성, 일관성, 정확성
                - 핵심 정보 항목
                    - 가장 중요한 데이터 정보 항목 
            - 프로파일링 
                - 컬럼 분석
                - 패턴 분석
                - 날짜 분석
                - 코드 분석
                - 참주무결성 분석
            - 업무규칙
                - Inside-Out 도출 방법
                    - 데이터 통계적인 분석 오류데이터 잠재원인 파악 
                    - Low-Level의 필수성, 유효성, 업무규칙 자동 생성
                - Outside-In 도출 방법
                    - 비즈니스 이슈, 외부고객의 소리, 내부고객의 소리, 설문조사 
                - 근거규정 도출 방법
                    - 기관의 법령을 기초로 설계된 데이터 모델을 기준으로 표준 데이터 항목 업무규칙 생성 
            - 품질 측정
                - 품질 점검 측정 주기 설정 
            - 품질 개선
                - 오류추정 데이터에 대해 개선 담당자, 오류 원인, 개선 내용 지속적 관리 
    - 활용 관점 데이터 품질 관리 
        - 정의 및 관리 목적
            - 추적된 데이터 활용 여부, 데이터 흐름 관리, 활용관점 데이터 품질 체계적 관리 
        - 세부 관리 대상 
            - 데이터 흐름 관리 기준
                - 안전성
                    - 데이터 이동이 필요한 모든 소스와 타깃을 정의, 매핑 규칙 정의 
                - 유효성
                    - 데이터 클린징 규칙 정의 
                - 데이터 정합성 
                    - 소스와 타깃의 데이터가 매핑 규칙을 준수하여 데이터 정합성 보장 
            - 데이터 흐름 관리 방법 
                - 소스, 타깃 데이터 매핑 리스트 작성
                - 데이터 매핑 여부 검사 
                - 삭제 매핑 소스 검사 
                - 소스, 타깃 데이터 구조 동일 조사 
                - 변환 규칙 데이터 무결성 준수 검사 
                - 데이터 정합성 보장 검사 
                

                

2. 데이터 관리 프로세스 이해 (p659)
    - 데이터 품질 관리 정책
        - 데이터 품질 관리 원칙
            - 준수성
                - 데이터 관리 목적을 달성할 수 있도록 정의 
            - 불가변성
                - 아키텍처 원칙 변화에 의한 불가피한 경우를 제외하고는 쉽게 바뀌지 않도록 정의 
            - 이해성
                - 쉽게 이해할 수 있어야 하며, 의미가 불분명하여 발생하는 혼란을 최소화 해야 한다. 
            - 완전성
                - 정책 수립에 필요한 모든 사항을 정의 
            - 일관성 
                - 원칙 간 충돌이 없도록 정의 
        - 데이터 품질 관리 조직
            - 명확성
                - 담당자별로 수행해야 할 역할이 명확하게 정의 
            - 운영성
                - 업무 수행하는 데 필요한 교육 프로그램 등의 운영을 통해 원할한 수행이 가능해야 한다. 
        - 데이터 품질 관리 프로세스
            - 준수성
                - 정의된 데이터 품질 관리 프로세스는 데이터 관리 원칙에 맞게 정의 
            - 완전성
                - 데이터와 관련된 모든 요소가 빠짐없이 관리될 수 있도록 정의 
            - 상호 운용성
                - 기존의 다른 프로세스와 상호 연관 관계가 명확하게 정의되어 있어 적용함에 문제가 없어야 한다. 
        - 데이터 품질 관리 정책 수립 프로세스 
            - 데이터 품질 관리 정책 수립 -> 데이터 품질 관리 정책 검토 -> 데이터 품질 관리 정책 승인
              -> 데이터 품질 관리 정책 공표 
    - 데이터 표준 품질 관리 프로세스
        - 정의 및 관리 목적
            - 지속적인 표준화에 대한 교육과 개선/모니터링 활동
            - 신규 개발 시점에서 이루어지고, 기존 시스템과의 중복 표준이 허용될 수 있다. 
            - 표준 관리 대상 및 적용 대상이 많을 경우 표준화 도구 등을 활용한 자동화를 고려 
        - 조직 구성 및 역할 
            - 데이터 관리 책임자
            - 데이터 표준 담당자
            - DBA
            - 데이터 구조 담당자
            - 업무 담당자 
        - 데이터 표준 품질 관리 체계도 
            - 데이터 표준 수립 프로세스
                - 표준 대상 선정 -> 설문 및 인터뷰 -> 자료 수집 -> 데이터 표준 종합 분석 -> 표준화 방향성 수립
                  -> 표준화 방향성 검토 -> 데이터 표준화 -> 표준화 검토 -> 표준 배포 -> 데이터 구조 표준 적용
                  -> 표준 준수 검토 -> 데이터베이스 반영 
            - 데이터 표준 변경 프로세스
                - 단어, 용어, 도메인 변경 프로세스 
                    - 업무 요건 발생 -> 데이터 표준 요건 접수 -> 데이터 표준 조회 및 검토 -> 변경 영향 협의 / 대안 수립 
                      -> 표준 반영 및 배포 
                - 코드 변경 프로세스 
                    - 업무 요건 발생 -> 데이터 표준 코드 요건 접수 -> 데이터 표준 코드 조회 및 유효값 검토 
                      -> 변경 영향 협의 / 대안 수립 -> 표준 코드 반영 및 배포 
            - 데이터 표준 점검 프로세스 
                - 데이터 표준 진단 -> 진단 결과 분석 -> 데이터 표준 보완 -> 데이터 구조 영향도 분석 
                  -> 데이터 구조 변경 검토 -> 데이터베이스 반영 
    - 데이터 모델 품질 관리 프로세스 
        - 데이터 모델 품질 관리 체계도
            - 주제영역 관리 프로세스 
                - 주제영역 변경 요청 -> 주제영역 변경 검토 및 승인 -> 주제영역 변경 확정 및 반영 
            - 개념 데이터 모델 관리 프로세스 
                - 개념 데이터 모델 변경 요청 -> 개념 데이터 모델 검토 및 승인 -> 개념 데이터 모델 확정 및 반영 
            - 논리/물리 데이터 모델 관리 프로세스 
                - 논리/물리 데이터 모델 변경 요청 -> 데이터 표준 검토 및 반영 -> 논리/물리 데이터 모델 검토 및 승인 
                  -> 논리/물리 데이터 모델 확정 -> DB 객체(테이블) 생성
            - 데이터 모델 품질 점검 프로세스 
                - 논리-물리 데이터 모델 매핑 -> 물리-DB 데이터 모델 매핑 -> 논리-물리 데이터 모델 얼라인 분석 
                  -> 물리-DB 데이터 모델 얼라인 분석 -> 점검 결과 취합 및 보고 -> 점검 결과 검토 및 승인 
    - 데이터 값(Value) 품질 관리 프로세스 
        - 데이터 값 품질 관리 체계도 
            - 데이터 값 품질 진단 프로세스
                - 데이터 값 품질 진단 계획 수립 프로세스 
                    - 진단 계획 검토 -> 진단 수요 분석 -> 진단 대상 선정 -> 진단 방향 수립 
                      -> 진단 일정 및 계획 수립 -> 진단 추진 계획 승인 
                - 데이터 값 품질 진단 실시 프로세스 
                    - 프로파일링 기준 제시 -> 도메인 분석 -> 프로파일링 수행 -> 프로파일링 결과 분석 
                      -> 프로파일링 결과 배포 -> 오류(추정) 데이터 식별 
            - 업무규칙 관리 프로세스 
                - 업무규칙 도출 프로세스 
                    - 업무규칙 도출 기준 제시 -> 업무규칙 작성 -> 업무규칙 검토 -> 승인 여부 
                      -> 업무규칙 등록 -> 업무규칙 등록 통보

       관리할 것인지 애플리케이션 수준에서 관리할 것인지를 먼저 결정해야 하며 트리거에 의한 자동 변경은 DBMS 오류 시에 추적의 어려움이 있으므로 
       최소화해야 한다.

367. 3

369. 4

380. 4 / 2

       지나치게 업무에 의존적이거나 방언을 사용하지 않아야 한 다. 또한 약어의 사용도 최소화해야 한다.

       활용할 수 있는 상태로 보유하고 있는지의 여부이다. 최신성은 저장 데이터가 가장 최신의 형태와 값을 유지하고 있는지의 여부이며, 일관성은 데이터가

       저장되어 있지만 최신의 값과 일치하지 않아 반송되는 경우는 의미하고 있다.

392. 1 / 3

       역할이 명확하게 정의되어 있어야 함을 의미한다. 문제에 명시한 설명은 누가 수행해야 하는지에 대해 명확하게 지정하고 있지 않기 때문에

       표준화 요구사항 수집 시 데이터 관리 정책에 영향을 받기 때문에 직접적인 연관이 있다고 할 수 있다.

395. 4
     - 저장된 데이터의 값이 업무 규칙 수립에 중요한 기 준이 되는 것이 아니라 업무 규칙에 맞게 데이터 값이 저장되고 관리되어야 한다.

- 데이터 품질 관리 이해 (p656)
1. 3
    - 데이터 품질 관리 프레임워크의 구성은 데이터 품질 관리 원칙, 데이터 품질 관리 조직, 데이터 품질 관리 프로세스, 인프라로 구성되어 있다.
    
2. 4
    - 데이터 프로파일링 분석 방법에는 컬럼 분석, 날짜 분석, 패턴 분석, 코드 분석, 참조무결성 분석 둥이 있다.

3. 4 / 1
    - 업무규칙 도출 방법은 3가지 방법이 있다. Inside-Out, Outside-In, 근거 규정을 통한 업무규칙 도출 방법이 해당한다.

4. 4
    - 프로파일링을 통해 Min, Max, 평균, 패턴, 데이터의 분포 등의 결과를 획득할 수 있다.

5. 2
    - 표준화 컨설팅을 통하여 표준 사전, 표준 용어, 표준 도메인, 표준 코드, 표준 지침 등의 결과물을 얻을 수 있다.

6. 1
    - 부모 자식 관계의 중요 키 값에 대한 데이터 정합성 점검의 프로파일링 분석 방법은 참조무결성 분석 방법이다.

7. 3
    - 비즈니스 이슈, 외부고객의 소리-VOC(Voice Of Customer), 내부고객의 소리-VOB(Voice Of Business), 설문조사, 인터뷰 등의 분석을 통하여 
      오류 잠재원인을 파악하고 업무규칙을 도출하는 방법은 Outside-In 방법이다. 즉 외부로부터 문제점을 확인하고 관련 데이터 이슈를 통해 
      업무규칙을 도출하는 방법이다.

8. 4 / 2
    - 데이터의 원천을 추적할 수 있어야 되며 데이터 흐름(데이터 계보관리)을 관리해야 하는 데이터 품질 관리 영역은 데이터 활용 영역이다.
    
    
- 데이터 품질 관리 프로세스 이해 (p695)
1. 4 / 1
    - 데이터 품질 관리를 위한 프로세스는 데이터 품질 정책을 수립하고 수립된 정책에 따라 데이터 표준, 모델, 값(Value) 그리고 데이터 활용을 위한 
      프로세스 기반의 관리 절차를 말한다.

2. 4
    - 데이터 활용 관리는 데이터의 활용 여부를 점검하거나 활용도를 높이기 위해 측정 대상 데이터와 품질 지표를 선정하여 품질을 측정•분석하고 
      품질을 충족시키지 못하는 경우 원인을 분석하여 담당자 로 하여금 조치하도록 하는 작업이다. 또한 애플리케이션에서 활용되지 않는 데이터를 점검하여 
      데이터베이스의 사용 환경을 개선하고 업무적 중요도가 높은 데이터에 대한 품질의 평가와 개선으로 데이 터의 활용도를 높인다.

3. 3
    - 데이터 모델 품질 관리란 데이터 요구 사항 관리에 의해 변경되는 데이터 구조를 모델에 반영하는 작업 절차와 데이터베이스 시스템 구조와 동일하게 
      데이터 모델을 유지하도록 하는 작업 절차를 말한다.

4. 1
    - 데이터의 효과적인 확보, 품질 유지 관리를 위해 수립된 규정이나 계획, 지침 등에 포함된 데이터 관리 방향이 의미하는 것은 데이터 품질 관리 원칙이다. 
      데이터 품질 관리 정책은 데이터 품질 관리 원칙, 데이터 품질 관리 조직, 데이터 품질 관리 프로세스를 포괄하는 개념이다.

5. 3
    - 데이터 품질 관리 조직 정의 시 이에 대한 관리 기준 중 명확성은 데이터 품질 관리를 담당할 관리자가 선정되어 있고 담당자별로 수행해야 할 역할이 
      명확하게 정의되어 있어야 함을 의미한다. 문제에 명시한 설명은 누가 수행해야 하는지에 대해 명확하게 지정하고 있지 않기 때문에 명확성 측면에서 
      보완이 필요하다.

6. 4
    - 데이터 품질 관리 프로세스는 변화 관리, 프로젝트 관리 등 기존의 다른 프로세스와 상호 연관관계가 명확하게 정의되어 적용함에 있어 문제가 없어야 한다.

7. 1
    - 업무 규칙 정의는 데이터 표준 관리 프로세스와 직접적인 연관성이 가장 적다. 데이터 모델 정의 및 데이터 표준 점검은 명명규칙, 데이터 표준 등에 
      영향을 받기 때문에 데이터 표준 품질 관리 프로세 스와 직접적인 연관이 있다. 데이터 품질 관리 정책 수립 프로세스는 표준화 요구 사항 수집 시 
      데이터 품질 관리 정책에 영향을 받기 때문에 직접적인 연관이 있다고 할 수 있다.

8. 3 / 4 
    - 데이터 값(Value) 관점에서의 품질 관리는 저장된 데이터에 대한 품질을 확보하기 위한 활동으로 데이터 프로파일링, 업무규칙 등의 품질 진단을 통해 
      오류데이터를 식별하고 식별된 데이터 값(Value) 개선이 목적이다. 데이터 표준 변경은 데이터 값(Value) 관점의 품질 관리와는 관련이 없다.

9. 2
    - 데이터 표준 품질 관리는 데이터 표준화 원칙에 따라 정의된 표준 단어 사전 및 도메인 사전, 표준 용어 사전, 표준 코드, 데이터 관련 요소 표준 등을 
      기관에 적합한 형태로 정의하고 관리하는 작업을 말한다. 데이터 모델의 품질 관리와 관련이 없다.

10. 4 
    - 저장된 데이터의 값이 업무규칙 수립에 중요한 기준이 되는 것이 아니라 업무규칙에 맞게 데이터 값(Value)이 저장되고 관리되어야 한다.

## 표준 관점 데이터 품질 관리 
### 정의 및 관리 목적 
- 표준 데이터 
    - 기관이나 기업 전사차원에서 단일화하고 표준화된 정보시스템 구현
    - 데이터 불일치 or 데이터 오류를 방지 
    - 표준화되지 않은 데이터로 인해 야기되는 데이터 보정 작업 최소화 -> 정보시스템 생산성 및 가독성
    - 데이터 구조에 대한 이해도 높이고
    - 의사소통을 원활
    - 데이터 통합을 효율적으로 수행
### 세부 관리 대상
- 표준 단어 사전 
    - 표준성
        - 일반적으로 사용되는 사전적 의미 단어 
        - 지나치게 업무에 의존적, 방언 사용 x 
    - 참조 가능성
        - 새로운 업무 정의 참조 
    - 일반성
        - 일반인도 해당 단어의 의미를 이해할 수 있어야 함
    - 대표성
        - 동의어들을 대표할 수 있어야 한다. 
- 표준 도메인 사전 
    - 표준성
        - 공통적으로 사용되는 속성을 대상
    - 유일성
        - 중복 도메인 x 
    - 업무 지향성
        - 지나치게 일반화x -> 업무 특성을 충분히 반영 
- 표준 용어 사전 
    - 표준성
        - 용어 사용의 차이에 따라 발생하는 전사 차원의 혼란 최소화 
    - 일반성 
        - 지나치게 업무 관점 및 의미상 혼란x
        - 일반적인 의미와 전혀 다른 용어는 다른 용어로 대체, 새로운 용어 개발 자제
    - 업무 지향성 
        - 기업의 업무 범위 내에서 약어를 사용, 내부에서 별도로 정의하여 사용
        - 지나친 약어 사용 업무 이해도 떨어뜨리니 주의 
- 표준 코드 
    - 재사용성
        - 표준화 기구나, 정부, 공공기관에서 정의한 코드를 재사용 o 
          -> 데이터 이해력 높이고, 코드 관리 용이 
    - 일관성
        - 업무 범위 내에서 가능한 유일하게 정의 
    - 정보 분석성 
        - 가능한 범위의 데이터는 모두 코드화 하여 관리 
        - 사용자 직접 입력 최소화
- 데이터 표준 요소 
    - 통합성
        - 전사적으로 통합하여 관리 및 적용
    - 일관성
        - 정의된 표준 데이터 일관되게 적용되는지 정기적 검토 
### 표준 데이터 상관도 
### 표준 관점 품질 측정 방법
- 측정산식
    - 데이터 표준 적용률(%) = 진단대상 DB의 데이터 표준 용어 적용된 컬럼 수 / 전체 컬럼 수 X 100



## 모델 관점 데이터 품질 관리 
### 정의 및 관리 목적 
- 목록 
    - 데이터 참조모델
    - 개념 데이터 모델 
    - 논리 데이터 모델 
    - 물리 데이터 모델 
    - 메타 데이터 및 DBMS 객체 정보 
- 목적 
    - 데이터 구조에 대한 최신 정보 유지 
    - 전사 차원에서 데이터 모델의 공유와 재사용성 극대화 
    - 체계적인 데이터 모델의 변경 관리 
### 세부 관리 대상 
- 완전성
    - 단계별 데이터 모델에 모든 메타 데이터를 포함
- 일관성
    - 단어, 용어, 도메인 및 데이터 관련 요소 표준 준수 
- 추적성
    - 데이터 모델 변경 이력 추적 용이, 과거 데이터 모델 활용 요구 충족 
- 상호 연계성
    - 데이터 모델 간의 상호 연관 관계 표현
- 최신성
    - 데이터 구조의 단계별 데이터 모델, 업무 규칙, 물리 데이터와 논리적 일치 
- 호환성
    - 다른 종류의 관리 데이터와도 상호 호환 가능
### 모델 관점 품질 측정 방법
- 측정기준
    - 현행화율(%) = 100% - 불일치율(%)
    - 불일치율(%) = ((물리모델 기준 미현행화 테이블 수 / 실제 DB 기준 전체 테이블 수 X 100) + 
                    (물리모델 기준 미현행화 컬럼 수 / 실제 DB 기준 전체 칼럼 수 X 100)) / 2
