# 표준화 정의서 


**공통원칙**  
	1. 관용화된 용어를 우선하여 사용한다.    
	2. 한글명에 대해서 복수개의 영문명을 허용하지 않는다.  

**표준용어**   
	1. "일자", "일" 등 날짜를 의미하는 용어는 '일자'로 통일한다.   
    2. 용어의 길이는 한글12자, 영문 24자이내로 제한한다.   
    
**표준코드**  
	1. 코드 속성 명칭에대하여 맨 끝에 '코드'를 붙여 사용한다.   
	2. 코드는 전사모델에서 유일하게 정의한다.   
		
**표준도메인**  
	1. 표준 도메인은 기본적으로 Number, String, Datetime으로 한다.   

# 엔터티 정의서

**고객 엔터티**  
	- 개요: 중고차 거래(매입 및 판매)를 위한 고객을 관리함  
	- 성격: 키 엔터티이며, 개인이나 법인 고객이라는 실체(개체집합)을 관리  
	- 구성상의 특징: 식별자 고객번호로 구분하고, 매매 엔터티의 부모 엔터티임  
	- 특이사항: 개인과 법인이라는 서브타입으로 구성  
  
**차량 엔터티**  
	- 개요: 중고차 거래의 대상인 차량을 관리함  
	- 성격: 키엔터티이며, 차량이라는 실체(개체집합)을 관리  
	- 구성상의 특징: 식별자 차량번호로 구분하고, 매매차량과 사고현황의 자식엔터티임  
	- 특이사항: 승합차, 화물차 등 차량종류와 국산차, 수입차 등 국산수입 구분이라는 서브타입으로 구성되며,     
	            선택 속성으로 사고차량여부가 존재(사고차량여부가 Optional이라 가정 시)  
    
**보유도서 엔터티**    
	- 개요: 도서관에서 보유하고 있는 도서를 관리  
	- 성격: 개체 엔터티로 키엔터티로 분류되며 개념이 아닌 도서라는 실체를 관리  
	- 구성상의 특징: 도서 종류로 구분될 수 있으며, 대출도서 엔터티의 부모가 된다.  
	- 특이사항: 도서상태는 예약, 대출중, 대출기능의 Sub Type으로 구성  

**시장 엔터티**    
	- 개요: 도서관에서 보유하고 있는 도서를 관리  
	- 성격: 개체 엔터티로 키엔터티로 분류되며 개념이 아닌 도서라는 실체를 관리  
	- 구성상의 특징: 도서 종류로 구분될 수 있으며, 대출도서 엔터티의 부모가 된다.  
	- 특이사항: 도서상태는 예약, 대출중, 대출기능의 Sub Type으로 구성
