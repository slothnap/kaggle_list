# 데이터 요건 분석(p175)	
## 정보 요구 사항 개요(p177)	
### 정보 요구 사항(p178)
    1. 정의
        - 사용자가 일상적으로 수행하는 업무의 개선 사항이나 신규 개발 사항으로 시스템을 통해서 가능상의 목적을
          달성하기 위해 요청하는 내용
        - 현행 시스템 분석, 사용자 요구 사항 수집, 제안 요청서, 사업 수행 계획서
    2. 정보 요구 사항 생명주기 모형
        - 가. 정보 요구 사항 도출
            - 사용자 인터뷰, 설문지, 워크숍, 현행 시스템 분석
        - 나. 정보 요구 사항 분석 및 정의
            - 수집된 정보 요구 사항 정리 및 분석 정의
        - 다. 정보 요구 사항 명세화
            - 확정된 정보 요구 사항 개별로 세밀하게 분석 기록
        - 라. 정보 요구 사항 검증
            - 비즈니스 관점, 조직 관점, 애플리케이션 관점과 상관분석을 통해 누락 없이 반영 검증
    3. 정보 요구 사항 유형
        - 외부 인터페이스 요건
        - 기능 개선 요건
        - 성능 개선 요건
        - 보안 개선 요건
### 정보 요구 사항 관리(p185)
    1. 정의 및 관리 목적
        - 애플리케이션 및 시스템 전반에 걸친 사용자의 요구를 수집하고 분류하여 반영하는 작업 절차
        - 데이터, 애플리케이션, 비즈니스 등의 요구 사항을 전부 포함하는 통합 관리 프로세스를 정립
    2. 정보 요구 사항 프로세스 
        - 가. 업무 흐름 프로세스
            - 요구 사항 발송
            - 요구 사항 수렴
            - 요구 사항 검토
            - 영향도 분석
            - 공식화
            - 반영 작업 계획 수립
        - 나. 수행 조직 및 수행 업무
            - 사용자
                - 정보 요구 사항 정의 및 상세화
                - 정보 요구 사항 변경 요청
                - 정보 요구 사항 반영을 위한 미팅
                - 정보 요구 사항 반영 여부 확인
                - 미결 사항에 대한 의사결정 실시
            - 담당자
                - 사용자 정보 요구 사항 접수
                - 사용자 정보 요구 사항에 대한 기본적인 검토
                - 반영 여부 결정을 위한 사용자와 1차 미팅
                - 접수 요건에 대한ㅂ 처리 방식 및 처리 기한 결정
                - 관련 부서별 담당자 수집 및 요건 협의 주도
                - 사용자 정보 요구 사항 반영
                - 테스트 및 검증
                - 사용자 반영 결과 통보
            - 데이터 아키텍처 전문가
                - 사용자 정보 요구 사항에 대한 표준, 데이터베이스, 애플리케이션 영향도 분석 및 보고
                - 접수된 요구 사항에 대한 표준 준수 여부 체크
                - 영향도 분석을 통한 수정 및 변경 계획 수립
                - 표준 제시 및 준수 여부 검토
## 정보 요구 사항 조사(p191)	
### 정보 요구 사항 수집(p192)
    1. 정보 요구 사항 수집 형태
        - 관련 문서 수집
        - 사용자 면담을 통한 수집
        - 워크숍을 통한 수집
        - 현행 업무 처리 메뉴얼을 통한 수집
        - 현행 정보시스템 관련 산출물을 통한 수집
    2. 관련 문서 수집
        - 가. 문서 수집 목적
            - 과거 실적 자료 및 향후 계획 등의 자료를 수집한다. 
        - 나. 문서 수집 자료
            - 경영 계획에 대한 자료
            - 정보시스템에 대한 자료
            - 과거 수행한 컨설팅 보고서
            - 전산 처리 업무 매뉴얼
            - 현업 부서 업무 자료 
        - 다. 문서 수집 원칙
            - 문서는 기존에 보유하고 있는 문서를 변형하지 않고 수집하고, 정보시스템에 대한 자료는
              별도의 정리 양식을 이용하여 작성한다. 
            - 수집된 문서를 바탕으로 경영 및 정보시스템 현황에 대한 요약표를 작성하여 그 내용을 숙지
            - 수집된 문서들은 계획 수립 기간, 문서 관리자를 지정하여 운영한다. 
            - 유형별 문서를 향후 활용을 위하여 문서 분류 방식을 결정한 후에 일정한 장소에 보관한다. 
            - 수집된 문서는 통상 대외비의 성격이 강하므로 개인별로 보관하는 것을 통제하고 문서 보안 관리
    3. 사용자 면담
        - 가. 사용자 면담 종류
            - 면담조사 방법
                - 개인 면담
                - 집단 면담
                - 표준화 면담
                - 비표준화 면담
            - 면담조사 방식
                - 피라미드 구조
                    - 구제적이며 선택전인 질문으로 시작해서 일반적이고도 서술적인 질문으로 전개
                - 퍼널 구조
                    - 서술적인 질문으로 시작하여 점차 구체적이고 선택적인 질문으로 시작해서 일반적이고도
                      서술적인 질문으로 전개해 나가는 방식
                - 다이아몬드 구조
                    - 피라미드 형태의 진행이 이루어지고 난 후, 다시금 퍼널 형태의 진행으로 마무리
        - 나. 사용자 면담 진행
            - 1) 계획 및 준비 
                - 가) 면담 주제 선정
                    - 면담 요지 미리 준비
                - 나) 면담 진행 팀 구성
                    - 면담자
                        - 면담 대상자에게 질문
                    - 기록자
                        - 면담 답변 내용 기록
                    - 관찰자 
                        - 주의를 환기
                        - 보충 질문
                        - 면담 종료 판단
                - 다) 면담 대상자 선정
                    - 업무의 명확한 이해 가능자 선정
                    - 프로젝트 후원자나 사용자측에서 추천 받는 것도 좋다
                    - 두 명 이상의 면담 대상자를 선정
                - 라) 면담 일정 수립
                    - 프로젝트 목적과 범위 통보, 문저 자료 요청
                    - 면담 시간 1.5 ~ 3시간
                - 마) 면담 준비 
                    - 면담 대상자 담당 업무 검토 필요
                    - 면담 시나리오 작성, 면담 대상자에게 배포 X
                    - 면담 진행 순서 준비
            - 2) 면담
                - 가) 면담 시작
                    - 진행 순서 및 역할 확인
                    - 모든 절차 및 면담 지침에 세세한 문구까지 모두 반영
                - 나) 면담 주제 토의
                    - 모든 면담 내용 기록
                    - 대화 흐름 안끊기게 
                    - 여담 내용도 반드시 기록
            - 3) 면담 결과 분석
                - 개인적 의견을 고려 면담 결과 정리
                - 의문 및 추가 사항 면담 대상자에게 확인, 추가 면담 가능
            - 4) 분석 결과 피드백
                - 면담 내용 면담 대상자에게 확인 가능
        - 다. 면담 수행 시 고려 사항
            - 면담 시간 준수
            - 비밀 보장
            - 기대 수준 설정
            - 면담 범위 준수
            - 적절한 대상자 선정
            - 응답 유도
            - 면담 내용 문서화
            - 잘못된 선입견의 배제
    4. 워크숍
        - 가. 워크숍 개요 및 목적
            - 경영층 또는 현업 부서장의 공통된 의견을 도출
            - 유사한 업무 또는 관련된 업무 등을 수행하는 부서에 대한 면담 노력 절감
            - 전문가들의 판단력을 이용하여 최적의 결론 도출
        - 나. 워크숍 준비
        - 다. 워크숍의 수행
            - 1) 워크숍 개시
            - 2) 워크숍 수행 준비
            - 3) 워크숍 수행
            - 4) 워크숍 종료
    5. 현행 업무 조사서
        - 업무 조사서 양식은 단순하고 이해가 쉬움
    6. 현행 프로그램 - 데이터 관련 문서
        - 현행 시스템 프로세스의 구조
        - 현행 시스템의 데이터 분석
        - 현행 데이터 저장소의 구조
    7. 관찰
    8. 브레인스토밍
        - 가. 브레인스토밍 개요 및 목적
        - 나. 브레인스토밍 4가지 기본 원칙
            - 양에 포커스를 참추기
            - 비판, 비난 자제
            - 특이한 아이디어 환영
            - 아이디어 조합 및 개선
        - 다. 진행방법
    9. 프로토타이핑
        - 1단계: 기본적인 사용자 요구 사항 분석
        - 2단계: 앞으로 개발될 시스템의 가장핵심적인 기능 위주로 개발
        - 3단계: 프로토타입의 보완을 위한 여러 가지 제안
        - 4단계: 프로토타입의 수정과 보완
### 정보 요구 사항 정리(p206)
    1. 정보 요구 사항 정리
        - 가. 사용자 면담 정리 
        - 나. 업무 조사서 정리
        - 다. 워크숍 정리
            - 워크숍의 목적
            - 워크숍 진행 내용
            - 해결 과제에 대한 상태
            - 기타 특이 사항
        - 라. 기타 기법 정리
    2. 정보 요구 우선순위 분석
        - 가. 화폐가치 산출 방법
            - 최종적으로 구해진 가치가 높을수록 우선순위가 높다
            - 정보 요구 사항 기업 차원의 중요성 1 ~ 3점
            - 정보 요구 사항 시스템 차원의 중요성 1 ~ 3점
            - 정보 요구 사항이 다른 정보 요구 사항에 도움 평가 1 ~ 5점
            - 세가지 점수 곱한다. 
        - 나. 상대적 중요도 산정 방법
            - 정보 요구 사항이 업무에 기여하는 수준 1 ~ 5점
                - 목적 5점, 목표 4점, 전략 3점
            - 정보 요구 사항 매트릭스를 작성
                - 가상 관련이 큰 정보 요구 사항 9점
            - 현재 정보시스템이 정보 요구 사항을 얼마나 충족 1 ~ 3점
                - 만족 3점, 보통 2점, 지원x 1점
### 정보 요구 사항 통합(p211)
    1. 정보 요구 사항 목록 검토
    2. 정보 요구 사항 목록 통합/분할
        - 가. 동일 부서 내 중복 요구 사항 검토
            - 부서 내 정보 요구 사항 목록을 작성
            - 정보 요구 사항 제목을 기준으로 부서 내 동일 요건의 요구 사항이 존재하는지 파악
            - 정보 요구 사항의 세부 요청 내용을 기준 세밀하게 중복 여부 파악
            - 부서 내 동일 요건 도출 -> 정보 요구 번호를 '비고'란에 기입
            - 동일 부서 내 중복성 배재 요구 사항 목록을 완성
        - 나. 서로 다른 부서 간 중복 요구 사항 검토
            - 부서 내 정리된 정보 요구 사항 목록과 부서 간 동일 정보 요구 사항 파악
            - 정보 요구 사항 세밀 검토 중복 여부 파악
            - 부서 간 동일 요건 도출 -> 정보 요구 번호를 기입
            - 전사 관점의 검토된 정보 요구 사항 목록 작성
## 정보 요구 사항 분석(p217)
### 분석 대상 정의(p218)
    1. 현행 업무 분석 대상 정의
        - 가. 분석 대상 자료
            - 현행 업무 흐름도
            - 현행 업무 설명서
            - 현행 업무 분장 기술서
        - 나. 분석 대상 업무 영역 선정
    2. 현행 시스템 분석 대상 정의
        - 가. 분석 대상 현행 시스템 선정
            - 업무 영역,현행 시스템 매트릭스
        - 나. 분석 대상 현행 시스템 관련 자료 
            - 현행 시스템 수집 문서
                - 현행 시스템 구성도
                - 현행 시스템의 분석, 설계 및 개발 보고서
                - 화면, 장표 및 보고서 레이아웃
                - 현행 시스템 테이블 목록 및 테이블 정의서
                - 프로그램 목록
                - 사용자 및 운영자 지침서
                - 시스템 지원 및 유지 보수 이력
                - 시스템 개선 요구 사항 등
            - 수집 문서 평가 
                - 유용성: 문서의 활용 가능성 여부
                - 완전성: 문서의 내용에 누락된ㅂ 부분이 없는지 여부
                - 정확성: 문서의 내용이 현재의 시스템과 일치하는지 여부
                - 유효성: 문서가 최신의 내용을 반영하고 있는지 여부
        - 다. 추가적인 분석 대상
            - 사용자 뷰: 화면, 수작업 파일, 수작업 전산 양식, 보고서 등
### 정보 요구 사항 상세화(p222)

            - 산출된 정보 항목 중 기본 로직을 정리
            - 표준화 과정을 통하여 검토 후 사용자의 정보 항목 목록을 정의 
        - 나. 수행 작업 내용
            - 프로세스 분해, 상세화
                - 단위 업무 기능별 하향식으로 프로세스를 분해 및 도출
                - 프로세스 계층도 및 프로세스 정의서를 작성
