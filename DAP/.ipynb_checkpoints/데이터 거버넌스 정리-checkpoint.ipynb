### 데이터거버넌스
- 기업이 데이터를 체계적으로 생산-관리하고, 해당 데이터를 사용자가 효과적으로 사용-활용할 수 있도록 하기 위한 일련의 업무

### 데이터 거버넌스 프레임워크
- 데이터 구조관리 (Data Architecture Management)
- 데이터 모델링과 디자인 (Data Modeling and Design)
- 데이터 저장관리 (Data Storage Management)
- 데이터 보안관리 (Data Security Management)
- 데이터 통합과 상호운영관리 (Data Integration and Interoperability)
- 콘텐츠 관리 (Content Management)
- 마스터 데이터 관리 (Master Data Management)
- 비즈니스 인텔리전스 관리 (Business Intelligence Management)
- 메타 데이터 관리 (Meta-data Management)
- 데이터 품질 관리 (Data Quality Management)

### 빅데이터 거버넌스 아키텍처 


### 빅데이터 거버넌스 체계 실현을 위한 고려사항
- Organization
    - Data Scientist, 유형별 Stewardship(관리) 지정
    - 데이터 소유자/권하자에 대한 명확한 정의(외부수집 데이터, 증간가공 데이터 등 수명주기/생태계)
- Data Quality
    - 실시간/속도 보장이 중요한 경우도 고려
    - 반구조화/비구조화 데이터 품질 보장 한계 고려
    - 외부수집 데이터 생성주체, 데이터 수집 방법론, 다른 데이터와의 상관관계 고려
    - 데이터/정보의 진위여부 등 신뢰성 확보
- Meta data Management
    - 다양성, 복잡성 등 빅데이터 특성 반영
    - 데이터 유형 및 출처 등 추가 메타정보 관리
- Privacy
    - 정형/비정형 데이터 내 민감한 정보에 대한 제도적, 기술적 안전장치 마련
- Biz. Process Integration
    - 빅데이터 분석결과와 비즈니스 최적화간 연계
    - 비즈니스 이익 및 규재준수 고려
    - 데이터 관리 비용 최적화
- Master data Management
    - MDM(Master Data Management) 과의 연계를 통한 빅데이터 분석 Insight 가치 향상
- Life cycle Management
    - Data Lineage (데이터 계보관리)
    - 대규모 데이터 보관 및 폐기정책 수립 및 실행
