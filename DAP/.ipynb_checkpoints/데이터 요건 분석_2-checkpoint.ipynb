# 정보 요구 검증

## 정보 요구 사항 상관분석 기법
- **주제별 분류**    
    - 요구 사항 분석가 수행   
    - 품질보증팀 수행   
    - 외부 감리 수행   
- **정보 요구 / 애플리케이션 상관분석**   
    - 생성				== 'C'   
    - 조회				== 'R'   
    - 수정				== 'U'   
    - 삭제 				== 'D'   
- **정보 요구 / 업무 기능 상관분석**   
    - 생성, 수정 삭제 	== 'C'   
    - 검색 				== 'U'   
- **정보 요구 / 조직 기능 상관분석**   
    - 생성, 수정, 삭제 	== 'C'   
    - 검색 				== 'U'  

## 추가 및 삭제 정보 요구 사항 도출 
- **정보 요구 / 애플리케이션 상관 분석**    
    - 애플리케이션 충족도 분석 매트릭스    
    - 매트릭스 분석   
        - 조회, 수정, 삭제 액션 중 하나가 발생해야 한다.   
        - 생성, 삭제하는 합은 7개를 초과하지 않는다.    
        - 조회 전용을 제외한 다른 프로세스는 생성, 수정, 삭제 하나를 수행   
- **정보 요구 / 업무 기능 상관분석**   
    - 매트릭스 분석   
        - 적어도 한번 이상의 'C'를 갖는가?
