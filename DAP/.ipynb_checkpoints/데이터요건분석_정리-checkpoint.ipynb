## 정보 요구 사항
### 정보 요구 사항
- 정의
    - 사용자가 일상적으로 수행하는 업무의 개선 사항이나 신규 개발 사항
    - 시스템을 통해 기능상의 목적을 달성하기 위해 요청하는 내용
- 수집
    - 현행 시스템 분석
    - 사용자 요구 사항 수집
    - 제안 요청서
    - 사업 수행 계획서
- 요구 사항 유형 분류
    - 비즈니스 요구 사항
    - 이해관계자 요구 사항
    - 해결책 요구 사항 
        - 기능적 요구 사항
        - 비기능적 요구 사항
    - 전환 요구 사항
    - 프로젝트 요구 사항
    - 품질 요구 사항
- 프로젝트 실패 원인 분류
    - 계획과 추정 요인
    - 실행 요인
    - 인간적 요인
### 정보 요구 사항 생명주기 모형
- 정보 요구 사항 도출         == 사용자 인터뷰·설문지·워크숍·현행 시스템 분석
- 정보 요구 사항 분석 및 정의 == 수집된 정보 요구 사항 정리, 분석 및 정의
- 정보 요구 사항 명세화       == 확정된 정보 요구 사항 기록 
- 정보 요구 사항 검증         == 사용자 정보 요구 사항 반영 검증 
### 정보 요구 사항 유형 
- 사용자 요구 사항 
    - 기능
        - 전체 및 개별 기능 요구 사항
    - 비기능
        - 시스템 동작 및 성능 목표 성과 필요 사항
        - 프로젝트 관리 및 지원상황
    - 기타 요구 사항
        - 유지관리 사항
        - 컨설팅
        - 공사
- 정보 요구 사항 유형 
    - 외부 인터페이스 요건
    - 기능 개선 요건
    - 성능 개선 요건
    - 보안 개선 요건
