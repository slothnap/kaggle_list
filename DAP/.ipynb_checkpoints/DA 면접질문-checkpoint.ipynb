1. 점이력, 선분이력 특징 
- 점이력: 특정 시점의 데이터 상태를 알 수 있다. 
- 선분이력: 특정 시점의 데이터 상태가 아닌 일정 기간 동안의 데이터의 상태를 관리 



2. 본질 식별자와 인조 식별자 특징 및 차이점에 대해 설명해주세요.  
- 본질 식별자: 비즈니스 프로세스에서 만들어지는 식별자
- 인조 식별자: 인위적을 만들어지는 식별자 



3. 단일코드와 목록성 코드의 특징 및 차이점
- 단일 코드  : 하나의 코드로 Key가 구성되며 코드와 코드명만 관리 
- 목록성 코드: 하나 이상의 코드로 Key가 구성되며 코드와 코드명 이외의 부가적인 속성도 관리 



4. 데이터 표준화 기대 효과에 대해 말씀해주십시오. 
- 명칭통일에 따른 의사소통 증대
- 데이터 소재 파악에 따른 시간감소
- 데이터 형식 규칙 적용으로 인한 품질 향상
- 데이터 변환, 정제 비용 감소
