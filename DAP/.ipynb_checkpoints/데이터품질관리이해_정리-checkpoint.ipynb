## 표준 관점 데이터 품질 관리 
### 정의 및 관리 목적 
- 표준 데이터 
    - 기관이나 기업 전사차원에서 단일화하고 표준화된 정보시스템 구현
    - 데이터 불일치 or 데이터 오류를 방지 
    - 표준화되지 않은 데이터로 인해 야기되는 데이터 보정 작업 최소화 -> 정보시스템 생산성 및 가독성
    - 데이터 구조에 대한 이해도 높이고
    - 의사소통을 원활
    - 데이터 통합을 효율적으로 수행
### 세부 관리 대상
- 표준 단어 사전 
    - 표준성
        - 일반적으로 사용되는 사전적 의미 단어 
        - 지나치게 업무에 의존적, 방언 사용 x 
    - 참조 가능성
        - 새로운 업무 정의 참조 
    - 일반성
        - 일반인도 해당 단어의 의미를 이해할 수 있어야 함
    - 대표성
        - 동의어들을 대표할 수 있어야 한다. 
- 표준 도메인 사전 
    - 표준성
        - 공통적으로 사용되는 속성을 대상
    - 유일성
        - 중복 도메인 x 
    - 업무 지향성
        - 지나치게 일반화x -> 업무 특성을 충분히 반영 
- 표준 용어 사전 
    - 표준성
        - 용어 사용의 차이에 따라 발생하는 전사 차원의 혼란 최소화 
    - 일반성 
        - 지나치게 업무 관점 및 의미상 혼란x
        - 일반적인 의미와 전혀 다른 용어는 다른 용어로 대체, 새로운 용어 개발 자제
    - 업무 지향성 
        - 기업의 업무 범위 내에서 약어를 사용, 내부에서 별도로 정의하여 사용
        - 지나친 약어 사용 업무 이해도 떨어뜨리니 주의 
- 표준 코드 
    - 재사용성
        - 표준화 기구나, 정부, 공공기관에서 정의한 코드를 재사용 o 
          -> 데이터 이해력 높이고, 코드 관리 용이 
    - 일관성
        - 업무 범위 내에서 가능한 유일하게 정의 
    - 정보 분석성 
        - 가능한 범위의 데이터는 모두 코드화 하여 관리 
        - 사용자 직접 입력 최소화
- 데이터 표준 요소 
    - 통합성
        - 전사적으로 통합하여 관리 및 적용
    - 일관성
        - 정의된 표준 데이터 일관되게 적용되는지 정기적 검토 
### 표준 데이터 상관도 
### 표준 관점 품질 측정 방법
- 측정산식
    - 데이터 표준 적용률(%) = 진단대상 DB의 데이터 표준 용어 적용된 컬럼 수 / 전체 컬럼 수 X 100



## 모델 관점 데이터 품질 관리 
### 정의 및 관리 목적 
- 목록 
    - 데이터 참조모델
    - 개념 데이터 모델 
    - 논리 데이터 모델 
    - 물리 데이터 모델 
    - 메타 데이터 및 DBMS 객체 정보 
- 목적 
    - 데이터 구조에 대한 최신 정보 유지 
    - 전사 차원에서 데이터 모델의 공유와 재사용성 극대화 
    - 체계적인 데이터 모델의 변경 관리 
### 세부 관리 대상 
- 완전성
    - 단계별 데이터 모델에 모든 메타 데이터를 포함
- 일관성
    - 단어, 용어, 도메인 및 데이터 관련 요소 표준 준수 
- 추적성
    - 데이터 모델 변경 이력 추적 용이, 과거 데이터 모델 활용 요구 충족 
- 상호 연계성
    - 데이터 모델 간의 상호 연관 관계 표현
- 최신성
    - 데이터 구조의 단계별 데이터 모델, 업무 규칙, 물리 데이터와 논리적 일치 
- 호환성
    - 다른 종류의 관리 데이터와도 상호 호환 가능
### 모델 관점 품질 측정 방법
- 측정기준
    - 현행화율(%) = 100% - 불일치율(%)
    - 불일치율(%) = ((물리모델 기준 미현행화 테이블 수 / 실제 DB 기준 전체 테이블 수 X 100) + 
                    (물리모델 기준 미현행화 컬럼 수 / 실제 DB 기준 전체 칼럼 수 X 100)) / 2
## 값(Value) 관점 데이터 품질 관리 
### 정의 및 관리 목적
- DBMS 상시적인 데이터 품질 측정/개선 활동
- 데이터 품질 기준, 프로파일링, 업무규칙, 품질 측정, 품질 개선 
### 세부 관리 대상 
- 품질 기준 
    - 품질 관리 대상
        - 품질 측정 대상 시스템 및 테이블 관리 
    - 데이터 품질 지표 
        - 완전성, 유효성, 일관성, 정확성
    - 핵심 정보 항목 
        - 가장 중요한 데이터 정보 항목(고객연락처정보, 매출정보)
- 프로파일링
    - 컬럼 분석
    - 패턴 분석 
    - 날짜 분석 
    - 코드 분석 
    - 참조무결성 분석 
- 업무규칙
    - Inside-Out 도출 방법
        - 데이터 통계적 분석 -> 오류데이터 잠재원인 파악, 업무 규칙 도출 
    - Outside-In 도출 방법
        - 비즈니스 이슈, 외부고객 소리-VOC, 내부고객의 소리-VOB, 설문조사, 인터뷰
    - 근거규정 도출 방법
        - 기관 법령 기초 데이터 모델 기준 -> 실제 운영 시스템 데이터 품질 검증을 위한 업무 규칙 생성
- 품질 측정
    - 프로파일링, 업무규칙 등을 지속적, 정기적으로 수행 -> 데이터베이스 성능과 측정 주기 설정 
- 품질 개선 
    - 오류추정 데이터 개선 담당자, 오류 원인, 개선내용 지속적 관리 절차 및 방법을 정의 
### 값 관점 품질 측정 방법
- 측정기준
    - 오류율(%) = 오류 데이터 건수 / 진단대상DB 전체 데이터 건수 X 100



## 활용 관점 데이터 품질 관리 
### 정의 및 관리 목적 
- 정의
    - 양질의 데이터를 추적하여 정보계 및 빅데이터로 활용을 극대화 하기 -> 데이터 원천 추적 필요 
- 목적
    - 데이터 활용 여부
    - 데이터 흐름
    - 데이터 품질 체계적 관리 
### 세부 관리 대상
- 데이터 흐름 관리 
    - 안전성
        - 데이터 매핑 규칙 
    - 유효성
        - 데이터 클린징 규칙 
    - 데이터 정합성
        - 매핑 규칙 준수 -> 데이터 정합성

# 데이터 품질 관리 프로세스 이해
## 데이터 품질 관리 정책 
### 정의 및 관리 목적 
- 정의 
    - 데이터 확보 계획과 데이터의 효과적인 운영 관리 체계 및 계획 정의 
- 목적
    - 비전과 목표에 맞는 데이터 확보
    - 안정적으로 서비스 기본원칙과 관리 체계 구축 
    - 체계적이고 일관성 있는 데이터 관리 
### 세부 관리 대상 
- **데이터 품질 관리 정책** 
    - **데이터 품질 관리 원칙** 
    - **데이터 품질 관리 조직**
    - **데이터 품질 관리 프로세스** 
- **데이터 품질 관리 원칙**
    - 정의
        - **데이터 효과적인 확보, 품질 유지 관리, 계획 지침 등** 
    - 준수성
        - 데이터 관리 달성할 수 있게 정의 
    - 불가변성
        - 불가피한 경우 제외 쉽게 바뀌지 x 
    - 이해성
        - 원칙은 쉽게 이해, 의미 불분명 혼란 최소화 
    - 완전성
        - 정책 수립 필요한 모든 사항 정의
    - 일관성 
        - 원칙 간 충돌 없도록, 충돌시 분명한 의사결정 할 수 있게 명시 
- 데이터 품질 관리 프로세스 
    - 준수성
        - 데이터 관리 원칙에 맞게 정의 
    - 완전성
        - 데이터와 관련된 모든 요소가 빠짐없이 관리 
    - 상호 운용성
        - 기존 다른 프로세스와 상호 연관 관계 명확하게 정의 
- 데이터 품질 관리 조직 
    - **명확성**
        - 담당자 선정 -> 수행 역할 명확하게 정의
    - **운영성**
        - 교육 프로그램 + 필요한 능력 갖춰야 한다. 
### 데이터 품질 관리 정책 수립 
- 데이터 품질 관리 정책 수립 프로세스 
    - 데이터 품질 관리 정책 수립
    - 데이터 품질 관리 정책 검토 
    - 데이터 품질 관리 정책 승인
    - 데이터 품질 관리 정책 공표
    
    데이터의 효과적인 확보, 품질 유지 관리를 위해 수립된 규정이나 계획, 지침 등에 포함된                        데이터 관리 방향이 의미하는 것은 데이터 품질 관리 원칙이다. 데이터 품질 관리 정책은                        데이터 품질 관리 원칙, 데이터 품질 관리 조직, 데이터 품질 관리 프로세스를 포괄하는 개념이다.
