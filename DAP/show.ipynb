1. 프로젝트 개요
- 프로젝트명
	- 입출금 시스템 개편 
- 목적
	- 입/출금 시스템의 거래소 분리를 위한 1차 목표인 블록체인 데이터 서비스 프로토타입 구현을 통해
	  기존 거래소 시스템에 종속되지 않는 독립적인 입/출금 시스템 설계 방향성 제시

1.1 프로젝트 배경
- 기존 거래소 내 구축된 시스템은 거래소 기능과 입출금 시스템 기능에 대한 명확한 분리가 되지 않은 상황임
	- 입/출금 시스템 기능에 대한 명확한 분리 및 역할 구분 필요
	- 관리 주체가 명확하지 않아 유지 보수 및 개선의 어려움 존재
	- 신규 상장 시 거래소 내 소스코드 및 DB에 대한 변경 작업 발생으로 인한 시스템 영향도 문제 발생
	- 크론탭을 사용한 스케쥴링 기반의 데이터 처리로 인한 기능적 한계 존재
- 당사 거래소의 시스템 개선 및 경쟁력 향상을 위해서 거래소 기능과 입출금 시스템 기능의 분리가 필요함
	- 각각의 기능에 대한 명확한 R&R 할당을 통해서 업무 효율성 및 생산성 증대 가능
	- 거래소 분리 시 입출금 서비스의 성능 향상 뿐만 아니라 향후 사업 다각화를 통한 신규 Biz 수립 가능
- 거래소 분리를 위해서는 전체 내부 시스템의 아키텍처 전환이 필요하기 때문에 단계적으로 개발 및 전환이 필요함
	- 블록체인 네트워크와의 통신하는 데이터 처리 시스템 구축을 시작으로 신규 입출금 시스템 개발, 관련 인터페이스 API 개발 등 점진적인 개발이 필요로 함
	- 개발 뿐만 아니라 전체적인 전환을 위한 신규 정책 설계, 사용자 (관리자) 업무 환경 분석 및 개선 등의 작업들이 필요로 함

1.2 프로젝트 목적 및 기대효과

- 집금 처리: 코인 내부 이동(집금)을 위한 기능 제공(내부 작업자에 의해 발생)
- 트랜잭션 관리: 서명된 트랜잭션 데이터를 관리 및 블록체인 노드로 요청
- 기준 정보 관리: 코인, 주소, 회원, 블록체인 노드 등의 입출금 시스템 관리를 위한 모든 기준 정보 관리
- 노티 서비스: 슬랙, 텔레그램, 문자 등의 알람 서비스
- 카프카 클러스터: 입출금 시스템 내 콘솔망을 통한 이벤트 데이터 처리
- 외부 API: 거래소 연동을 위한 API 서비스
	- 이벤트 기반: 거래소측에서 입출금 시스템 이벤트에 대한 실시간 데이터 처리가 필요한 경우
	- REST API: 거래소에서 조회가 필요한 경우
- 관리 페이지: 기존 B Manage 화면과 유사한 백오피스 기능 제공
