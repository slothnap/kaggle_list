- str (suspicious transaction report)
    - 의심거래보고제도
    - 금융거래와 관련하여 수수한 재산이 불법재산이라고 의심되는 합당한 근거가 있거나 금융거래의 상대방이
      자금세탁행위를 하고 있다고 의심되는 합당한 근거가 있는 경우 이를 금융정보분석원장에게 보고
    - 구성
        - str 보고서 생성, str 보고, str보고 모니터링
- ctr (currency transaction report)
    - 고객현금거래보고제도
    - 1천만원 이상 거래 주목, 하루 1천만원 이상 현금 거래 시 금융정보분석원에 보고
- kyc (know your customer)
    - 고객확인의무
    - 자금세탁방지를 위해 필요한 정보 수준
    - CDD (customer due diligence)
        - 금융회사가 고객과 거래 시 고객의 성명, 주소, 연락처를 확인하고
          자금 세탁 행위 등의 우려가 있는 경우 실제 당사자 여부 및 금융거래 목적을 확인하는 제도
    - EDD (enhanced due diligence)
        - 고객 개인별, 금융 상품별 자금세탁 위험도에 따라 위험도가 높다고 판단될 시
          직업, 거래목적, 거래 자금의 원천, 그 외에도 우려를 해소하기 위해 판단해야하는 자료들까지
        함께 검토하는 '강화된 고객확인의무'

- 업무
    - 전사데이터 관리체계 정의
    - 데이터 표준 정의 및 프로세스 수립
        - 단어
            - 테이블과 컬럼의 한글명 및 영문명을 구성하는 최소한의 원소
            - 용어는 반드시 이러한 단어의 조합으로 이루어 진다. 
            - 명사형으로 정의하며 동사, 접속사, 복수 표시 또는 소유격형태의 단어는 사용하지 않는다. 
            - 예시
                - 단일어 == 사용
                - 합성어 == 미 + 사용
                - 분류어 == 여부
            - 명명규칙
                - 약어 최대 4자리
                - 단어물리명 숫자시작허용 여부
        - 도메인
            - 번호 도메인
                - 식별번호
                    - 엔터티에서 독립적으로 식별자 역할을 하는 속성을 정의하기 위한 도메인으로  
                      자기자신만의 도메인을 갖는 도메인 (단, 코드 도메인 제외)
                - 일반번호
                    - 식별번호 도메인을 제외한 각 자리별 특정 의미를 가지거나 체계를 가지고 관리  
                      되어야 하는 속성을 정의하기 위한 도메인
            - 제한형 도메인 
                - 데이터 값을 제한하는 도메인 (데이터의 범위, 데이터의 타입, 데이터의 값 등)
                - 금액
                    - 화폐의 가치를 수로 표현하기 위한 도메인
                - 건수
                    - 객체의 개수나 량을 수로써 표현하기 위한 도메인
                - 율 
                    - 비율을 수로 표현하기 위한 도메인
                - 일자
                    - 특정 사건이 일어난 시점 또는 시점과 시점간의 시간을 표현하기 위한 도메인
                - 내용 
                    - 서술 형식으로 상세 내용을 표현한 것으로 자유 형식의 텍스트를 표현하기 위한 도메인
            - 코드 도메인
                - 코드화 하여 데이터를 분류, 구분하기 위한 속성을 정의하기 위한 도메인
                - 단일코드
                    - 하나의 코드로 Key가 구성되며 코드와 코드명만 관리한다. 
                        - ex) 원장구분코드 -> 1:출금, 2: 입금, 3: 가수금
                - 목록성코드
                    - 하나 이상의 코드로 Key가 구성되며 코드와 코드명 이외의 부가적인 속성도 관리한다.
                        - ex) 우편주소코드 -> 우편번호, 시도, 구군, 동, 리, 번지 등
                - 여부코드
                    - 'Y'와 'N' 또는 '0'과 '1'의 두가지 인스턴스만 관리한다. 
                    - 'Y'값이 긍정적으로 표시할 수 있게 명명함
        - 용어
            - 반드시 단어사전에 등록된 단어를 사용하고 분류어로 끝나야 한다. 
            - 한글 30자 이내로 지정하고, 영문변환시 축약하여 25자 이내로 생성한다. 
            - 영문변환시 띄어쓰기를 허용하지 않으며 Under Bar (_)를 사용하여 연결한다. 
            - 동일한 의미를 가지는 용어는 하나만 등록하여 사용한다. 
            - 단일어/합성어 + ... + 단일어/합성어 + 분류어
            - 명명규칙
                - 용어 최대 30자리
        - 분류어
            - 분류어는 칼럼을 분류하는 속성분류어(도메인)와 테이블을 분류하는 엔터티분류어로 관리한다. 
            - 분류어는 데이터의 특성/ 엔터티에 특성을 그룹화 하여 관리한다. 
            - 속성분류어
                - 속성분류어(도메인)는 특정 값을 제한함으로써 값에 대한 표준화와 명명규칙의 표준화를 준수한다.
                - 속성분류어(도메인)는 인포타입과 결합하여 관리한다. 
                    - 도메인 + 인포타입 == 고객명 + VARCHAR2(40) == 고객명_V40
                                        == 금액   + NUMBER(22)   == 금액N22
            - 엔터티분류어
                - 기본 (BASE)
                    - 현재나 최종의 상태를 표시하는 기본적인 정보를 담은 엔터티타입으로 각 주제영역의 대표성 엔터티 기본과 이력이 혼합되어 있는 경우에는 기본을 사용한다.
                    - ex) 국가기본, 회원기본
                - 이력 (LOG)
                    - 기본 엔터티가 존재하고 별도로 과거 정보를 시점별로 보관하는 엔터티인 경우, 즉 정보의 변경사항을 관리하기 위한 목적으로 생성하는 엔터티
                    - ex) 연락처이력
                - 상세 (DTL)
                    - 기본 상위엔터티가 존재하고 기본 엔터티에 대하여 일대다의 형식으로 상세 정보를 관리하는 엔터티
                    - ex) 업체상세
                - 내역 (LIST)
                    - 작업 또는 거래시점에 생성되고, 동일 형식의 정보가 낱개 형태로 변경/추가되는 엔터티(transaction 성격의 엔터티)
                    - 거래내역, 거래상세내역, 입금내역
                - 관계 (REL)
                    - 엔터티와 엔터티 사이의 관계를 정의하기 위하여 생성하는 엔터티
                    - ex) 회원관계자관계
                - 집계 (AGRG)
                    - 이미 존재하는 엔터티로부터 데이터 값을 추출하여 통계, 합계 등을 산출하기 위한 속성으로 구성된 엔터티 타입
                - 코드 (CD)
                    - 코드 정보를 관리하는 엔터티타입
                - 채번 (SEQN)
                    - 일련번호 채번용 엔터티
                - 백업 (BKUP)
                    - 데이터의 보관을 위하여 별도로 저장하는 엔터티
                - 임시 (TEMP)
                    - 임시로 사용하는 엔터티
                - 이행 (MIG)
                    - 데이터 이행용 엔터티
                - 스냅샷 (SNST)
                    - 정기적 또는 부정기적으로 특정 시점의 데이터의 상태를 보관하기 위한 엔터티,  
                      '이력'과의 차이는 거래와 관계 없이 전신상의 목적으로 관리하는 엔터티
                - 차원 (DMSN)
                    - 데이터웨어하우스를 설계하기위해 사용되는 다차원모델링의 차원 엔터티
                - 팩트 (FACT)
                    - 데이터웨어하우스를 설계하기위해 사용되는 다차원모델링의 차원 엔터티
        - 코드
            - 여부
                - 그러함과 그러하지 아니 (확인^여부)
            - 구분
                - 따로따로 갈라서 나눔 (계약^구분^코드)
            - 상태
                - 사물현상이 놓여 있는 모양이나 형편 (계좌^상태^코드)
            - 종류
                - 사물의 부문을 나누는 갈래 (카드^종류^코드)
            - 유형
                - 어떤 비슷한 것들의 본질을 개체로서 나타낸것, 또는 그것들의 공통되는 성질이나 모양 (평가^유형^코드)
            - 분류
                - 사물을 공통되는 성질에 따라 종류별로 나눔 (업종^분류^코드)
            - 기타
                - 정의된 유형에 속하지 않고 개별적으로 관리되는 경우 (상품^코드)
    - 업무에 따른 데이터 모델링 수행 및 검토
        - 개발자 데이터 모델링 미팅 
            - 신규 서비스 파악 
            - 논리모델 작성
            - 테이블 영향도 파악 
        - 표준화 수행
            - 표준등록(단어, 도메인, 용어, 코드)
            - 논리모델 표준화 수행
        - 논리모델 배포
        - 물리모델 생성
        - 테이블 스크립트 추출
        - 스크립트 검토(DBA)
        - 테이블 생성(DBA)
