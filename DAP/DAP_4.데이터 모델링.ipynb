# 데이터 모델링(p319)
## 데이터 모델링 이해(p321)
### 데이터 모델링 개요(p322)
    1. 데이터 모델링 정의
        - 가. 데이터 모델링 탄생 배경
        - 나. 정보시스템 개발 방법론
            - 정보 공학 방법론
                - 정보 전략 계획 (ISP)
                - 업무 영역 분석 (BAA)
                - 업무 시스템 설계 (BSD)
                - 구축 (CON)
            - 객체지향 컴포넌트 기반 방법론
    2. 데이터 모델링의 중요성
        - 데이터 모델링이란 조직의 업무 수행에 필요한 데이터베이스를 생성하기 위하여 업무에서 필요로 하는
          정보의 구조와 규칙을 명확하게 표현하기 위한 지적인 작업 과정
### 데이터 모델링 기법 이해(p332)
    1. 엔터티-관계 데이터 모델
        - 가. 추상화
            - 추상화란 현실세계에 존재하는 구체적 개체(Entity) 또는 객체(Object)들의 공통된 특징
            - 유형화
                - 현실세계에 존재하는 같은 성질을 갖는 멤버들을 타입 또는 유형(Class)이라는 하나의 개념으로
                  정의하는 데 사용된다. 
            - 집단화
                - 값(Value)을 유형화하면 속성이라는 타입이 된다. 
            - 일반화
                - 여러 엔터티 타입 간의 공통적인 특징을 파악하는 과정
        - 나. 타입과 인스턴스
            - 타입: 실제 발생된 사례들을 추상화한 집합
            - 인스턴스: 현실 세계에서 실제로 발생된ㅂ 사례
    2. 엔터티-관계 데이터 모델 구성요소
        - 가. 엔터티
            - 조직의 업무를 수행하는데 필요로 하는 사물, 사건 또는 개념을 나타내는 어떤 것
        - 나. 관계
            - 하나 또는 두 개의 엔터티에서 인스턴스를 연관시키는 업무적인 이유
        - 다. 속성과 도메인
            - 속성은 엔터티에 저장되는 인스턴스들의 특성을 설명하는 항목 
        - 라. 식별자
            - 속성 중에는 엔터티에 저장되는 어떤 특정 인스턴스 하나를 식별할 수 있는 하나 또는 하나 이상의 속성을
              유일 식별자라고 한다. 
            - 일련의 식별자 속성은 반드시 값을 갖는다. 
            - 일련의 식별자 속성의 값은 유일하다
            - 식별자를 구성하는 일련의 속성은 최소한의 개수로 해야 한다. 
### 데이터 모델링 표기법 이해(p340)
    1. 엔터티-관계 데이터 모델 표기법
        - 가. 엔터티 표기법
            - 정보공학(IE) 엔터티 표기법
            - CASE*Method 엔터티 표기법
        - 나. 속성 표기법
        - 다. 관계 표기법
            - 1) 관계 기수성(Cardinality, Degree) 표기법
                - 일대일 관계
                - 일대다 관계
                - 다대다 관계
            - 2) 관계 선택성 표기법
                - 필수 관계
                - 선택 관계
            - 3) 관계 식별성 표기법
                - 식별 관계
                - 비식별 관계
### 관계형 모델 이론(p349)
    1. 릴레이션 개념
            - 릴레이션 정의
                - 두 개의 집합 x,y 가 있을 때 두 집합간의 관계 R을 만족하는 모든 순서쌍
            - 속성과 도메인의 개념
    2. 관계형 모델 이론
        - 가. 데이터 구조
            - 1) 각 열은 하나의 값을 가진다. 
            - 2) 각 열의 값은 동일한 종류이다. 
            - 3) 각 행은 유일하다. 
            - 4) 열의 순서는 의미가 없다. 
            - 5) 행의 순서는 의미가 없다. 
            - 6) 각 열은 유일한 이름을 갖는다. 
        - 나. 데이터 조작
        - 다. 데이터 무결성
            - 1) 엔터티 무결성 규칙
                - 주 키는 널 값을 포함하지 않는다. 
                - 식별자를 구성하는 각 속성이 널이 아닐 것
                - 엔터티 내 특정 건의 유일성을 보장할 것
                - 최소한의 속성 집합일 것
                    - 최소한의 속성 집합이 아니면 유일 식별자의 유일성이 깨질 수 있다. 
            - 2) 참조무결성 규칙
                - 관계 엔터티의 모든 외래 키 값은 관련 있는 관계 엔터티의 모든 주 키 값이 존재해야 한다. 
                - 입력규칙
                - 삭제규칙
            - 3) 도메인 무결성 규칙
                - 엔터티 내의 모든 열에 관한 무결성 규칙으로 데이터 타입, 길이, 허용 값, 기본값, 유일성,
                  널 여부 등에 관한 제한이 있다. 
                - 같은 도메인의 값들끼리만 비교 가능
                - 속성이 취할 수 있는 값의 제한
            - 4) 연쇄 작용 또는 업무 규칙 
                - 입력, 삭제, 수정 또는 조회 등의 작업이 동일 엔터티 혹은 다른 엔터티의 속성에 영향을 미치는
                  업무규칙을 정의하는 것이다. 
## 데이터 모델링(p381)
### 논리 데이터 모델링 이해(p382)
    1. 논리 데이터 모델링 핵심
        - 가. 논리 데이터 모델링 필수 성공 요소
            - 가능한 많은 시간을 업무에 능통한 현업 사용자와 함께 데이터 모델링을 진행하라
            - 절차보다는 데이터에 초점을 두고 모델링을 진행하라
            - 데이터의 구조와 무결성을 함께 고려하라 
            - 개념화와 정규화 기법을 적용하라 
            - 가능하면 다이어그램을 이용하여 업무를 표현하라
            - 데이터 모델링을 지원하는 데이터 사전을 구축하다 
        - 나. 논리 데이터 모델링 목적 및 효과
            - 해당 비즈니스에 대한 데이터 관점에서의 명확한 이해 
            - 전사적인 통합 데이터 체계 확립
            - 데이터의 일관성 및 정확성 유지를 위한 규칙 도출
            - 안정적인 데이터베이스 설계의 토대 마련
            - 정보시스템 관계자와의 명확한 의사소통을 위한 수단으로 활용
    2. 논리 데이터 모델링 절차
            - 주제 영역 정의
                - 조직이 사용하는 데이터의 최상위 집합
                - 데이터를 하향식으로 분석하기 위한 개념으로 유용 
            - 엔터티 정의
            - 관계 정의
            - 속성 정의
                - 조직이 사용하는 데이터를 저장하는 속성을 정의
            - 식별자 확정
                - 엔터티의 하나의 인스턴스를 식별할 수 있는 일련의 속성
            - 정규화
                - 엔터티 내에 속성이 잘못 위치하여 발생하는 데이터 이상 현상 입력이상, 수정이상, 삭제이상을 제거하기
                  위하여 속성의 위치를 정확하게 찾아주는 과정
            - 이력관리
                - OLAP 환경에서의 시간의 이력 관점의 데이터 분석을 위해
### 주제 영역 정의(p387)
    1. 주제 영역 개념
    2. 주제 영역 도출
        - 가. 주제 영역 발견 기법
            - 주제 영역 발견 기법은 경영학에서 사용하는 가치 사슬 분석 기법을 활용
        - 나. 주제 영역 도출 방안
            - 1) 업무 기능의 이름으로부터 도출
                - 업무활동과 데이터의 상호보완관계를 유지하며 조직의 업무를 파악
            - 2) 업무에서 사용하는 데이터의 명사형 도출
                - 조직에서 무엇을 하는지 직관적으로 알 수 있는 대표적인 명사형 단어
            - 3) 하향식 접근방법
                - 최상위 주제 영역에서 출발하여 단위 주제 영역으로, 단위 주제 영역에서 엔터티로 전개
            - 4) 상향식 접근방법
                - 어떤 엔터티들을 모아서 하나의 다른 주제 영역으로 정의할 때 상향식 접근방법으로 도출
        - 다. 주제 영역 명명
            - 1) 업무에서 사용하는 용어를 부여한다. 
            - 2) 유일한 단수형 명사를 사용한다. 
            - 3) 데이터의 그룹을 의마하는 이름을 부여한다. 
    3. 주제 영역 활용
        - 가. 목적
            - 업무 분석의 효율을 높인다. 
            - 업무 기능과 병행하여 분석하는 경우에 모델의 품질 확보에 기여
            - 주제 영역 계층과 업무 기능 계층 간의 대응 관계를 확인
            - 주제 영역은 기업의 전사 업무를 위한 전체 데이터 구성에 대한 청사진을 제공한다. 
            - 데이터 구성 및 통합에 대한 방향을 제시한다. 
            - 효율적 데이터 관리를 위한 기준을 제공한다. 
        - 나. 장점
            - 데이터 및 업무 활동 모델의 품질을 보증한다. 
            - 프로젝트 관리: 업무 범위에 대한 기준을 확보한다. 
            - 분할 후 정복으로 집중화한 토의를 할 수 있다. 
            - 모델 개발 조정을 용이하게 한다. 
            - 저장소 관리를 좀 더 용이하게 한다. 
            - 상세 사항의 전개 혹은 축약이 가능하다. 
    4. 주제 영역 정의 내용 
        - 가. 주제 영역 목록
            - 레벨: 주제 영역의 계층 수준
            - 주제 영역 명
            - 설명(단위 주제 영역의 경우)
            - 대표 엔터티: 해당 주제 영역 내에서 대표적인 엔터티를 기술
        - 나. 주제 영역 정의서 샘플 양식
    5. 주제 영역 사례
        - 여신 업무
        - 여신 상담 및 신청
        - 신용조사
        - 여신심사
        - 여신승인
        - 여신 약정 체결
        - 여신실행
        - 여신사후관리
### 엔티티 정의(p394)
    1. 엔터티 개념
        - 엔터티란 조직에서 업무를 수행하는데 필요로 하는 사물, 사건 또는 개념을 나타내는 어떤 것
    2. 엔터티 분류
        - 가. 일반적인 엔터티 분류
            - 유형 엔터티: 물리적으로 존재하는 대상(고객, 상품)
            - 활동 엔터티: 어떤 사건에 관한 정보(주문, 계약, 장비 고장)
            - 개념 엔터티: 관리할 정보가 있는 무형의 개념(계정과목, 성적)
        - 나. 모델 관점 엔터티 분류
            - 독립 엔터티
                - 인스턴스를 식별하는 데 있어서 어떤 다른 인스턴스에 의존하지 않는 엔터티
                - 고객, 제품, 사원 
            - 종속 엔터티 
                - 인스턴스를 식별하는 데 있어서 하나 또는 하나 이상의 다른 인스턴스에 의존하는 엔터티
                - 사원가족, 주문상품)
                - 특성 엔터티
                    - 하나의 인스턴스에 여러 번 발생하는 속성의 그룹
                    - 다른 인스턴스에 의하여 직접적으로 인식되지는 않는 엔터티 (사원경력)
                - 연관 엔터티
                    - 두 개 이상의 연관된 다른 엔터티로부터 식별자를 상속 받는 엔터티 (관계 해소시)
                - 서브타입 엔터티
                    - 다른 부분집합과 구별되는 공통 속성이나 관계를 공유하는 엔터티의 부분집합
        - 다. 발생 시점 엔터티 분류
            - 1) 키 엔터티
                - 자신의 부모를 가지지 않는 엔터티
                - 사원, 부서, 고객, 상품, 자재
            - 2) 메인 엔터티
                - 키 엔터티를 제외한 엔터티 중에서 업무 중심에 해당하는 엔터티
                - 보험계약, 사고, 예금원장, 청구, 구매의뢰, 출하지시, 공사, 주문, 예산편성, 매출 
            - 3) 액션 엔터티
                - 키 엔터티와 메인 엔터티가 아닌 것 모두 
                - 상태 이력, 차량 수리 내역, 상세 주문 내역
    3. 엔터티 도출
        - 현 시스템 분석, 기업의 전략-목표 분석, 사용자 인터뷰, 정보요구 분석, 문서-보고서 분석 
        - 가. 현행 데이터베이스로부터 엔터티 도출
            - 현행 데이터베이스가 조직의 정보 요구 사항을 빠짐없이 정확하게 지원하고 있는지 알기 위해
            - 현행 데이터베이스의 구조적인 문제점 및 해결방안을 도출하기 위해
            - 목표 정보 요구 사항을 지원하기 위하여 논리 데이터 모델을 어떻게 생성할지를 결정하기 위해
            - 현행 데이터베이스에서 목표 데이터베이스로의 데이터 이행이 필요한 경우 이에 대한 원천 데이터를
              정확하게 파악하기 위해 
        - 나. 정보요구 분석
        - 다. 업무에서 사용하는 문서로부터 도출
        - 라. 사용자 인터뷰로부터 도출
        - 마. 아키텍처 모델로부터 도출
        - 바. 업무를 수행하는 프로세스로부터 도출
        - 사. 엔터티 분류 기법을 활용하여 도출 
    4. 엔터티 검증
        - 가. 조직의 업무를 수행하는데 필요한 의미 있는 정보를 나타내야 한다. 
        - 나. 특정 사례가 아닌 유사한 사물들을 대표하는 집합체
        - 다. 인스턴스가 포함할 내부 연관성 있는 속성에 의해 결정된 단일개념을 대표해야 한다. 
        - 라. 엔터티 내 인스턴스의 출현을 구별할 수 있는 능력을 제공해야 한다. 
        - 마. 정규화 규칙을 만족해야 한다. 
    5. 엔터티 정의 사항
        - 가. 엔터티 명
            - 지켜야 할 명명규칙
                - 엔터티 명은 현실 세계의 인스턴스 및 정보를 대표하도록 한다. 
                - 엔터티 명 자체로 의미를 표현하도록 한다. 
                - 물리적인 사항을 개념화한 논리적 사물을 반영하도록 한다. 
                - 최소한의 어휘를 사용하여 의미를 전달할 수 있도록 한다. 
            - 지키지 말아야 할 명명규칙
                - 데이터 관리자의 승인을 받지 않은 약어 또는 두문자어
                - 테이블, 파일, 메뉴, 보고서, 입력화면 같은 정보시스템 구성요소 
                - 두 가지 이상의 개념을 동시에 사용하는 표현
        - 나. 엔터티 설명
            - 엔터티 그 자체가 무엇이며, 업무에서 이 엔터티가 왜 중요한가, 엔터티의 범위에 대하여 기술
        - 다. 엔터티 분류
            - 업무를 좀 더 잘 이해하기 위한 개념
        - 라. 현재 발생 건수
        - 마. 발생 건수 변화
        - 바. 권한
        - 사. 식별자 속성
        - 아. 외래 식별자 속성
        - 자. 식별자 이외 속성
    6. 엔터티 일반화
        - 가. 일반화 정의
            - 현실세계의 사물, 사건을 단순화하여 표현하는 추상화 기법 중 하나로 엔터티의 부분집합을 정의하는 것
            - 슈퍼타입 엔터티
                - 일반화 계층의 가장 상위에 있는 엔터티
            - 서브타입 엔터티
                - 다른 부분집합과 구별되며, 공통의 속성이나 관계를 공유하는 인스턴스들의 부분집합
        - 나. 일반화 특성
            - 슈퍼타입은 두 개 이상의 독립적인 서브타입으로 구성된다. 
            - 슈퍼타입은 각 서브타입들의 공통적인 속성과 관계를 가진다. 
            - 서브타입은 자신의 속성이나 독립적인 관계를 가진다. 
            - 자신의 속성이나 관계를 가지지 않는 서브타입도 존재할 수 있으나 이런 경우는 일반적인 속성으로
              처리하는 것이 좋다. 
        - 다. 일반화 시점
            - 인스턴스가 공통적인 속성 집합을 가지고 있을 때
            - 인스턴스가 공통적인 관계 집합을 가질 때 
        - 라. 일반화 장단점
            - 일반화
                - 장점
                    - 데이터가 통합되어 있어 조회가 간편하다
                    - 한 개의 프로그램으로 업무를 처리할 수 있다. 
                    - 동일한 종류의 업무 추가 시 최소한의 변경으로 지원 가능하다. 
                    - 전체적인 업무 로직 변화의 경우 신속하게 반영할 수 있다. 
                - 단점
                    - 공통적이지 않은 속성이 많은 경우 효과가 반감된다. 
                    - 여러 업무의 데이터가 서로 섞여 있어 내용을 정확하게 파악하기 힘들다. 
                    - 모든 업무를 이해하는 인원 만이 데이터 관리를 수행할 수 있다. 
                    - 특정 업무에만 관련된 데이터 변경이 있을 경우 타 업무에 영향을 미치게 된다. 
            - 상세화
                - 장점
                    - 업무에 대한 명확한 파악이 가능하다. 
                    - 데이터의 추가나 변경이 자유롭다 
                    - 특화된 업무에 대한 속성 관리나 프로그램이 용이하다. 
                    - 장애나 에러에 대한 파급효과가 작다
                    - 데이터 구조가 명확하여 업무 파악이나 인수인계를 신속하게 수행할 수 있다. 
                - 단점
                    - 엔터티의 수효가 늘어나서 관리에 어려움이 발생한다. 
                    - 데이터 간의 불일치를 초래할 수 있다. 
                    - 통합적인 정보 도출이 어렵다. 
                    - 동일한 로직도 별도의 프로그램으로 개발해야 하므로 관리가 비효율적이다.

        - 관계 읽기

            - 교수는 학생을 가르친다.
        - 사건 관계

        - 마. 참조무결성

                - 계층 구조 변경 시 과거 데이터의 수정 없음

        - 나. 다대다 관계

        - 다. 배타적 관계

              하나의 엔터티의 행과 관계를 맺는 형태
### 속성 정의(p432)
    1. 속성 개념
        - 엔터티에 속하는 모든 인스턴스들이 공통으로 가지는 이러한 가각의 특성을 속성이라고 부른다. 
    2. 속성 도출
        - 가. 현행 시스템 자료
        - 나. 현업 장표 보고서
        - 다. 사용자와 협의
        - 라. 데이터 흐름도의 데이터 저장소
        - 마. 전문 서적 및 자료
        - 바. 다른 시스템 자료
    3. 속성 정의 사항
        - 속성 상세 정보 중요성
            - 속성 상세 정보 정의는 업무 관련 데이터의 본질과 목적을 이해하는데 도움을 준다. 
            - 속성 수준의 무결성을 설정하고 강화하는 데 도움을 준다. 
            - 데이터 무결성을 개선하여 데이터 품질을 향상시킨다. 
            - 데이터 사전을 구성한다. 
        - 가. 속성 명
            - 1) 의미가 명확한 속성 명칭 부여
            - 2) 유일한 복합명사 사용
            - 3) 단수형으로 속성 명 사용
            - 4) 표준 단어 및 용어 제정
        - 나. 속성 설명
            - 분석자가 업무 전문가에게 질문을 통하여 알게 된 유용하다고 생각하는 정보
        - 다. 선택성 및 선택성 조건
        - 라. 속성 유형
            - 기본 속성
                - 속성 값이 해당 인스턴스에 원래 존재하여, 다른 속성 값으로부터 유도될 수 없는 속성
            - 유도 속성
                - 속성 값이 항상 다른 속성의 값으로부터 유도되거나 계산되는 속성
            - 설계 속성
                - 업무 제약사항을 반영하거나 시스템 운영을 단순화하기 위하여 생성하는 속성
        - 마. 유도 알고리즘
            - 유도형 속성에 값을 지정할 수 있다. 
        - 바. 도메인
            - 하나의 속성에 다형 지정 가능한 값의 집합
        - 사. 허용 값
        - 아. 기본 값
        - 자. 소유 권한
            - 데이터 거버넌스 차원의 속성 정의 및 승인에 대한 권한을 가지는 개인, 기능 또는 조직에 대한 기록을 관리
            - 데이터를 생성하고 사용하는 것에 대한 권한을 가지는 개인
    4. 속성 검증 및 확정
        - 가. 원자 값 단위까지 분할 
            - 1) 일자 형태의 속성 : 매출일자, 생년월일
            - 2) 외부에서 만들어진 공인된 속성 : 우편번호 유형
            - 3) 전화번호 유형 : 전화번호, 팩스번호
        - 나. 하나의 값만을 가지는지 검증
            - 1) 계약일자
            - 2) 차량번호
            - 3) 취미
        - 다. 유도 속성인지 검증
            - 1) 현재 정보만 관리하는 형태 : 현주소, 현고객등급코드 등
            - 2) 최초 정보를 보관하는 형태 : 최초가입일자, 모집사원번호 등
            - 3) 집계 정보를 관리하는 형태 : 인원 수, 가족 수, 총직원 수 등
            - 4) 추출 관계만 관리하는 형태 : 가입, 계약번호, 관리 사원 등
            - 5) 대표 정보만 관리하는 형태 : 대표 전사메일 ID, 취미, 법인의 대표자 정보
            - 6) 다른 속성의 일부 정보만 분리한 형태 : 성별, 결혼 여부 등 
            - 7) 일부분만 추출 값인 형태 : 인원 수, 법인의 대표자 정보 등
    5. 유도 속성 규칙
        - 유도된 속성이나 계산된 속성은 중복으로 인식하고 이를 데이터 모델에 표현하지 말 것을 권고
        - 유도 속성이란 하나 이상의 다른 데이터 속성에 축적된 여러 사례의 값을 토대로 어떤 추가 계산 작업을
          수행함으로써 선택적으로 주제를 도출하는 속성에 대하여 새로운 값을 창출하는 속성
        - 계산 속성은 엔터티의 단일 사례에 어떤 특성을 기술하며, 일반적으로 관련 속성의 또다른 단일 사례로 계산
        - 유도 속성은 경영층이 진실로 원하고 필요로 하는 정보를 대표 
        - 유도 속성은 사용자들의 데이터 요구 사항을 나타낸다. 
        - 데이터 모델에 유도 속성을 포함시키는 것은 그들의 물리적 구현에 대한 어떤 것도 내포하지 않는다. 
        - 유도 속성은 향후 과정에서 결코 식별자로서의 역할을 맡지 않아야 한다. 
### 식별자 확정(p446)
    1. 식별자 개념
        - 가. 식별자 종류
            - 1) 기준 정보 엔터티의 본질 식별자
            - 2) 거래 처리 엔터티의 본질 식별자
        - 나. 후보 식별자 도출
            - 1) 널이 될 수 없다. 
            - 2) 각 인스턴스를 유일하게 식별할 수 있어야 한다. 
            - 3) 나머지 속성들을 직접 식별할 수 있어야 한다. 
            - 4) 후보 식별자로 속성 집합을 선택하는 경우에는 개념적으로 유일해야 한다. 
            - 5) 후보 식별자의 데이터는 자주 변경되지 않는 것이어야 한다. 
        - 다. 보조 식별자
            - 원래의 식별자를 대신할 수 있는 또 다른 속성들이나 관계를 말한다. 
        - 라. 인조 식별자 지정
            - 식별자 확정 시 기존의 본질 식별자를 그대로 인정할 수 없는 여러 가지 상황이 발생했을 때 
              전부 혹은 일부의 임의의 값을 가진 속성들로 대체하여 새롭게 구성한 식별자를 말한다. 
            - 1) 최대한 범용적인 값을 사용한다. 
            - 2) 유일한 값을 만들기 위한 인조 식별자를 사용한다. 
            - 3) 하나의 인조 식별자 속성으로 대체할 수 없는 형태를 주의한다. 
            - 4) 편의성, 단순성 확볼르 위한 인조 식별자를 사용할 수 있다. 
            - 5) 의미의 체계화를 위한 인조 식별자를 사용할 수 있다. 
            - 6) 내부적으로만 사용하는 인조 식별자
        - 마. 식별자 확정
            - 1) 식별 관계의 두 가지 의미
                - 식별자로서의 역할
                    - 엔터티 자신의 입장에서 자신의 인스턴스들을 다른 것들과 구별될 수 있도록 유일한 값을 
                      만드는데 일조한다는 의미
                - 정보로서의 역할
                    - 참조하는 엔터티의 입장에서 상대방의 식별자를 상속 받았기 때문에 자신이 보유한 정보가
                      증가했다는 의미도 가지고 있다. 
            - 2) 식별자 상속과 단절의 원리
                - 식별자의 적절한 상속과 단절 전략은 실질적인 처리의 단순화를 가져다 줄 수 있으므로 
                  그 전략적 가치를 가지고 있다. 
            - 3) 식별자 확정 절차
                - 기준 정보 엔터티 식별자 확정
                    - 부모를 가지지 않는 최상위 엔터티이므로 서로 독립적으로 식별자를 확정할 수 있다. 
                - 중요 거래 처리 엔터티 식별자 확정
                    - 해당 업무의 근본이 되는 엔터티라고 할 수 있으므로 자신이 하위에 거느리고 있는
                      수많은 엔터티의 상황을 종합적으로 감안한 전략적으로 도출 해야 한다. 
                - 기타 거래 처리 엔터티 식별자 확정
                    - 가능하다면 인조 속성을 많이 사용하지 않는 것이 바람직하다. 
### 정규화(p454)
    1. 정규화 정의
        - 엔터티에 데이터의 입력, 수정, 삭제 연산을 수행할 때 발생하는 이상 현상을 제거하여 논리 데이터 모델링의 
          목적인 정확성, 일관성, 단순성, 비중복성, 안정성을 만족시키는 최적의 데이터 구조를 만들어가는 과정
    2. 정규화 장점
        - 중복 값이 줄어든다. 
        - 널 값이 줄어든다. 
        - 복잡한 코드로 데이터 모델을 보완할 필요가 없다. 
        - 새로운 요구 사항의 발견 과정을 돕는다. 
        - 업무 규칙의 정밀한 포착을 보증한다. 
        - 데이터 구조의 안정성을 최대화한다. 
    3. 정규화 과정
        - 입력 이상
            - 어떤 데이터를 입력하려고 할 때 불필요하게 원하지 않는 데이터도 함께 입력해야 하는 구조 
        - 수정 이상
            - 일부 속성 값을 수정함에 있어서 원하지 않는 정보의 이상 현상이 발생
        - 삭제 이상
            - 일부 정보를 삭제함으로써 유지되어야 할 정보까지도 연쇄 삭제되는 현상이 발생
        - 가. 제1정규형
            - 모든 속성은 반드시 하나의 값을 가져야 한다. 
            - 각 속성의 모든 값은 동일한 형식이어야 한다. 
            - 각 속성들은 유일한 이름을 가져야 한다. 
            - 행들은 서로 간의 식별 가능해야 한다. 
        - 나. 제2정규형
            - 식별자가 아닌 모든 속성은 식별자 전체 속성에 완전 종속되어야 한다. 
            - 주 키가 아닌 모든 칼럼이 주 키에 종속적이어야 제2정규형을 만족시킨다. 
        - 다. 제3정규형
            - 제2정규형을 만족하고 식별자를 제외한 나머지 속성들 간의 종속이 존재하면 안된다. 
### 이력 관리(p463)
    1. 이력 관리란?
        - 데이터는 현재의 프로세스만 처리하고 버리는 것이 아니라 마치 후손에게 물려주어야 할 귀중한
          문화유산처럼 오랜 기간의 데이터를 유지시켜 좀 더 가치 있는 정보를 제공할 수 있는 밑거름
    2. 이력 관리 대상 선정
        - 가. 사용자 조사
            - 변경 내역을 감시할 필요가 있는가?
            - 시간의 경과에 따라 데이터가 변할 수 있는가? 
            - 시간의 경과에 따라 관계가 변할 수 있는가?
            - 과거의 데이터를 조회할 필요가 있는가?
            - 과거 버전을 보관할 필요가 있는가? 
        - 나. 이력 데이터의 종류
            - 1) 발생 이력 데이터
                - 어떤 데이터가 발생할 때마다 이력 정보를 남겨야만 하는 이력
            - 2) 변경 이력 데이터
                - 데이터가 변경될 때마다 변경 전과 후의 차이를 확인 필요 이력
            - 3) 진행 이력 데이터 
                - 업무의 진행에 따라 데이터의 이력 정보를 남겨야 하는 이력
        - 다. 이력 관리 형태
            - 1) 시점 이력
                - 데이터의 변경이 발생한 시각만을 관리
            - 2) 선분 이력
                - 데이터 변경의 시작 시점부터 그 상태의 종료 시점까지 관리
        - 라. 선분 이력 관리 유형 
            - 1) 인스턴스 레벨 이력 관리
                - 하나의 인스턴스의 어떤 변경이라도 발생하면 전체 인스턴스를 새롭게 생성하는 방식의 이력 관리 
                - 한 번의 액세스로 스냅 샷을 참조하는 것이 가능
                - 로그성 데이터를 저장할 목적인 경우 적당함
                - 다른 이력 관리 유형에 비해 저장하기가 쉽다. 
                - 가장 큰 단점 중 하나는 하나 이상의 컬럼에 변경이 발생했을 때 이벤트가 모호해짐
                - 만약 이벤트가 자식 정보를 가지게 된다면 매우 치명적
                - 이력 관리의 다른 유형들에 비해 저장 공간의 낭비를 초래할 수 있다. 
                - 실제 어떤 데이터가 변경된 것인지를 찾기 위해서는 과거의 데이터를 Merge해서 비교 가능 
                - 특정 순간의 스냅 샷만 보는 게 아니라면 처리가 복잡해지는 경향
                - 변화가 빈번하게 발생하는 상황이라면 고려해 볼 수 있는 유형
            - 2) 속성 레벨 이력 관리
                - 이력을 관리할 속성에서 변화가 생길 때만 이력을 생성하는 방식 
                - 변경 이벤트가 매우 분명하게 관리된다. 
                - 하나의 이력 관리 엔터티에서 다른 엔터티와 통합 이력 관리가 가능 
                - 변경된 것만 처리하기 때문에 독립적 처리가 가능
                - 변화가 발생할 가능성은 매우 낮으면서 이력 관리 대상 속성은 매우 많은 경우에 사용하는 것이 유리
                - 특정 속성들에 변화가 집중되는 경우에 해당 속성에 대해서만 이력을 관리할 수 있기 때문에 유리
                - 여러 속성에 대한 이력이 필요할 때 많은 Merge가 발생
                - 이력 관리의 다른 유형들에 비해서 향후 사용될 데이터 액세스 쿼리에서 조건 검색이 조금 어렵다. 
                - 변화가 너무 많은 경우에는 적용이 곤란
            - 3) 주제 레벨 이력 관리
                - 내용이 유사하거나 연동될 확률이 높은 것별로 로우 레벨 이력을 관리하는 방법
                - 로우 레벨과 속성 레벨의 장점을 모두 수용하는 형태의 이력 관리 형태
                - 목적이 분명한 엔터티를 생성함으로써 확장성을 확보할 수 있는 용도로 사용
                - 변경 부분만 처리가 가능
                - 다른 엔터티와 통합 이력 관리가 가능
                - 속성 레벨의 단점을 해소 
                - 전체를 참조할 때 로우 레벨에 비해 Merge가 발생하는 문제점이 존재
                - 부문에 따라 변경 정도의 차이가 심한 경우 유리 
    3. 선분 이력 관리용 식별자 확정
        - 가. 선분 이력에서 식별자 결정 시 고려 사항
        - 나. 선분 이력에서 종료점 처리 시 주의 사항
            - 1) 종료점이 미정이므로 NULL
                - 논리적으로 타당하지만 비교가 불가능
                - 인덱스를 사용하지 못하므로 수행 속도 저하
            - 2) 수렴하므로 최대치 부여 
                - 아직 종료되지 않았으므로 무한히 계속되는 것으로 간주
                - 최대치 부여 (9999-12-31)
                - 가능한 테이블 생성 시 기본 값 제약 부여
                - 수행 속도에 유리

            - 업무 영역 또는 요구 사항이 정확하게 반영

            - 데이터 모델의 구성요소를 정의하는데 있어서 누락을 최소화

            - 제반 준수 요건들이 누락 없이 정확하게 준수

        - 활용성

            - 업무 변화 시에 설계 변경이 최소화 하도록 유연하게 설계
    2. 논리 데이터 모델 품질 검토 체크리스트의 활용

## 물리 데이터 모델링(p481)	
### 물리 데이터 모델링 이해(p482)
    1. 물리 데이터 모델 정의
        - 논리적 모델을 특정 데이터베이스로 설계함으로써 생성된 데이터를 저장할 수 있는 물리적인 스키마
    2. 물리 데이터 모델 의의 
        - 논리 데이터 모델에서 도출된 내용 변환을 포함하여 데이터의 저장 공간, 데이터의 분산, 데이터 저장 방법
          등을 함께 고려하는 단계 
    3. 논리 데이터 모델-물리 데이터 모델
        - 가. 분산 데이터베이스 구축 시 
            - 분산 데이터베이스 구축 시 노드별로 자신이 원하는 형태의 물리적 모델을 생성하고자 할 때 적용
        - 나. 물리 데이터 모델 비교
            - 각자 나름대로의 특징을 가지고 있는 여러 개의 물리적 모델을 생성하여 종합적인 비교, 검토하기 
              위하여 적용
        - 다. 물리적 환경의 변화
            - 논리적인 모델에는 변화가 발생하지 않지만 물리적인 환경에는 변경이 발생했을 경우 기존의
              물리적 모델을 새로운 목표 물리적 모델로 개선하고자 할 때 적용 
        - 라. 물리적 모델의 형상 관리
            - 물리적 모델이 세월의 흐름에 따라 조금씩 변해갈 경우 그 이력을 관리할 목적으로 여러 개의
              버전을 보유하고자 할 때 사용

            - 현재 입출력 컨트롤러의 성능 및 적절하게 운용되고 있는가를 파악한다.

### 논리-물리 모델 변환(p486)
    1. 논리 데이터 모델-물리 데이터 모델 변환 용어
    2. 엔터티-테이블 변환
        - 가. 테이블 설명
            - 1) 테이블
            - 2) 로우
            - 3) 칼럼
            - 4) 기본키
            - 5) 외래키
        - 나. 서브타입 변환
            - 1) 슈퍼타입 기준 테이블 변환
                - 정의
                    - 서브타입을 슈퍼타입에 통합하여 하나의 테이블로 만든다. 
                    - 이 통합된 테이블에는 모든 서브타입의 데이터를 포함해야 한다. 
                    - 주로 서브타입에 적은 양의 속성이나 관계를 가진 경우에 적용한다. 
                - 절차 
                    - 슈퍼타입으로 테이블 명칭 부여
                    - 서브타입을 구분ㅂ할 수 있도록 컬럼 추가 
                    - 슈퍼타입의 속성을 컬럼명으로 
                    - 서브타입의 속성을 컬럼명으로
                    - 슈퍼타입의 관계를 FK로
                    - 서브타입의 관계를 FK로
                - 하나의 테이블로의 통합이 유리한 경우
                    - 데이터 액세스가 좀 더 간편
