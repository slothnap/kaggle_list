## 정보 요구 사항
### 정의
- 정의
    - 사용자가 일상적으로 수행하는 업무의 개선 사항이나 신규 개발 사항
    - 시스템을 통해 기능상의 목적을 달성하기 위해 요청하는 내용
- 수집
    - 현행 시스템 분석
    - 사용자 요구 사항 수집
    - 제안 요청서
    - 사업 수행 계획서
- 요구 사항 유형 분류
    - 비즈니스 요구 사항
    - 이해관계자 요구 사항
    - 해결책 요구 사항 
        - 기능적 요구 사항
        - 비기능적 요구 사항
    - 전환 요구 사항
    - 프로젝트 요구 사항
    - 품질 요구 사항
- 프로젝트 실패 원인 분류
    - 계획과 추정 요인
    - 실행 요인
    - 인간적 요인
### 정보 요구 사항 생명주기 모형
- 정보 요구 사항 도출         == 사용자 인터뷰·설문지·워크숍·현행 시스템 분석
- 정보 요구 사항 분석 및 정의 == 수집된 정보 요구 사항 정리, 분석 및 정의
- 정보 요구 사항 명세화       == 확정된 정보 요구 사항 기록 
- 정보 요구 사항 검증         == 사용자 정보 요구 사항 반영 검증

        - 프로젝트 관리 및 지원상황
    - 기타 요구 사항
        - 유지관리 사항
        - 컨설팅
        - 공사
- 정보 요구 사항 유형 
    - 외부 인터페이스 요건
        - 기존과 동일한 형태의 인터페이스 존재 여부
    - 기능 개선 요건
        - 많은 사용자가 편리하게 사용할 수 있는 요건의 우선 적용 여부

### 정보 요구 사항 프로세스 
- 업무 흐름 프로세스 (발송->수렴->검토->영향도->공식화->반영계획)
    - 요구 사항 발송
    - 요구 사항 수렴
    - 요구 사항 검토
    - 영향도 분석 
    - 공식화 
    - 반영 작업 계획 수립 
- 수행 조직 및 수행 업무
    - 사용자
        - 정보 요구 사항 정의 및 상세화
        - 정보 요구 사항 변경 요청
        - 정보 요구 사항 반영을 위한 미팅
        - 정보 요구 사항 반영 여부 확인
        - 미결 사항에 대한 의사결정 실시
    - 담당자
        - 사용자 정보 요구 사항 접수 
        - 사용자 정보 요구 사항에 대한 기본적인 검토

- 애플리케이션 충족도 분석 매트릭스 
    - 정보 항목 생성 기본 프로세스 반드시 존재 
    - 정보 항목 상태 종료 기본 프로세스 존재 
    - 정보 항목 조회, 수정, 삭제 중 하나 발생 필요 
    - 하나의 정보 항목 생성, 수정, 삭제 합 7개 초과 x 
    - 조회 전용이 아니고선 반드시 생성, 수정, 삭제 하나 수행
- 매트릭스 분석 
    - 기본 프로세스 사용 x
        - 정보 항목 누락            -> 정보 항목 도출
        - 기본 프로세스 필요없음    -> 기본 프로세스 삭제
        - 기본 프로세스 업무 영역 x -> 해당 업무 영역으로 이동
    - 정보 항목 7개 이상
        - 정보 항목 너무 큼         -> 정보 항목 세분화 필요 
    - 정보 항목 생성 없음
        - 누락 정보 항목 필요 x     -> 도출 정보 항목 삭제
        - 정보 항목 업무 영역 x     -> 해당 업무 영역으로 이동
    - 엔터티 삭제 프로세스 없음
        - 기본 프로세스 누락        -> 기본 프로세스 도출
        - 업무 삭제 존재 x          -> 전사 오류 or 삭제 필요 확인
        - 기본 프로세스 업무 영역 x -> 해당 업무 영역으로 이동
    - 삭제 프로세스 둘 이상 존재
