# 데이터 모델링(p319)
## 데이터 모델링 개요
### 데이터 모델링 정의 
- 데이터 모델
    - 현실 세계를 데이터베이스로 표현하는 중간 관정
    - 컴퓨터에 저장할 데이터의 구조 및 업무 규칙을 표현하기 위해 사용하는 지적인 도구 
- 데이터 모델링 탄생 배경 
- 정보시스템 개발 방법론 
    - 구조적 방법론
        - 데이터 흐름도(DFD)로부터 업무를 구조적으로 분석, 설계 -> 소프트웨어 개발하는 절차 
    - 정보 공학 방법론
        - 계획 
            - 정보 전략 계획   == ISP (Information Strategy Planning)
                - 정보 아키텍처
                    - 데이터 아키텍처
                    - 프로세스 아키텍처
                - 시스템 아키텍처 
                    - 시스템 단위의 화면 설계
                    - 데이터베이스 설계 
                - 기술 아키텍처 
                - 조직 아키텍처 
        - 분석 
            - 업무 영역 분석   == BAA (Business Area Analysis)
        - 설계 
            - 업무 시스템 설계 == BSD (Business System DEsing)
        - 구축 
            - 구축             == CON (Construction)
    - 객체지향 방법론
    - 컴포넌트 기반 방법론
### 데이터 모델링의 중요성 
- 데이터 모델링
    - 조직의 업무 수행에 필요한 데이터베이스를 생성하기 위하여 업무에서 필요로 하는 정보의 구조와 
      규칙을 명확하게 표현하기 위한 지적인 작업 과정 
- 데이터 무결성
    - 사용자가 관계형 테이블에 입력,수정,삭제,조회라는 관계형 연산을 수행할 때, 테이블 내 데이터 값의 
      정확성과 일관성을 보장하는 일련의 업무 규칙 
- 데이터 모델은 데이터베이스 설계에 대한 계획 또는 청사진이다. 






## 데이터 모델링 기법 이해 
### 엔터티-관계 데이터 모델 
- 추상화 
    - 현실세계 존재하는 구체적 개체(Entity) 또는 객체(Object)들의 공통된 특징 
    - 추상적 특징을 파악하여 인식의 대상으로 삼는 행위 
- 논리 데이터 모델링 추상화 
    - 유형화 (주제영역)
        - 현실세계에 존재하는 같은 성질을 갖는 멤버들을 타입 또는 유형이라는 하나의 개념으로 정의 
        - 조직의 업무가 수행되는 관점에 따라서 하나의 엔터티를 여러 가지 엔터티 타입으로 유형화 
            - ex) 소나타 -> 상품, 고정자산, 운송수단
    - 집단화 (속성)
        - 값(Value)을 유형화하면 속성이라는 타입
        - 속성이라는 타입들의 세트로 구성되는 새로운 타입(엔터티 타입)을 정의 
    - 일반화 (서브타입)
        - 여러 엔터티 타입 간의 공통적인 특성을 파악하는 과정 
        - 둘 또는 그 이상의 엔터티 타입 요소 간에 서브세트(부분집합)를 정의
- 타입과 인스턴스 
    - 타입: 실제 발생된 사례들을 추상화한 집합
    - 인스턴스: 현실 세계에서 실제로 발생된 사례(타입의 멤버)
- 표준 용어 정의 
    - 엔터티-관계 데이터 모델
        - 업무에서 다루는 사물(thing)
            - 엔터티 (Entity)-본서 표준
                - 엔터티 타입 (Entity Type)
                - 개체 타입 (Entity Type)
                - 실체 유형 (Entity Type)
            - 인스턴스(Instance)-본서 표준
                - 엔터티 (Entity)
                - 개체 (Entity)
                - 실체 (Entity)
        - 사물(thing) 사이 연관
            - 관계 (Relationship)-본서 표준
            - 페어링 (Pairing)-본서 표준
        - 사물(thing)이 갖는 특성
            - 속성(Attribute)-본서 표준
            - 값(Value)-보넛 표준
    - 오브젝트 컴포넌트 모델
        - 업무에서 다루는 사물(thing)
            - 클래스(Clss)-본서 표준
            - 객체(Object)-본서 표준
### 엔터티-관계 데이터 모델 구성요소
- 엔터티 
    - 조직의 업무를 수행하는데 필요로 하는 사물, 사건 또는 개념을 나타내는 것 
    - 동일한 업무 행위를 하는 단일 개념으로 정의한 인스턴스들의 집합체 
    - 유사한 속성을 갖는 단일 개념으로 정의한 인스턴스들의 집합체 
- 관계
    - 하나 또는 두 개의 엔터티에서 인스턴스를 연관시키는 업무적인 이유 
- 속성과 도메인 
    - 속성은 엔터티에 저장되는 인스턴스들의 특성을 설명하는 항목 
    - 도메인은 하나의 속성이 갖는 같은 종류의 모든 값들의 집합 
- 식별자 
    - 속성 중에서 엔터티에 저장되는 어떤 특징 인스턴스 하나를 식별할 수 있는 하나 또는 하나 이상의 속성을 유일 식별자 
    - 특성
        - 일련의 식별자 속성은 반드시 값을 갖는다. 
        - 일련의 식별자 속성의 값은 유일(Unique)하다. 
        - 식별자를 구성하는 일련의 속성은 최소한의 개수로 해야 한다. 






## 데이터 모델링 표기법 이해 
### 엔터티-관계 데이터 모델 표기법 
- 엔터티 표기법 
    - 정보공학(IE) 엔터티 표기법
        - 종속 엔터티 
            - 하나의 인스턴스를 식별하는 데 있어서 하나 또는 하나 이상의 다른 인스턴스에 의존하는 엔터티 
            - 특성, 연관, 서브타입 
        - 특성 엔터티
            - 하나의 인스턴스에 여러 번 발생하는 속성의 그룹 (사원경력, 고객주소)
        - 연관 엔터티 
            - 두 개 이상의 연관된 다른 엔터티로부터 식별자를 상속 받는 엔터티(다대다 관계 해소 시 생성됨 예, 주문상품 등).
        - 서브타입 엔터티 
            - 다른 부분집합과 구별되는 공통 속성이나 관계를 공유하는 엔터티의 부분집합(예, 개인고객, 법인고객 등)
    - CASE*Method 엔터티 표기법 
- 속성 표기법 
    - 속성은 엔터티에 저장되는 인스턴스들의 특성을 설명하는 항목으로, 단어들을 조합한 명사적 용어로 표현한다. 
    - 엔터티 명 + 수식어 + 도메인 명
- 관계 표기법 
    - 관계란 하나 또는 두 개의 엔터티에서 인스턴스를 연관시키는 업무적인 이유 
    - 관계 기수성(Cardinality, Degree) 표기법
        - 일대일 관계
        - 일대다 관계 
        - 다대다 관계 
    - 관계 선택성 표기법 
        - 필수 관계 
        - 선택 관계 
    - 관계 식별성 표기법 
        - 식별 관계 
        - 비식별 관계
        
        
        
        
        
        
## 관계형 모델 이론 
### 릴레이션 개념 
- 용어 비교 
    - 관계형(집합) 이론 
        - 릴레이션
        - 튜플
        - 애트리뷰트 
    - 논리 데이터 모델
        - 엔터티
        - 인스턴스 
        - 속성 
    - 관계형 데이터베이스
        - 테이블 
        - ROW 
        - 컬럼 
    - 파일(File) 시스템
        - 파일 
        - 레코드 
        - 필드 
- 릴레이션
    - 두 개의 집합 x, y가 있을 때 두 집합간의 관계 R을 만족하는 모든 순서쌍을 말합니다. 
    - 차수(Degree) == 애트리뷰트(속성) 개수 
    - 기수성(Cardinality) == 튜플(Row) 
- 속성과 도메인의 개념 
### 관계형 모델 이론 
- 관계형 모델 이론 
    - 데이터 구조
        - 사용자가 인식하는 데이터의 구성
    - 데이터 조작 
        - 사용자가 관계형 데이터 구조에 행하는 일련의 연산(처리) 형태 
    - 데이터 무결성 
        - 사용자가 일련의 관계형 연산을 수행할 때 관계형 테이블의 데이터 값이 어떻게 되야 하는 지를 통제하는 일련의 업무 규칙 
- 데이터 구조
    - 각 열은 하나의 값을 가진다.
        - 정보 접근의 유연성
        - 테이블 구조 변경의 유연성
        - 정규화 이론에 나오는 입력이상, 수정이상, 삭제이상 현상을 없애고 데이터 무결성 확보 
    - 각 열의 값은 동일한 종류이다. 
        - 데이터 무결성 규칙 
    - 각 행은 유일하다. 
        - 엔터티 무결성 규칙 
    - 열의 순서는 의미가 없다. 
    - 행의 순서는 의미가 없다. 
    - 각 열은 유일한 이름을 가진다. 
- 데이터 조작 
    - 집합 처리 
    - 관계 연산자: 조회 
        - Select (or Restrict): 열에 의거한 행의 부분집합
        - Project: 열의 부분집합
        - Product: 두 관계 테이블간 행의 조합을 묶음
        - Join: 열의 기준에 의거하여 각 행을 수평적으로 묶음
        - Union: 중복을 없이하여 각 행을 수직적으로 묶음
        - Intersection: 관계 테이블간의 공통된 행
        - Difference: 하나의 관계 테이블에만 있는 행
        - Division: 다른 관계 테이블의 모든 행에 대응하는 열을 제외한 열 
    - 처리 연산자: 관계 테이블의 내용에 변화
        - Insert: 행의 입력
        - Update: 행의 수정
        - Delete: 행의 삭제 
- 데이터 무결성 
    - 사용자가 일련의 관계형 처리(입력, 수정, 삭제, 조회)를 수행할 때 관계형 엔터티의 데이터 값이 어떻게 작용해야 하는지를 통제하는 일련의 
      업무규칙으로 엔터티, 관계, 도메인 무결성 규칙 및 연쇄 작용이 있다. 
    - 엔터티 무결성 규칙 (Entity Integrity rule)
        - 주 키는 NULL 값을 포함하지 않는다. 
        - 엔터티 내 특정 건(Instance)의 유일성을 보장할 것 
        - 최소한의 속성 집합(Minimal Set)일 것 
    - 참조 무결성 규칙 (Referential integrity rule)
        - 관계 엔터티(테이블)의 모든 외래 키 값은 관련 있는 관계 엔터티(테이블)의 모든 주 키 값이 존재해야 한다. 
        - 입력 규칙
            - 자식 엔터티의 행이 입력 될 때 
        - 삭제 규칙 
            - 부모 엔터티의 행을 삭제할 때 
    - 도메인(속성) 무결성 규칙 (Domain Integrity rule)
        - 엔터티(테이블) 내의 모든 열에 관한 무결성 규칙으로 데이터 타입, 길이, 허용 값, 기본값, 유일성, NULL 여부 등에 관한 제한
        - 같은 도메인의 값들끼리만 비교 가능 
        - 속성이 취할 수 있는 값의 제한 
    - 연쇄 작용(Triggering operation) 또는 업무 규칙 (Business Rule)
        - 입력, 삭제, 수정 또는 조회 등의 작업이 동일 엔터티(테이블) 혹은 다른 엔터티(테이블)의 속성에 영향을 미치는 업무 규칙을 정의
