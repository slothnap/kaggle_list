# 데이터베이스 설계와 이용(p531)
## 데이터베이스 설계(p533)
### 저장공간 설계(p534)
    1. 테이블
        - 가. 힙 테이블
            - 인덱스와 데이터 저장 공간이 분리되어 있는 구조
            - 대량의 트랜잭션이 발생하는 로그성 테이블은 힙 테이블을 사용하는 것이 바람직
        - 나. 클러스터형 인덱스 테이블
            - Primary Key 값이나 인덱스 키 값의 순서로 데이터가 정렬되어 저장되는 테이블을 클러스터형 인덱스
            - 데이터 입력시 정렬에 대한 부하가 수반된다. 
        - 다. 파티션 테이블
            - 파티셔닝은 대용량의 테이블을 파티션이라는 더 작은 논리적인 단위로 나눔으로써 성능이 저하되는 것을
              방지하고 관리를 수월하게 하고자 하는 기능
        - 라. 외부 테이블
            - 외부 파일을 마치 데이터베이스 안에 존재하는 일반 테이블처럼 이용할 수 있는 데이터베이스 객체 
        - 마. 임시 테이블
            - 트랜잭션이나 세션별로 데이터를 저장하고 처리할 수 있는 임시 테이블
    2. 파티션 설계
        - 가. 파티셔닝 개요
            - 정의
                - 테이블과 인덱스 데이터를 어떠한 기준의 단위로 분할하여 저장하는 것을 의미
            - 성능적 관점
                - 액세스 패턴에 따라 파티셔닝이 이루어질 수 있도록 파티션 키를 선정해야 한다. 
            - 관리적 관점
                - 이력 데이터를 파티셔닝할 경우 파티션의 생성주기와 소멸주기를 일치시켜야 한다. 
        - 나. 범위 파티션 테이블
            - 날짜 컬럼을 기준으로 데이터를 일, 월, 년 등의 기준으로 파티셔닝할 때 많이 사용한다. 
        - 다. 목록 파티션 테이블
            - 미리 정해진 코드성 컬럼의 값을 키로 파티셔닝하는 경우
        - 라. 해시 파티션 테이블
            - 성능상의 관점에서만 접근하는 것이 좋다. 
            - 파티션 키로 정의한 컬럼에 해시 함수를 적용하여 결과 값이 동일한 데이터를 같은 파티션에 저장 
        - 마. 복합 파티션 테이블
            - 범위, 목록, 해시 파티셔닝 방법을 결합하여 적용한 파티션 테이블 
        - 바. 참조 파티션 테이블
            - 자식 테이블에 부모 테이블과 동일한 컬럼이 없어도 부모 테이블과 동일한 기준으로 자식 테이블을
              파티셔닝 할 수 있다. 
        - 아. 기타 파티션 테이블
            - 범위 파티션을 자동으로 생성해 주는 인터벌 파티션
            - 사용자가 데이터를 입력하는 시점에 데이터가 저장될 파티션을 정의할 수 있는 시스템 파티션
            - 테이블 생성 시 가상 컬럼을 정의하고 가상 컬럼에 파티셔닝을 적용하는 가상 컬럼 파티션
        - 자. 파티션 제약사항
    3. 칼럼
        - 가. 데이터 타입
            - 칼럼은 데이터 타입과 길이로 정의 
        - 나. 데이터 타입과 길이 결정
            - CHARACTER vs VARCHAR 타입
                - character 반드시 여부 값을 저장하는 컬럼과 같이 입력되는 모든 데이터의 길이가 동일 및 not null
            - DATE vs CHARACTER 타입
            - 정밀도
            - 입력 데이터의 최대 길이
            - 기본 값의 정의
        - 다. 이기종 DBMS 간의 데이터 타입 변환
            - 문자 데이터 타입
            - 숫자 데이터 타입
            - 날짜 데이터 타입
    4. 용량 설계
        - 목적
            - 정확한 데이터 용량을 예측하여 저장 공간을 효과적으로 사용하고 저장 공간에 대한 확장성 보장
            - H/W 특성을 고려하여 디스크 채널 명복을 최소화하기 위함
            - 디스크 I/O를 분산하여 접근 성능을 향상하기 위함
            - 테이블이나 인덱스에 맞는 저장 옵션을 지정하기 위함
        - 테이블 저장 옵션 고려 사항
            - 초기 사이즈, 증가 사이즈
            - 최대 사이즈와 자동 증가 옵션
            - 트랜잭션 관련 옵션
        - 저장 용량 설계 절차
            - 용량 분석: 데이터 증가 예상 건수, 주기, 행 하나의 사이즈 등을 고려
            - 객체별 용량 산정 : 테이블, 인덱스에 대한 크기
            - 테이블스페이스별 용량 산정 : 테이블스페이스별 객체 용량의 합계
            - 디스크 용량 산정 : 테이블스페이스에 따른 디스크 용량과 I/O 분산 설계 
### 무결성 설계(p546)
    1. 데이터 무결성
        - 가. 데이터 무결성 종류
            - 엔터티 무결성
                - 엔터티는 각 인스턴스를 유일하게 식별할 수 있는 속성이나 그룹을 가져야 한다. 
            - 영억 무결성
                - 칼럼 데이터 타입, 길이, 유효 값이 일관되게 유지되어야 한다. 
            - 참조 무결성
                - 데이터 모델에서 정의된 엔터티 간의 관계 조건을 유지하는 것이다. 
            - 사용자 정의 무결성
                - 다양하게 정의될 수 있는 비즈니스 규칙이 데이터적으로 일관성을 유지하는 것이다. 
        - 나. 데이터 무결성 강화 방법
            - 애플리케이션
                - 데이터를 조작하는 프로그램 내에 데이터 생성, 수정, 삭제 시 무결성 조건을 검증하는 코드를 추가
                - 장점
                    - 사용자 정의 같은 복잡한 무결성 조건을 구현함
                - 단점
                    - 소스 코드에 분산되어 관리의 어려움 있음
                    - 개별적으로 시행되므로 적정성 검토에 어려움
            - 데이터베이스 트리거 
                - 트리거 이벤트 시 저장 SQL을 실행하여 무결성 조건을 실행함
                - 장점
                    - 통합 관리가 가능함
                    - 복잡한 요건 구현 가능
                - 단점
                    - 운영 중 변경이 어려움
                    - 사용상 주의가 필요함
            - 제약조건
                - 데이터베이스 제약조건 기능을 선언하여 무결성을 유지함
                - 장점
                    - 통합 관리가 가능함
                    - 간단한 선언으로 구현 가능
                    - 변경이 용이하고 유효/무효 상태 변경 가능함
                    - 원천적으로 잘못된 데이터 발생을 막을 수 있음
                - 단점
                    - 복잡한 제약조건 구현이 불가능
                    - 예외적인 처리가 불가능
    2. 엔터티 무결성
        - PK 제약조건
        - UNIQUE 제약조건
    3. 영역 무결성
        - 데이터 타입 & 길이
        - 유효 값
        - NOT NULL
    4. 참조무결성
        - 입력 참조무결성
            - DEPENDENT
            - AUTOMATIC
            - DEFAULT
            - CUSTOMIZED
            - NULL
            - NO EFFECT
        - 수정/삭제 참조무결성
            - RESTRICT
            - CASCADE
            - DEFAULT
            - CUSTOMIZED
            - NULL
            - NO EFFECT
### 인덱스 설계(p552)
    1. 인덱스 기능
    2. 인덱스 구조
        - 가. 트리 기반 인덱스
            - B+tree는 루트에서 리프 노드까지 모든 경로의 깊이가 같은 밸런스 트리
            - 대용량 처리의 데이터 입력과 삭제 등에 좋은 성능을 유지
            - 범위 검색 시에 좋은 성능을 낸다. 
            - 삭제가 빈번하게 발생하는 경우 치명적
            - OLTP
            - 데이터 분포도가 좋은 컬럼에 적합
        - 나. 해시 기반 인덱스
            - 키 값에 해시 함수를 적용하여 해당 키 값에 대응하는 버킷을 식별하고 탐색
            - 버킷에 새로운 데이터 엔트리를 위한 저장 공간이 없을 때는 새로운 오버플로우 페이지를 할당하고
              그 페이지에 데이터를 삽입하여 해당 버킷과 연결한다. 
            - 오버플로우 체인이 길어지면 버킷 탐색하는 성능이 저하되므로 초기 생성 시 80% 정도 채우고,
              주기적으로 다시 해싱해 오버플로우 페이지를 제거한다. 
        - 다. 비트맵 인덱스
            - 행에 대한 논리적 주소정보를 제공
            - 넓은 분포도를 갖고 있는 데이터에 대해 다중 조건을 만족하는 행의 개수를 계산하는 데 적합
            - DW, DM
            - 데이터 분포도가 나쁜 컬럼에 적합
        - 라. 함수 기반 인덱스
            - 함수나 수식으로 계산된 결과에 대해 B+tree 인덱스나 비트맵 인덱스를 생성하여 사용할 수 있는 기능
        - 마. 파티션 인덱스
            - 비 파티션 인덱스
                - 파티션 테이블에 생성된 비 파티션 인덱스는 글로벌 인덱스라 부르며 파티셔닝되지 않은 인덱스가
                  파티셔닝된 전체 테이블과 대응되는 구조
            - 로컬 파티션 인덱스
                - 테이블이 파티셔닝된 기준 그대로 파티셔닝된 파티션 인덱스
            - 글로벌 파티션 인덱스 
                - 테이블의 파티션과 독립적으로 인덱스 파티션을 구성하는 것
                - 글로벌 파티션 인덱스는 파티셔닝되지 않은 테이블에도 적용할 수 있다. 
    3. 인덱스 설계 절차
        - 가. 접근 경로 수집
            - 반복 수행되는 접근 경로
                - 조인 컬럼
            - 분포도가 양호한 칼럼
                - 주문번호, 청구번호, 주민등록 등과 같이 단일 컬럼 인덱스로도 충분한 수행 속도를 보장 가능한 것
                - 분포도(%) = 데이터별 평균 행(Row) 수 / 테이블의 총 행(Row) 수 X 100
            - 조회 조건에 사용되는 칼럼
                - 조회 조건으로 많이 이용되는 컬럼
            - 자주 결합되어 사용되는 칼럼
                - 조합에 의해 사용되는 컬럼 유형
            - 데이터 정렬 순서와 그루핑 칼럼
                - ORDER BY에 의한 정렬 작업을 피할 수 있다. 
            - 일련번호를 부여한 칼럼
            - 통계 자료 추출 조건
            - 조회 조건이나 조인 조건 연산자 
        - 나. 분포도 조사에 의한 후보 컬럼 선정
            - 분포도가 10 ~ 15% 정도이면 인덱스 컬럼 후보로 사용
            - 분포도가 1차적으로 단일 컬럼을 대상으로 조사 
            - 인덱스 후보는 중복이 없게 최소 조합으로 선별
            - 빈번하게 값의 변경이 발생하는 컬럼은 업무적으로 중요하지 않다면, 가능한 인덱스 후보에서 제외
        - 다. 접근 경로 결정
            - 접근 유형에 따라 어떤 인덱스 후보를 사용할 것인지 결정
        - 라. 칼럼 조합 및 순서 결정
            - 항상 사용되는 컬럼을 선두 컬럼으로 한다. 
            - 등치(=) 조건으로 사용되는 컬럼을 선행 컬럼으로 한다. 
            - 분포도가 좋은 컬럼(Distinct Value가 많은 컬럼)을 선행 컬럼으로 한다.
            - ORDER BY, GROUP BY 순서를 적용한다. 
        - 마. 적용 시험
### 분산 설계(p562)
    1. 분산 데이터베이스 개요
        - 하나의 논리적 데이터베이스가 네트워크상에서 여러 컴퓨터에 물리적으로 분산되어 있지만 사용자가 
          하나의 데이터베이스처럼 인식할 수 있도록 논리적으로 통합되어 공유되는 데이터베이스
    2. 분산 데이터베이스 관리 시스템
        - 분할 투명성
            - 사용자가 입력한 전역 질의를 여러 개의 단편 질의로 변환해 주기 때문에 사용자는 전역 스키마가
              어떻게 분할되어있는지를 알 필요가 없게 된다. 
            - 분할 방법
                - 수직 분할
                    - 한 테이블 속성들의 부분 집합으로 이루어진 테이블들로 나눈다
                - 수평 분할
                    - 행 들의 집합으로 이루어진 테이블들로 나눈 것
        - 위치 투명성
            - 사용자나 애플리케이션에서 어떤 작업을 수행하기 위해 분산 데이터베이스를 존재하는
              어떠한 데이터의 물리적인 위치도 알 필요가 없어야 한다는 것
        - 중복 투명성
            - 어떤 데이터가 중복되었는지, 또는 어디에 중복 데이터를 보관하고 있는지 사용자가 알 필요가 없다
        - 장애 투명성
            - 데이터베이스가 분산되어 있는 각 컴퓨터 시스템이나 네트워크에 장애가 발생하더라도 데이터의
              무결성이 보장되어야 한다는 것
        - 병행 투명성
            - 다수 트랜잭션이 동시에 수행되어도 결과의 일관성이 유지되어야 하는 성질
            - 분산 트랜잭션의  일관성을 유지하기 위해 잠금과 타임스탬프의 두가지 방법 사용
    3. 분산 설계 전략
        - 지역 데이터베이스에 데이터를 복제하고 실시간으로 복제본을 갱신하는 방식
        - 지역 데이터베이스에 데이터를 복제하고 주기적으로 복제본을 갱신하는 방식
        - 데이터 분할 시 전체 지역 데이터베이스를 하나의 논리적 데이터베이스로 유지하는 방식
        - 데이터 분할 시 각각의 지역 데이터베이스들을 독립된 데이터베이스로 유지하는 방식
    4. 분산 설계 방식
        - 가. 분할
            - 완전성
                - 분할 시에 전역 테이블 내의 모든 데이터가 어떠한 손질도 없이 분할조각으로 사상되어야 하는 것
                - 분할 대상이 전체 데이터를 대상으로 이뤄져야 함
            - 재구성
                - 분할조각은 관계 연산을 사용하여 원래의 전역 테이블로 재구성이 가능하여야 한다. 
            - 상호 중첩 배제
                - 상호 중첩 배제는 분할 시에 하나의 분할조각에 속한 데이터 항목이 다른 분할조각의 데이터 항목에 
                  속하지 않아야 한다는 의미
            - 수평 분할
                - 특정 컬럼의 값을 기준으로 각 행을 분할
                - 컬럼의 값에 의해 분리된 행들은 서버 간 데이터가 상호 배타적으로 존재하므로 분할조각을
                  통합하여도 식별자의 중복이 발생하지 않는다.
            - 수직 분할
                - 컬럼을 기준으로 분할을 수행
                - 행을 기준으로 분리하지 않기 때문에 각 분할조각에는 동일한 식별자를 가진 행들이 중복되어 존재
                - 동일한 식별자 구조를 가지므로 식별자를 이용하여 분할 전 테이블을 재구성 할 수 있다.
        - 나. 할당
            - 할당은 분할조각을 복수 서버에 생성하는 분산방법
            - 부분 복제
                - 전역 서버 테이블의 일부 분할조각을 지역 서버에 복제하는 방식
            - 광역 복제
                - 전역 서버 테이블의 전체를 지역 서버에 복제하는 방식
    5. 데이터 통합
### 보안 설계(p567)
    1. 접근 통제 기능
        - 임의적 접근 통제
            - 사용자의 신원에 근거를 두고 권한을 부여하고 취소하는 매커니즘 기반
            - 사용자가 어떤 객체에 대해 특정 행위를 할 수 있도록 허용하는 것
            - 객체를 생성한 사용자는 그 객체에 대해 적용 가능한 모든 권한을 가진다. 
        - 강제적 접근 통제
            - 주체와 객체를 보안 등급 중 하나로 분류하고, 주체가 자신보다 보안 등급이 높은 객체를 읽거나 쓰는거 방지
    2. 보안 모델
        - 가. 접근 통제 행렬
            - 주체
                - 데이터베이스에 접근할 수 있는 조직의 개체
                - 객체에 대하여 접근을 시도하는 사용자
            - 객체
                - 보호되고 접근이 통제되어야 하는 데이터베이스의 객체
                - 테이블, 컬럼, 뷰, 프로그램, 논리적인 정보의 단위 등
            - 규칙
                - 주체가 객체에 대하여 수행하는 데이터베이스의 조작
                - 입력, 수정, 삭제, 읽기와 객체의 생성자 파괴 등
        - 나. 기밀성 모델
            - 단순 보안 규칙
                - 주체는 자신보다 높은 등급의 객체를 읽을 수 없다.
            - *(스타)-무결성 규칙
                - 주체는 자신보다 낮은 등급의 객체에 정보를 쓸 수 없다. 
            - 강한 *(스타) 보안 규칙
                - 주체는 자신과 등급이 다른 객체에 대하여 읽거나 쓸 수 없다. 
        - 다. 무결성 모델
            - 단순 보안 규칙
                - 주체는 자신보다 높은 등급의 객체를 읽을 수 없다.
            - *(스타)-무결성 규칙
                - 주체는 자신보다 높은 등급의 객체에 정보를 쓸 수 없다. 
    3. 접근 통제 정책
        - 신분-기반 정책
            - 개인 또는 그들이 속해 있는 그룹들의 신분에 근거하여 객체에 대한 접근을 제한하는 방법
            - IBP(한 사용자가 하나의 객체에 대해 허가를 부여)
            - GBP(다수의 사용자가 하나의 객체에 대하여 동일한 허가를 부여)
        - 규칙-기반 정책
            - 주체가 갖는 권한에 근거하여 객체에 대한 접근을 제한
            - MLP(사용자 및 객체별로 부여된 기밀 분류에 따른 정책)
            - CBP(조직 내의 부서별로 구분된 기밀 허가에 따르는 정책)
        - 역할-기반 정책
            - GBP의 변경된 형태
            - 정보에 대한 사용자의 접근이 개별적인 신분이 아니라 개인의 직무 또는 직책에 따라서 결정
    4. 접근 통제 메커니즘
        - 패스워드
        - 암호화
            - 1) 운영체제에서의 암호화
                - 운영체제의 파일 시스템에 파일로 데이터가 저장
            - 2) DBMS 엔진에서의 암호화
                - 데이터를 입출력하는 과정에서 데이터를 암복호화하는 방법
            - 3) 응용프로그램에서의 암호화
                - 응용프로그램 개발 시 외부의 암호화 라이브러리를 이용하여 보안 대상 테이블/컬럼에 대해서 암호화
        - 접근 통제 목록
            - 객체를 기준으로 접근 통제 정보를 저장하는 방식
        - 능력 리스트
            - 주체가 접근할 수 있는 객체와 접근 권한을 주체에 저장하는 방식
            - 비교적 객체가 적을 경우에 적합
        - 보안등급
            - 주체나 객체 등에 부여된 보안 속성의 집합
            - 각 주체와 객체를 보안등급 중 하나로 분류
            - 데이터 접근 요청을 처리할 때 주체/객체의 보안등급을 기반으로 접근 승인 여부를 결정
        - 통합 정보 메커니즘
            - 위의 복합된 특성을 두 가지 이상 구현
    5. 접근 통제 조건
        - 값 종속 통제
            - 객체에 저장된 값에 따라 접근 통제 허용을 관리 
        - 다중 사용자 통제
            - 지정된 객체에 대해 다수의 사용자가 연합하여 접근을 요청할 경우 접근 통제를 지원하기 위한 수단
        - 컨텍스트 기반 통제
            - 특정 시간, 네트워크 주소 등 확인이 가능한 접근 경로나 위치, 인증 수준 등과 같은 외부적인

### 데이터베이스 관리 시스템(DBMS)(p578)
    1. 데이터베이스 관리 시스템 개요
        - 가. 데이터베이스 관리 시스템 추상화
            - 메타 데이터 관리
            - 데이터의 독립성
            - 데이터의 추상화
                - 사용자 입장에서 데이터베이스 관리 시스템의 물리적인 저장 구조를 알지 못해도 쉽게 사용
            - 트랜잭션과 동시성 제어
                - 데이터베이스 관리 시스템은 여러 사용자가 동시에 동일한 데이터베이스에 접근할 수 있는 기능 제공
        - 나. 데이터베이스 관리 시스템 사상
            - 개념적 단계
                - 추상화의 최상위 단계로 최종 사용자들이 관심을 갖는 데이터베이스의 부분만을 정의 
            - 논리적 단계
                - 데이터베이스의 전체 구조를 추상화하는 단계로 데이터베이스에 저장된 데이터와 데이터 간의 관계, 
                  권한, 무결성과 같은 부가적인 정보를 정의 
            - 물리적 단계
                - 가장 낮은 추상화 단계로 데이터가 실제 어떻게 저장되어 있는지 원시 수준의 데이터 구조를 정의 
            - 개념적-논리적 사상
                - 논리적 스키마에 변화가 생겨도 그 변화를 개념적-논리적 사상에 반영하면 개념적 스키마에 
                  변화가 없으므로 논리적 단계의 데이터 독립성을 유지할 수 있다. 
            - 논리적-물리적 사상
                - 데이터 파일의 이동, 구조 등의 물리적 변화가 발생해도 논리적 스키마에는 변화가 없으므로 
                  물리적 단계의 데이터 독립성을 유지할 수 있다. 
        - 다. 데이터베이스 관리 시스템 구성요소
            - 데이터베이스 파일
                - 데이터 사전을 저장하는 파일, 사용자의 데이터를 저장하는 파일
            - 데이터 저장 관리자
            - 질의 처리기 
            - 트랜잭션 관리자
                - 데이터베이스 파일에 여러 명의 사용자가 동시에 접근하여 데이터를 처리하더라도 데이터에 
                  이상 현상이 없도록 데이터의 일관성을 유지하는 책임
    2. 데이터베이스 관리 시스템 종류
            - MySQL
            - MariaDB
            - Oracle
            - SQL Server
            - PstgreSQL
            - Cubrid
    3. 데이터베이스 구조
        - 가. 데이터 딕셔너리 
            - 데이터베이스의 모든 스키마 객체 정보
            - 스키마 객체에 대해 할당된 영역의 사이즈와 현재 사용 중인 영역의 사이즈
            - 컬럼에 대한 기본 값
            - 무결성 제약조건에 대한 정보
            - 사용자 이름, 사용자에게 부여된 권한과 역할
            - 기타 일반적인 데이터베이스 정보
        - 나. 데이터베이스, 테이블스페이스 및 데이터 파일
            - 테이블이 저장되는 테이블스페이스는 업무별로 지정한다. 
            - 대용량 테이블은 독립적인 테이블스페이스를 지정한다. 
            - 테이블과 인덱스는 분리하여 저장한다. 
            - LOG 타입 데이터는 독립적인 공간을 지정한다. 
        - 다. 데이터 블록, 확장 영역 및 세그먼트 간의 관계
            - 데이터 블록
                - DBMS가 데이터를 저장하는 가장 작은단위
                - 오라클: header -> row directory -> free space -> row data
                - MSSQL: header -> row data -> free space -> row directory
            - 데이터 확장 영역
                - Extent -< Block
            - 세그먼트
                - 테이블스페이스 내에 논리적인 구조를 정의
                - 테이블, 인덱스, 임시 세그먼트가 지원 
                - Segment -< Extent
    4. 메모리 구조
            - DBMS 정보 저장
                - 실행되는 프로그램 코드
                - 현재 사용하지 않더라도 접속되어 있는 세션 정보
                - 프로그램이 실행되는 동안 필요한 정보
                - 프로세스 간에 공유하거나 교환되는 정보
                - 보조 메모리에 영구적으로 저장된 캐시 데이터 
                - 소프트웨어 코드 영역
                    - 수행될 수 있는 소프트웨어 코드 저장 영역
                - 시스템 메모리 영역
                    - 데이터베이스 버퍼 + 로그 버퍼
                - 프로그램 영역
                    - 프로세스에서 필요로 하는 데이터나 제어 정보를 저장
                - SGA
                    - Shared Pool
                    - Java Pool
                    - Redo Log Buffer
                    - Fixed SGA
                    - Large Pool
                    - DB Buffer
            - 데이터베이스 버퍼
                - 데이터 파일로부터 읽어들인 데이터 블록의 복사본을 가진다. 
                - Dirty list
                    - 수정되었지만 아직 디스크에 기록되지 않은 데이터
                - LRU (Least Recently Used)
                    - 빈 버퍼, 현재 액세스 중인 고정된 버퍼, Dirty list
                - 데이터 액세스 순서
                    - 데이터베이스 버퍼에 있는 데이터를 검색
                    - 데이터를 찾으면 메모리에서 직접 데이터를 읽는다. 
                    - 버퍼에 없으면 디스크에서 버퍼로 복사한다. 
                    - 데이터 블록은 디스크에서 읽기전에 빈 데이터베이스 버퍼를 찾는다. 
                    - 끝에서 부터 LRU 목록을 검색 후 계속 찾는다.
                    - 빈 버퍼를 찾으면 버퍼에 복사후 LRU 목록 끝으로 이동
            - 로그 버퍼
                - 데이터베이스의 변경 사항 정보를 유지하는 것
                - insert, delete, create, alter , drop으로 변경된 사항을 REDO 입력 항목을 가진다. 
                - REDO 입력 항목은 데이터베이스 복구에 사용 
            - 공유 풀
                - 라이브러리 캐시
                    - 수행 SQL 저장
                    - 사용자 정의 함수/저장 프로시저/트리거 저장
                - 딕셔너리 캐시
                    - 데이터 딕셔너리 정보를 공유
                - 제어 구조
            - 그외 영역
                - Fixed Area
                    - DBMS의 환경 변ㅂ수를 저장 
                - Large Pool
                    - 대량의 데이터가 로딩되었을 때 임시적으로 활용하기 위한 영역
                - Java Pool
                    - 오라클에서 Java Stored Procedure를 실행하기 위한 영역
                - PGA
                    - 데이터 정렬을 위한 Sort Area, Hash Area 공간
    5. 프로세스 구조
            - 사용자 프로세스 
            - 서버 프로세스
            - 백그라운드 프로세스
                - PMON
                    - 사용자 프로세스에 장애가 발생하면 프로세스 복구를 수행
                - SMON
                    - 인스턴스 시작 시 필요한 경우 고장 복구를 수행, 임시 세그먼트 정리 수행
                - DBWn
                    - 버퍼의 내용을 데이터 파일에 기록
                - CKPT
                    - 체크 포인트가 발생하면 데이터 파일의 헤더 갱신
                - LGWR
                    - 로그 버퍼를 관리하여 로그 버퍼를 디스크의 로그파일에 기록

### 트랜잭션(p596)
    1. 트랜잭션 관리
        - 동시서 제어
        - 고장 획복 
    2. 트랜잭션 특성
        - 원장성 (Atomicity)
            - 트랜잭션은 완전히 수행되거나 전혀 수행되지 않은 상태
        - 일관성 유지 (Consistency)
            - 데이터베이스가 하나의 일관된 상태에서 또 다른 일관된 상태로 바뀐다. 
        - 고립성 (Isolation)
            - 나의 트랜잭션이 완료될 때까지 자신이 갱신한 값들을 다른 트랜잭션들이 보게 해서는 안된다. 
        - 영속성 (Durability)
            - 트랜잭션으로 변경이 완료되면 변경된 결과는 이후의 어떠한 고장에도 손실되지 않아야 한다. 
    3. 트랜잭션의 일관성
        - 가. 낮은 단계 트랜잭션 고립화 수준에서 발생할 수 있는 현상들
            - Dirty Read (= Uncommitted Dependency)
                - 다른 트랜잭션이 변경 중인 데이터를 읽었을 때 Rollback을 수행하여 현재 트랜잭션과 
                  비일관성 상태 
            - Non-Repeatable Read (= Inconsistent Analysis)
                - 한 트랜잭션 내에서 동일 SQL를 두 번 수행할 때 
                  다른 트랜잭션이 값을 수정 또는 삭제함으로써 두 SQL의 결과가 상이하게 나타남
            - Phantom Read
                - 동일 SQL을 두 번 수행할 떄 첫번쨰 수행시 존재하지 않던 유령 레코드가
                  두 번째 SQL 수행 시 나타나는 현상
        - 나. 트랜잭션 고립화 수준
            - 레벨 0 (= Read Uncommitted)
                - 트랜잭션에서 처리 중인 아직 Commit되지 않은 데이터를 다른 트랜잭션이 읽는 것을 허용
                - Dirty Read, Non-Repeatable Read, Phantom Read 현상 발생
            - 레벨 1 (= Read Committed)
                - 트랜잭션이 Commit되어 확정된 데이터만 읽는 것을 허용
                - Non-Repeatable Read, Phantom Read 현상 발생
            - 레벨 2 (= Repeatable Read)
                - 선행 트랜잭션이 읽은 데이터는 트랜잭션이 종료될 떄까지 수행 트랜잭션이 갱신하거나
                  삭제하는 것을 불허함으로써 같은 데이터를 두 번 질의했을 때 일관성 있는 결과를 반환
                - Phantom Read 현상 발생
            - 레벨 3 (= Serializable Read)
                - 선행 트랜잭션이 읽은 데이터를 후행 트랜잭션이 갱신하거나 삭제하지 못할 뿐만 아니라
                  중간에 새로운 레코드를 입력하는 것도 막아줌으로써 완벽한 읽기 일관성을 제공
                - 오라클은 Undo 데이터를 이용해서 Serializable Read를 제공
    4. 동시성 제어
        - 정의
            - 다수의 사용자가 데이터베이스에 동시에 접근하여 같은 데이터를 조회 또는 갱신할 때
              데이터 일관성을 유지하기 위한 조치 
        - 가. 낙관적 동시성 제어
            - 다수 사용자가 동시에 같은 데이터에 접근할 경우가 적다고 생각
            - 사용자들이 동일 데이터를 동시에 수정하지 않을 것
            - 잡금(Locking)을 설정하지 않는다.
            - 동시성 증가 유리
        - 나. 비관적 동시성 제어
            - 다수 사용자가 동시에 같은 데이터에 접근할 경우가 많다고 생각
            - 사용자들이 같은 데이터를 동시에 수정할 것이라고 가정
            - 잡금(Lcking)을 설정
    5. 동시성 제어 기능
            - 잠금 단위
                - 자금 대상의 크기
            - 잠근 확산
                - 자금 단위의 개수를 설정
            - 잠금의 유형
                - 공용 잠금 (Shared Lock)
                - 배타적 잠금 (Exclusive Lock)
            - RS
                - row share
            - RX
                - row exclusive
            - S
                - share
            - SRX
                - share row exclusive
            - X
                - exclusive
        - 2PC
            - 2개 이상의 트랜잭션들이 병행 처리되었을 떄의 데이터베이스 결과는 그 트랜잭션들을 임의의
              직렬적인 순서로 처리했을 떄의 결과와 논리적으로 일치해야 한다. 
        - 교착 상태
            - 정의
                - 다른 사용자가 잠근 자원이 해제되기를 기다리면서 자신이 잠금 자원을 해제하지 않는 상태
            - 상호 배제 (Mutual Exclusive)
                - 어느 자원에 대해 한 프로세스가 이미 사용 중이면 다른 프로세스는 기다려야 하는 것
            - 점유와 대기 (Wait for)
                - 하나 이상의 자원을 할당받은 채로 나머지 자원을 할당받기 위해 다른 프로세스의 자원이
                  해제되기를 기다리는 프로세스가 존재하는 경우
            - 비중단 (no Preemption)
                - 자원을 할당받은 프로세스로부터 자원을 강제로 빼앗지 못하는 것
            - 환형 대기 (Circular wait)
                - 자원 할당 그래프 상에서 프로세스의 환형 사슬이 존재하는 것
    6. 동시성 구현 사례
    7. 고장 회복
    8. 자금 지속 시간

### 애플리케이션 성능 개선(p611)
    1. 실행 구조
        - 가. 옵티마이저
            - 조인 순서, 조인 방법, 테이블 액세스 방법을 선택하는 것을 실행 계획이라고 한다. 
            - 옵티마이저는 가능한 실행 계획들을 검토하고 이 중에서 가장 효과적인 것을 결정
        - 나. SQL 실행 단계
            - 파싱 단계
                - SQL의 구문과 의미가 정확한지 검사하고, 참조된 테이블에 대해 사용자의 접근 권한 검사
            - 옵티마이저 단계
                - 결과 정보를 이용해 최적의 실행 계획을 수립
            - 행 소스 생성 단계
                - 옵티마이저에게 넘겨받은 실행 계획을 내부적으로 처리하는 자세한 방법을 생성하는 단계
            - SQL 실행
                - 생성된 Row Source를 SQL 수행 엔진에서 수행해 결과를 사용자에게 돌려주는 과정
                - 소프트 파싱
                    - 이미 최적화를 한 번 수행한 SQL 질의에 대해 옵티마이저 단계와 Row Source 생성 단계를 생략
                - 하드 파싱
                    - 최적화와 Row Source 생성 단계를 새로 수행
    2. 조인
        - 가. Nested-Loop 조인
            - 조인 조건
                - Indexed Nested-Loop 조인으로 수행되려면 Inner 테이블에 조인 조건으로 사용된 컬럼에 
                  반드시 인덱스가 존재해야 한다. 
            - 출력 순서
                - Nested-Loop 조인에서 결과 집합이 출력되는 순서는 드라이빙 테이블을 액세스한 순서와 동일
            - 조인 순서
                - 옵티마이저는 조인 시에 먼저 드라이빙 테이블을 결정하고 나머지 집합의 조인 순서를 결정
                - 드라이빙 테이블은 범위가 가장 작은 집합으로 선정
        - 나. Sort-Merge 조인
            - 조인하려는 두 집합을 조인 속성으로 정렬하여 Sorted List를 만든 후 이들을 병합하는 조인 기법
            - 조인 조건
                - 정렬 단계와 병합 단계로 나누어 수행
            - 출력 및 연결 순서
                - 정렬된 결과를 순차적으로 비교하여 병합을 수행하므로 Outer 집합의 정렬 순서와 Inner 집합의 
                  정렬 순서가 병합되어 출력된다. 
            - 조건절
            - 이용
                - 독립적으로 처리 범위를 줄인 후 조인에 참여하므로 테이블 각각의 조건에 의해 대상 집합을 
                  줄일 수 있을 때 유리하다. 
                - 처리 대상이 전체 테이블일 때 랜덤 I/O 부하가 큰 Nested-Loop 조인보다 유리하다. 
                - 조인 속성에 인덱스가 없을 때 Simple Nested-Loop 조인보다 성능이 우수하다. 
                - 효과적인 수행을 위해서는 적정한 정렬 영역 사이즈가 확보되어야 한다. 
                - 정렬에 대한 부하가 많이 발생하므로 대용량 처리 시 수행 속도가 저하될 수 있다. 
        - 다. Hash 조인
            - 정의
                - 관계형 데이터베이스에서 비용이 가장 많이 들어가는 조인 방법
                - 대용량의 데이터 조인시에 Sort-Merge 조인이나 Nested-Loop 조인보다 더 좋은 성능
            - 조인 알고리즘 단계
                - 파티션 단계
                    - 조인 대상 테이블 데이터들의 조인 칼럼에 해시함수를 적용하고 반환된 해시 값에 따라
                      동적으로 파티셔닝을 하고 디스크에 기록
                - 조인 단계
                    - 둘 중 더 작은 파티션을 Build Input으로 지정하여 메인 메모리에 해시테이블로 로딩하고 
                      다른 파티션을 Probe Input으로 하여 데이터를 읽으면서 Hash 조인을 수행
    3. 온라인 프로그램 성능 개선
        - 가. 온라인 프로그램 성능 개선 방안
            - 온라인 프로그램 특징
                - 화면 조회가 가능할 정도로 1회 조회 데이터가 소량이다. 
                - 신속한 트랜잭션 처리가 요구된다. 
                - 조회 조건이 단순하다. 
                - 업무 형태에 따라 데이터 액세스 패턴이 고정되어 있다. 
            - 성능 개선 작업 고려 
                - 사용 빈도가 높은 SQL문을 개선하는 것이 효과적
                - 인덱스를 이용하여 데이터 액세스 범위를 줄이는 것이 효과적
                - 부분 범위 처리로 응답 시간을 단축
                - 부분 범위 처리를 하기 위해서 Nested-Loop 조인과 인덱스를 이용한 정렬을 유도
                - 장기 트랜잭션 처리를 억제 
        - 나. 온라인 프로그램의 성능 개선 사례
            - 1) 상수 바인딩에 의해 발생되는 파싱 부하
            - 2) 웹 게시글 형태의 인터페이스 시 부분 범위 처리
            - 3) 과다한 함수 사용으로 부하 발생
    4. 배치 프로그램의 성능 방안
        - 배치 작업 이슈
            - 절대 수행 시간 부족
            - 수행 결과 검증 시간 확보의 어려움
            - 오류에 따른 재처리 시간 확보가 불가능
            - 미완료 시 대안 제시에 어려움
            - 미처리 또는 지연으로 파급되는 문제 해결에 장시간 소요됨
        - 가. 배치 프로그램 성능 개선 방안
            - 1) 절차적인 처리 방식의 비효율
                - SQL이 루프 내에서 반복적으로 수행되는 구조미으몰 DBMS Call이 과도하게 발생
                - 단위 SQL이 반복적으로 수행되는 구조이므로 Random I/O 발생이 증가
                - 동일 데이터를 반복해서 읽는다. 
                - 업무 규칙을 절차적으로 구성하였기 때문에 업무 규칙이 변경되면 프로그램 구조의 수정이 불가피
                - 다수의 단위 SQL로 구성되어 있어 개별적인 단위 SQL의 개선만 가능
            - 2) 절차적인 처리 방식의 보완 요소
                - 이중 커서 사용을 하지 않고 조인을 이용하여 단일 커서를 사용
                - 동일 모듈 내에서는 같은 데이터를 2회 이상 읽지 않게 프로그램을 구조화
                - 최소한 개별적인 SQL 단위 비효율은 제거
            - 3) 집합적인 처리 방식의 고려 사항
                - 대용량 데이터를 처리해야 하는 경우 Hash 조인을 사용하는 것이 유리
                - 분포도가 나쁘면 Random I/O 비효율이 급속도로 증가 인덱스 스캔보다 Full Table Scan 방식이 유리
                - 절대적인 작업량이 정해져 있는 대용량 데이터 처리 시 병행 처리를 사용하여 시간을 단축
                - Hash 조인이나 집계를 위한 소트 작업을 고려하여 추가 메모리를 세션에 할당
                - 집합 처리에 의한 작업은 병행 처리, Full Table Scan, 메모리 영역 확보 등으로 
                  짧은 시간에 데이터베이스의 자원을 확보하여 처리를 해야 한다. 
            - 4) 분석 함수를 통한 성능 개선 방안
                - Ranking Family
                    - 대상 집합에 대하여 특정 컬럼 기준으로 순위나 등급을 매기는 분석 함수 그룹
                    - RANK, DENSE_RANK, CUME_DIST, PERCENT_RANK, NTILE, ROW_NUMBER
                - Window Aggregate Family
                    - SUM, AVG, MIN, MAX, STDDEV, VARIANCE, COUNT, FIRST_VALUE, LAST_VALUE
                - Reporting Aggregrate Family
                - LEAD/LAG Family
            - 5) 파티션 스토리지 전략을 통한 성능 향상 방안
                - 데이터 액세스 시에 파티션 단위로 액세스 범위를 줄여 I/O에 대한 성능 향상을 가져올 수 있다.
                - 여러 분할 영역으로 나눔으로써 전체 데이터의 훼손 가능성이 감소하고 데이터의 가용성이 향상
                - 각 파티션 영역을 독립적으로 백업하고 복구할 수 있다. 
                - 디스크 스트라이핑으로 I/O 성능을 향상(디스크 암에 대한 경합의 감소) 시킬 수 있다. 
        - 나. 배치 프로그램 성능 개선 사례
            - 1) 절차적 처리 방식의 개선
                - One SQL을 사용하는 것이 유리
                - 반복문을 쓸때도 COMMIT이 개별 레코드 처리 떄마다 반복수행되지 않게 한다. 
### 서버 성능 개선(p630)
    1. 객체 튜닝
        - 객체는 성능을 고려하여 설계되어야 한다. 
        - 저장 장치를 이루는 블록, 확장 영역, 세그먼트에 관련된 사항을 튜닝한다. 
        - 인덱스는 삭제, 갱신으로 인해 저장 공간 효율성이 떨어져 비효율이 발생하는 경우 인덱스 재구성 작업을 한다
        - I/O 병목이 발생하지 않게 물리적인 배치를 실시한다. 
    2. 인스턴스 튜닝
        - 메모리
            - 메모리는 Buffer 캐시, Library 캐시 등의 히트율(HIT ratio)에 의해서 평가하여 조정한다. 
            - Sort Area, Hash Area는 스와핑 발생 여부에 따라 사이즈를 결정
            - 특정한 대형 작업은 작업을 실시하는 세션에서 조정하여 작업을 실시
        - 프로세스
            - 대부분의 DBMS가 다중 프로세스 시스템이고 필요에 따라서 추가적인 프로세스 가동이 가능하다. 
        - Latch 경합
            - 트랜잭션 처리를 위한 경합이 발생한다. 
            - 객체 생성이나 변경 등으로 경합이 발생할 수 있다. 
    3. 환경 튜닝
        - CPU
            - 성능 평가하기 위한 기준으로 CPU 사용율을 평가한다. 
        - 메모리 튜닝
        - I/O 튜닝
            - 데이터베이스 병목은 I/O에 의해서 발생한다. 
            - 물리적인 디스크와 디스크 채널을 분산하므로 성능을 개선할 수 있다. 
            - 읽기/쓰기 작업에 따른 분산이 필요하다 
            - Block(File System)보다는 Raw Device가 I/O 성능에 유리하다. 
        - 네트워크 튜닝
            - 최근 네트워크 전송속도의 경우 1Gbps가 보편화되었으나, 많은 양의 데이터를 빈번하게 전송해야 한다면
              전송속도와 대역폭을 증가시키는 것을 고려한다.
