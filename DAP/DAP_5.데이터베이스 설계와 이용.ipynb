# 데이터베이스 설계와 이용(p531)
## 데이터베이스 설계(p533)
### 저장공간 설계(p534)
    1. 테이블
        - 가. 힙 테이블
            - 인덱스와 데이터 저장 공간이 분리되어 있는 구조
            - 대량의 트랜잭션이 발생하는 로그성 테이블은 힙 테이블을 사용하는 것이 바람직
        - 나. 클러스터형 인덱스 테이블
            - Primary Key 값이나 인덱스 키 값의 순서로 데이터가 정렬되어 저장되는 테이블을 클러스터형 인덱스
            - 데이터 입력시 정렬에 대한 부하가 수반된다. 
        - 다. 파티션 테이블
            - 파티셔닝은 대용량의 테이블을 파티션이라는 더 작은 논리적인 단위로 나눔으로써 성능이 저하되는 것을
              방지하고 관리를 수월하게 하고자 하는 기능
        - 라. 외부 테이블
            - 외부 파일을 마치 데이터베이스 안에 존재하는 일반 테이블처럼 이용할 수 있는 데이터베이스 객체 
        - 마. 임시 테이블
            - 트랜잭션이나 세션별로 데이터를 저장하고 처리할 수 있는 임시 테이블
    2. 파티션 설계
        - 가. 파티셔닝 개요
            - 정의
                - 테이블과 인덱스 데이터를 어떠한 기준의 단위로 분할하여 저장하는 것을 의미
            - 성능적 관점
                - 액세스 패턴에 따라 파티셔닝이 이루어질 수 있도록 파티션 키를 선정해야 한다. 
            - 관리적 관점
                - 이력 데이터를 파티셔닝할 경우 파티션의 생성주기와 소멸주기를 일치시켜야 한다. 
        - 나. 범위 파티션 테이블
            - 날짜 컬럼을 기준으로 데이터를 일, 월, 년 등의 기준으로 파티셔닝할 때 많이 사용한다. 
        - 다. 목록 파티션 테이블
            - 미리 정해진 코드성 컬럼의 값을 키로 파티셔닝하는 경우
        - 라. 해시 파티션 테이블
            - 성능상의 관점에서만 접근하는 것이 좋다. 
            - 파티션 키로 정의한 컬럼에 해시 함수를 적용하여 결과 값이 동일한 데이터를 같은 파티션에 저장 
        - 마. 복합 파티션 테이블
            - 범위, 목록, 해시 파티셔닝 방법을 결합하여 적용한 파티션 테이블 
        - 바. 참조 파티션 테이블
            - 자식 테이블에 부모 테이블과 동일한 컬럼이 없어도 부모 테이블과 동일한 기준으로 자식 테이블을
              파티셔닝 할 수 있다. 
        - 아. 기타 파티션 테이블
            - 범위 파티션을 자동으로 생성해 주는 인터벌 파티션
            - 사용자가 데이터를 입력하는 시점에 데이터가 저장될 파티션을 정의할 수 있는 시스템 파티션
            - 테이블 생성 시 가상 컬럼을 정의하고 가상 컬럼에 파티셔닝을 적용하는 가상 컬럼 파티션
        - 자. 파티션 제약사항
    3. 칼럼
        - 가. 데이터 타입
            - 칼럼은 데이터 타입과 길이로 정의 
        - 나. 데이터 타입과 길이 결정
            - CHARACTER vs VARCHAR 타입
                - character 반드시 여부 값을 저장하는 컬럼과 같이 입력되는 모든 데이터의 길이가 동일 및 not null
            - DATE vs CHARACTER 타입
            - 정밀도
            - 입력 데이터의 최대 길이
            - 기본 값의 정의
        - 다. 이기종 DBMS 간의 데이터 타입 변환
            - 문자 데이터 타입
            - 숫자 데이터 타입
            - 날짜 데이터 타입
    4. 용량 설계
        - 목적
            - 정확한 데이터 용량을 예측하여 저장 공간을 효과적으로 사용하고 저장 공간에 대한 확장성 보장
            - H/W 특성을 고려하여 디스크 채널 명복을 최소화하기 위함
            - 디스크 I/O를 분산하여 접근 성능을 향상하기 위함
            - 테이블이나 인덱스에 맞는 저장 옵션을 지정하기 위함
        - 테이블 저장 옵션 고려 사항
            - 초기 사이즈, 증가 사이즈
            - 최대 사이즈와 자동 증가 옵션
            - 트랜잭션 관련 옵션
        - 저장 용량 설계 절차
            - 용량 분석: 데이터 증가 예상 건수, 주기, 행 하나의 사이즈 등을 고려
            - 객체별 용량 산정 : 테이블, 인덱스에 대한 크기
            - 테이블스페이스별 용량 산정 : 테이블스페이스별 객체 용량의 합계
            - 디스크 용량 산정 : 테이블스페이스에 따른 디스크 용량과 I/O 분산 설계 
### 무결성 설계(p546)
    1. 데이터 무결성
        - 가. 데이터 무결성 종류
            - 엔터티 무결성
                - 엔터티는 각 인스턴스를 유일하게 식별할 수 있는 속성이나 그룹을 가져야 한다. 
            - 영억 무결성
                - 칼럼 데이터 타입, 길이, 유효 값이 일관되게 유지되어야 한다. 
            - 참조 무결성
                - 데이터 모델에서 정의된 엔터티 간의 관계 조건을 유지하는 것이다. 
            - 사용자 정의 무결성
                - 다양하게 정의될 수 있는 비즈니스 규칙이 데이터적으로 일관성을 유지하는 것이다. 
        - 나. 데이터 무결성 강화 방법
            - 애플리케이션
                - 데이터를 조작하는 프로그램 내에 데이터 생성, 수정, 삭제 시 무결성 조건을 검증하는 코드를 추가
                - 장점
                    - 사용자 정의 같은 복잡한 무결성 조건을 구현함
                - 단점
                    - 소스 코드에 분산되어 관리의 어려움 있음
                    - 개별적으로 시행되므로 적정성 검토에 어려움
            - 데이터베이스 트리거 
                - 트리거 이벤트 시 저장 SQL을 실행하여 무결성 조건을 실행함
                - 장점
                    - 통합 관리가 가능함
                    - 복잡한 요건 구현 가능
                - 단점
                    - 운영 중 변경이 어려움
                    - 사용상 주의가 필요함
            - 제약조건
                - 데이터베이스 제약조건 기능을 선언하여 무결성을 유지함
                - 장점
                    - 통합 관리가 가능함
                    - 간단한 선언으로 구현 가능
                    - 변경이 용이하고 유효/무효 상태 변경 가능함
                    - 원천적으로 잘못된 데이터 발생을 막을 수 있음
                - 단점
                    - 복잡한 제약조건 구현이 불가능
                    - 예외적인 처리가 불가능
    2. 엔터티 무결성
        - PK 제약조건
        - UNIQUE 제약조건
    3. 영역 무결성
        - 데이터 타입 & 길이
        - 유효 값
        - NOT NULL
    4. 참조무결성
        - 입력 참조무결성
            - DEPENDENT
            - AUTOMATIC
            - DEFAULT
            - CUSTOMIZED
            - NULL
            - NO EFFECT
        - 수정/삭제 참조무결성
            - RESTRICT
            - CASCADE
            - DEFAULT
            - CUSTOMIZED
            - NULL
            - NO EFFECT
### 인덱스 설계(p552)
    1. 인덱스 기능
    2. 인덱스 구조
        - 가. 트리 기반 인덱스
            - B+tree는 루트에서 리프 노드까지 모든 경로의 깊이가 같은 밸런스 트리
            - 대용량 처리의 데이터 입력과 삭제 등에 좋은 성능을 유지
            - 범위 검색 시에 좋은 성능을 낸다. 
            - 삭제가 빈번하게 발생하는 경우 치명적
            - OLTP
            - 데이터 분포도가 좋은 컬럼에 적합
        - 나. 해시 기반 인덱스
            - 키 값에 해시 함수를 적용하여 해당 키 값에 대응하는 버킷을 식별하고 탐색
            - 버킷에 새로운 데이터 엔트리를 위한 저장 공간이 없을 때는 새로운 오버플로우 페이지를 할당하고
              그 페이지에 데이터를 삽입하여 해당 버킷과 연결한다. 
            - 오버플로우 체인이 길어지면 버킷 탐색하는 성능이 저하되므로 초기 생성 시 80% 정도 채우고,
              주기적으로 다시 해싱해 오버플로우 페이지를 제거한다. 
        - 다. 비트맵 인덱스
            - 행에 대한 논리적 주소정보를 제공
            - 넓은 분포도를 갖고 있는 데이터에 대해 다중 조건을 만족하는 행의 개수를 계산하는 데 적합
            - DW, DM
            - 데이터 분포도가 나쁜 컬럼에 적합
        - 라. 함수 기반 인덱스
            - 함수나 수식으로 계산된 결과에 대해 B+tree 인덱스나 비트맵 인덱스를 생성하여 사용할 수 있는 기능
        - 마. 파티션 인덱스
            - 비 파티션 인덱스
                - 파티션 테이블에 생성된 비 파티션 인덱스는 글로벌 인덱스라 부르며 파티셔닝되지 않은 인덱스가
                  파티셔닝된 전체 테이블과 대응되는 구조
            - 로컬 파티션 인덱스
                - 테이블이 파티셔닝된 기준 그대로 파티셔닝된 파티션 인덱스
            - 글로벌 파티션 인덱스 
                - 테이블의 파티션과 독립적으로 인덱스 파티션을 구성하는 것
                - 글로벌 파티션 인덱스는 파티셔닝되지 않은 테이블에도 적용할 수 있다. 
    3. 인덱스 설계 절차
        - 가. 접근 경로 수집
            - 반복 수행되는 접근 경로
                - 조인 컬럼
            - 분포도가 양호한 칼럼
                - 주문번호, 청구번호, 주민등록 등과 같이 단일 컬럼 인덱스로도 충분한 수행 속도를 보장 가능한 것
                - 분포도(%) = 데이터별 평균 행(Row) 수 / 테이블의 총 행(Row) 수 X 100
            - 조회 조건에 사용되는 칼럼
                - 조회 조건으로 많이 이용되는 컬럼
            - 자주 결합되어 사용되는 칼럼
                - 조합에 의해 사용되는 컬럼 유형
            - 데이터 정렬 순서와 그루핑 칼럼
                - ORDER BY에 의한 정렬 작업을 피할 수 있다. 
            - 일련번호를 부여한 칼럼
            - 통계 자료 추출 조건
            - 조회 조건이나 조인 조건 연산자 
        - 나. 분포도 조사에 의한 후보 컬럼 선정
            - 분포도가 10 ~ 15% 정도이면 인덱스 컬럼 후보로 사용
            - 분포도가 1차적으로 단일 컬럼을 대상으로 조사 
            - 인덱스 후보는 중복이 없게 최소 조합으로 선별
            - 빈번하게 값의 변경이 발생하는 컬럼은 업무적으로 중요하지 않다면, 가능한 인덱스 후보에서 제외
        - 다. 접근 경로 결정
            - 접근 유형에 따라 어떤 인덱스 후보를 사용할 것인지 결정
        - 라. 칼럼 조합 및 순서 결정
            - 항상 사용되는 컬럼을 선두 컬럼으로 한다. 
            - 등치(=) 조건으로 사용되는 컬럼을 선행 컬럼으로 한다. 
            - 분포도가 좋은 컬럼(Distinct Value가 많은 컬럼)을 선행 컬럼으로 한다.
            - ORDER BY, GROUP BY 순서를 적용한다. 
        - 마. 적용 시험
### 분산 설계(p562)
    1. 분산 데이터베이스 개요
        - 하나의 논리적 데이터베이스가 네트워크상에서 여러 컴퓨터에 물리적으로 분산되어 있지만 사용자가 
          하나의 데이터베이스처럼 인식할 수 있도록 논리적으로 통합되어 공유되는 데이터베이스
    2. 분산 데이터베이스 관리 시스템
        - 분할 투명성
            - 사용자가 입력한 전역 질의를 여러 개의 단편 질의로 변환해 주기 때문에 사용자는 전역 스키마가
              어떻게 분할되어있는지를 알 필요가 없게 된다. 
            - 분할 방법
                - 수직 분할
                    - 한 테이블 속성들의 부분 집합으로 이루어진 테이블들로 나눈다
                - 수평 분할
                    - 행 들의 집합으로 이루어진 테이블들로 나눈 것
        - 위치 투명성
            - 사용자나 애플리케이션에서 어떤 작업을 수행하기 위해 분산 데이터베이스를 존재하는
              어떠한 데이터의 물리적인 위치도 알 필요가 없어야 한다는 것
        - 중복 투명성
            - 어떤 데이터가 중복되었는지, 또는 어디에 중복 데이터를 보관하고 있는지 사용자가 알 필요가 없다
        - 장애 투명성
            - 데이터베이스가 분산되어 있는 각 컴퓨터 시스템이나 네트워크에 장애가 발생하더라도 데이터의
              무결성이 보장되어야 한다는 것
        - 병행 투명성
            - 다수 트랜잭션이 동시에 수행되어도 결과의 일관성이 유지되어야 하는 성질
            - 분산 트랜잭션의  일관성을 유지하기 위해 잠금과 타임스탬프의 두가지 방법 사용
    3. 분산 설계 전략
        - 지역 데이터베이스에 데이터를 복제하고 실시간으로 복제본을 갱신하는 방식
        - 지역 데이터베이스에 데이터를 복제하고 주기적으로 복제본을 갱신하는 방식
        - 데이터 분할 시 전체 지역 데이터베이스를 하나의 논리적 데이터베이스로 유지하는 방식
        - 데이터 분할 시 각각의 지역 데이터베이스들을 독립된 데이터베이스로 유지하는 방식
    4. 분산 설계 방식
        - 가. 분할
            - 완전성
                - 분할 시에 전역 테이블 내의 모든 데이터가 어떠한 손질도 없이 분할조각으로 사상되어야 하는 것
                - 분할 대상이 전체 데이터를 대상으로 이뤄져야 함
            - 재구성
                - 분할조각은 관계 연산을 사용하여 원래의 전역 테이블로 재구성이 가능하여야 한다. 
            - 상호 중첩 배제
                - 상호 중첩 배제는 분할 시에 하나의 분할조각에 속한 데이터 항목이 다른 분할조각의 데이터 항목에 
                  속하지 않아야 한다는 의미
            - 수평 분할
                - 특정 컬럼의 값을 기준으로 각 행을 분할
                - 컬럼의 값에 의해 분리된 행들은 서버 간 데이터가 상호 배타적으로 존재하므로 분할조각을
                  통합하여도 식별자의 중복이 발생하지 않는다.
            - 수직 분할
                - 컬럼을 기준으로 분할을 수행
                - 행을 기준으로 분리하지 않기 때문에 각 분할조각에는 동일한 식별자를 가진 행들이 중복되어 존재
                - 동일한 식별자 구조를 가지므로 식별자를 이용하여 분할 전 테이블을 재구성 할 수 있다.
        - 나. 할당
            - 할당은 분할조각을 복수 서버에 생성하는 분산방법
            - 부분 복제
                - 전역 서버 테이블의 일부 분할조각을 지역 서버에 복제하는 방식
            - 광역 복제
                - 전역 서버 테이블의 전체를 지역 서버에 복제하는 방식
    5. 데이터 통합
### 보안 설계(p567)
    1. 접근 통제 기능
        - 임의적 접근 통제
            - 사용자의 신원에 근거를 두고 권한을 부여하고 취소하는 매커니즘 기반
            - 사용자가 어떤 객체에 대해 특정 행위를 할 수 있도록 허용하는 것
            - 객체를 생성한 사용자는 그 객체에 대해 적용 가능한 모든 권한을 가진다. 
        - 강제적 접근 통제
            - 주체와 객체를 보안 등급 중 하나로 분류하고, 주체가 자신보다 보안 등급이 높은 객체를 읽거나 쓰는거 방지
    2. 보안 모델
        - 가. 접근 통제 행렬
            - 주체
                - 데이터베이스에 접근할 수 있는 조직의 개체
                - 객체에 대하여 접근을 시도하는 사용자
            - 객체
                - 보호되고 접근이 통제되어야 하는 데이터베이스의 객체
                - 테이블, 컬럼, 뷰, 프로그램, 논리적인 정보의 단위 등
            - 규칙
                - 주체가 객체에 대하여 수행하는 데이터베이스의 조작
                - 입력, 수정, 삭제, 읽기와 객체의 생성자 파괴 등
        - 나. 기밀성 모델
            - 단순 보안 규칙
                - 주체는 자신보다 높은 등급의 객체를 읽을 수 없다.
            - *(스타)-무결성 규칙
                - 주체는 자신보다 낮은 등급의 객체에 정보를 쓸 수 없다. 
            - 강한 *(스타) 보안 규칙
                - 주체는 자신과 등급이 다른 객체에 대하여 읽거나 쓸 수 없다. 
        - 다. 무결성 모델
            - 단순 보안 규칙
                - 주체는 자신보다 높은 등급의 객체를 읽을 수 없다.
            - *(스타)-무결성 규칙
                - 주체는 자신보다 높은 등급의 객체에 정보를 쓸 수 없다. 
    3. 접근 통제 정책
        - 신분-기반 정책
            - 개인 또는 그들이 속해 있는 그룹들의 신분에 근거하여 객체에 대한 접근을 제한하는 방법
            - IBP(한 사용자가 하나의 객체에 대해 허가를 부여)
            - GBP(다수의 사용자가 하나의 객체에 대하여 동일한 허가를 부여)
        - 규칙-기반 정책
            - 주체가 갖는 권한에 근거하여 객체에 대한 접근을 제한
            - MLP(사용자 및 객체별로 부여된 기밀 분류에 따른 정책)
            - CBP(조직 내의 부서별로 구분된 기밀 허가에 따르는 정책)
        - 역할-기반 정책
            - GBP의 변경된 형태
            - 정보에 대한 사용자의 접근이 개별적인 신분이 아니라 개인의 직무 또는 직책에 따라서 결정
    4. 접근 통제 메커니즘
        - 패스워드
        - 암호화
            - 1) 운영체제에서의 암호화
                - 운영체제의 파일 시스템에 파일로 데이터가 저장
            - 2) DBMS 엔진에서의 암호화
                - 데이터를 입출력하는 과정에서 데이터를 암복호화하는 방법
            - 3) 응용프로그램에서의 암호화
                - 응용프로그램 개발 시 외부의 암호화 라이브러리를 이용하여 보안 대상 테이블/컬럼에 대해서 암호화
        - 접근 통제 목록
            - 객체를 기준으로 접근 통제 정보를 저장하는 방식
        - 능력 리스트
            - 주체가 접근할 수 있는 객체와 접근 권한을 주체에 저장하는 방식
            - 비교적 객체가 적을 경우에 적합
        - 보안등급
            - 주체나 객체 등에 부여된 보안 속성의 집합
            - 각 주체와 객체를 보안등급 중 하나로 분류
            - 데이터 접근 요청을 처리할 때 주체/객체의 보안등급을 기반으로 접근 승인 여부를 결정
        - 통합 정보 메커니즘
            - 위의 복합된 특성을 두 가지 이상 구현
    5. 접근 통제 조건
        - 값 종속 통제
            - 객체에 저장된 값에 따라 접근 통제 허용을 관리 
        - 다중 사용자 통제
            - 지정된 객체에 대해 다수의 사용자가 연합하여 접근을 요청할 경우 접근 통제를 지원하기 위한 수단
        - 컨텍스트 기반 통제
            - 특정 시간, 네트워크 주소 등 확인이 가능한 접근 경로나 위치, 인증 수준 등과 같은 외부적인
              요소에 의존하여 객체의 접근을 제어
    6. 감사 추적
