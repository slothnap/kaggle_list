### 데이터 관리 핵심 요소
- 데이터 수집, 처리, 검증, 저장
- 정형, 비정형 데이터를 비롯해 다양한 소스와 유형의 데이터 통합
- 데이터 고가용성과 재해 복구 지원
- 사람과 앱이 데이터를 사용하고 액세스하는 방법 관리
- 데이터 보호 및 보안, 데이터 개인정보보호 보장


## 데이터 관리 시스템 및 구성요소
- 데이터 관리 시스템은 데이터 관리 플랫폼에 구축되며, 여러 구성요소와 프로세스가 함께 작동해 데이터로부터 정확한 가치를 추출할 수 있도록 지원합니다. 여기에는 데이터베이스 관리 시스템, 데이터 웨어하우스 및 데이터 레이크, 데이터 통합 툴, 분석 등이 포함될 수 있습니다.

### 데이터베이스 관리 시스템(DBMS)
- 데이터베이스 관리 시스템의 종류는 매우 다양합니다. 가장 일반적인 데이터베이스 관리시스템에는 관계형 데이터베이스 관리 시스템(RDBMS), 객체 지향 데이터베이스 관리 시스템(OODMBS), 인메모리 데이터베이스, 컬럼 데이터베이스 등이 있습니다.




- **관계형 데이터베이스 관리 시스템(RDBMS)**: 관계형 데이터베이스 관리 시스템은 데이터 정의가 포함되어 있어 프로그램과 검색 시스템이 매번 데이터의 구조와 위치를 설명하는 대신 이름별로 데이터 항목을 참조할 수 있도록 하는 데이터베이스 관리 시스템입니다. 관계형 모델을 기반으로 하는 관계형 데이터베이스 관리 시스템은 데이터 항목 간의 관계를 유지관리해 액세스를 향상하고 중복을 방지합니다. 예를 들어 항목의 기본 정의와 특징을 저장하고 나면 고객 주문 세부정보 라인과 가격 테이블에 연결됩니다.


- **객체 지향 데이터베이스 관리 시스템(OODBMS)**: 객체 지향 데이터베이스 관리 시스템은 객체 지향 프로그래밍 시스템(OOPS) 개발자들이 개발하고 사용하는 데이터 정의와 스토리지에 대한 다른 접근 방식을 말합니다. 데이터는 관계형 데이터베이스 관리 시스템의 테이블과는 다르게 객체, 즉 독립적이고 자기 기술적인 엔티티로 저장됩니다.


- **인메모리 데이터베이스**: 인메모리 데이터베이스(IMDB) 는 디스크 드라이브 대신 컴퓨터의 메인 메모리(RAM)에 데이터를 저장합니다. 메모리의 검색 속도는 디스크 기반 시스템에서보다 훨씬 빠르며, 따라서 인메모리 데이터베이스는 주로 응답 시간이 빨라야 하는 애플리케이션에서 사용됩니다. 예를 들어 리포트에 컴파일하는 데 며칠이 걸리던 프로세스에 몇 분 내지는 몇 초 만에 액세스해 분석을 마칠 수 있습니다.


- **컬럼 방식 데이터베이스**: 컬럼 방식 데이터베이스는 빠른 액세스를 위해 서로 연관된 데이터 그룹(정보의 '열')을 함께 저장합니다. 컬럼 방식 데이터베이스는 최신 인메모리 비즈니스 애플리케이션 및 검색 속도(제한된 범위의 데이터에 대한)가 중요한 많은 독립형 데이터 웨어하우스 애플리케이션에 사용됩니다.

### 데이터 웨어하우스와 데이터 레이크
- **데이터 웨어하우스**: 데이터 웨어하우스는 리포팅과 분석을 목적으로 다양한 소스에서 축적된 데이터의 중앙 저장소입니다.
- **데이터 레이크**: 데이터 레이크는 데이터가 미가공 또는 원래 형식으로 저장되는 방대한 데이터 풀입니다. 데이터 레이크는 일반적으로 정형, 비정형, 반정형 데이터를 포함한 빅데이터를 저장하는 데 사용됩니다.

### 마스터 데이터 관리(MDM)
- 마스터 데이터 관리는 제품 데이터, 고객 데이터, 자산 데이터, 재무 데이터 등 중요한 모든 비즈니스 데이터의 신뢰할 수 있는 단일 마스터 참조(단일 버전의 데이터)를 생성하는 분야입니다. 마스터 데이터 관리는 프로세스, 운영, 분석, 리포팅을 비롯하여 비즈니스의 다양한 부분에서 일관되지 않은 여러 데이터 버전을 사용하지 않도록 도와줍니다. 효과적인 마스터 데이터 관리의 세 가지 핵심 요소는 데이터 통합, 데이터 거버넌스, 데이터 품질 관리입니다.

### 빅데이터 관리
- 오늘날 비즈니스에 범람하는 방대한 양의 정형, 비정형, 반정형 데이터와 같은 빅데이터를 관리하기 위해 새로운 유형의 데이터베이스와 툴이 개발되어 왔습니다. 방대한 분량과 속도를 처리하기 위한 고효율 처리 기술과 클라우드 기반 시설은 데이터의 다양성을 해석하고 관리하기 위해 형성된 새로운 접근 방식입니다. 데이터 관리 툴이 다양한 종류의 비정형 데이터를 이해하고 활용할 수 있도록 데이터 항목을 식별하고 분류하는 새로운 전처리 프로세스로 스토리지와 검색을 용이하게 하는 등의 역할을 합니다.

### 데이터 통합
- 데이터 통합은 데이터가 필요한 위치와 시기에 데이터를 수집, 변환, 결합, 프로비저닝하는 활동입니다. 이러한 통합은 기업 내에서뿐 아니라 파트너와 제3자 데이터 소스 및 적용 사례에도 일어나며 모든 애플리케이션과 비즈니스 프로세스의 데이터 사용 요건을 충족합니다. 통합 기술에는 벌크/배치 데이터 이동, 추출, 변환, 로드(ETL), 변경 데이터 캡처, 데이터 복제, 데이터 가상화, 스트리밍 데이터 통합, 데이터 조율 등이 있습니다.

### 데이터 거버넌스, 보안, 규제준수
- 데이터 거버넌스는 조직 전반의 데이터 가용성, 품질, 규제준수, 보안을 보장하는 일련의 규칙과 책임을 말합니다. 데이터 거버넌스는 인프라를 구축하고 특정 종류 및 유형의 데이터를 처리하고 보호할 권한과 책임을 지닌 조직 내 개인(또는 직위)을 임명합니다. 데이터 거버넌스는 규제준수의 핵심 부분입니다. 시스템은 저장, 처리, 보안 메커니즘을 관리합니다. 이는 사용자, 즉 거버넌스 측면에서 데이터를 시스템에 입력하기 이전, 데이터를 사용하는 동안, 데이터를 사용하거나 다른 곳에 저장하기 위해 시스템에서 검색할 때 모두 데이터가 처음부터 정확하며 올바르게 처리 및 보호받도록 보장합니다. 거버넌스는 책임자가 데이터를 관리 및 보호하기 위해 프로세스와 기술을 어떻게 사용해야 하는지 구체적으로 명시합니다.


- 물론 데이터 보안은 해커, 바이러스, 사이버 공격, 데이터 침해 문제가 대두되는 오늘날의 세계에서 주요한 관심사입니다. 시스템과 애플리케이션에 보안 기능이 내장되어 있기는 하지만 데이터 거버넌스는 이러한 시스템이 올바르게 설치 및 관리되어 데이터를 보호하고 여러 절차와 책임을 시행해 시스템과 데이터베이스 외부의 데이터를 보호하는 환경을 보장하기 위해 존재합니다.

### 비즈니스 인텔리전스 및 분석
- 대부분의 데이터 관리 시스템에는 기본 데이터 검색 및 리포팅 툴이 포함되어 있으며 많은 시스템이 강력한 검색, 분석, 리포팅 애플리케이션과 통합되어 있거나 번들로 제공됩니다. 리포팅 및 분석 애플리케이션은 제3사 개발자도 사용할 수 있으며 기본 기능 또는 고급 기능을 위한 선택형 추가 구성 모듈로 애플리케이션 번들에 포함되는 경우가 많습니다.
 

- 오늘날 데이터 관리 시스템의 강점은 대부분 최소한의 교육을 받은 사용자가 자체적으로 온스크린 데이터를 가져오고 형식 지정, 계산, 분류, 요약 면에서 유연성 높게 리포트를 인쇄할 수 있는 임시 검색 툴입니다. 또한 전문가는 동일한 툴 또는 보다 정교한 분석 툴 세트를 사용한 계산, 비교, 고등 수학, 형식 지정을 통해 더 많은 작업을 수행할 수 있습니다. 새로운 분석 애플리케이션으로는 기존 데이터베이스, 데이터 웨어하우스, 데이터 레이크를 서로 연결하여 빅데이터를 비즈니스 애플리케이션 데이터와 통합함으로써 예측, 분석, 계획을 더욱 개선할 수 있습니다.

### 기업 데이터 전략은 무엇이고 왜 필요한가?
- 많은 기업이 데이터 전략에 관해서는 비즈니스 애플리케이션 공급업체가 시스템에 구축한 모든 것을 그대로 수용하는 수동적인 태도를 취해 왔습니다. 하지만 이제는 상황이 바뀌었습니다. 데이터가 폭발적으로 늘어나고 모든 기업의 운영에 중요해지면서 더 능동적이고 포괄적인 접근 방식으로 데이터를 관리해야 할 필요성이 점점 커지고 있습니다. 실용적인 측면에서 이는 다음과 같은 데이터 전략을 설정하는 준비를 뜻합니다.


- 기업에 필요하고 기업이 사용하게 될 특정 데이터 유형을 식별
- 데이터 유형별 책임자 지정
- 데이터의 획득, 수집, 처리를 관리하는 절차 수립


- 기업 데이터 관리 전략 및 인프라의 주요 이점 가운데 하나는 기업의 목적을 지원하는 모든 활동과 의사결정을 조율함으로써 조직을 하나로 연결하여 고객에게 고품질의 제품과 서비스를 효과적이고 효율적인 방식으로 제공할 수 있다는 점입니다. 종합적인 데이터 전략과 원활한 데이터 통합은 정보의 단절을 없애 줍니다. 이를 통해 각 부서, 관리자, 직원은 기업의 성공에 대한 개개인의 기여도를 확인하고 파악하여 이러한 목표에 맞춰 의사결정과 조치를 취할 수 있습니다.

### 데이터 관리의 진화
- 효과적인 데이터 관리는 50년이 훨씬 넘는 세월 동안 비즈니스의 성공에 핵심적인 역할을 해 왔습니다. 기업이 정보 리포팅의 정확성을 개선하고 동향을 파악하며 더 나은 의사결정을 내리도록 지원하는 데서부터 디지털 혁신을 이끌고 오늘날의 새로운 기술과 비즈니스 모델에 힘을 불어넣는 일까지 그 역할은 매우 다양합니다. 데이터는 새로운 종류의 자산이 되었으며 미래지향적인 조직은 언제나 데이터를 조직에 유리하게 활용할 새롭고 더 나은 방안을 모색하고 있습니다. 비즈니스와 업계의 관련성을 주의 깊게 탐색하는 데 중요한 데이터 관리의 최신 동향은 다음과 같습니다.


- **데이터 패브릭**: 오늘날 대부분의 조직은 온프레미스와 클라우드에 다양한 유형의 데이터를 배포하여 보유하고 있으며 여러 데이터베이스 관리 시스템, 처리 기술, 툴을 사용합니다. 데이터 패브릭은 아키텍처와 기술의 사용자 정의 조합으로 역동적 데이터 통합과 조율을 사용해 분산된 환경 전체의 데이터에 원활한 액세스와 공유를 지원합니다.


- **클라우드에서의 데이터 관리**: 많은 기업이 데이터 관리 플랫폼의 일부 또는 전체를 클라우드로 이전하고 있습니다. 클라우드에서 데이터를 관리하면 확장성, 고급 데이터 보안, 데이터 액세스 개선, 자동 백업 및 재해 복구, 비용 절감 등 클라우드가 지닌 모든 이점을 활용할 수 있습니다. 클라우드 데이터베이스와 DBaaS(Database-as-a-Service) 솔루션, 클라우드 데이터 웨어하우스, 클라우드 데이터 레이크는 모두 크게 인기를 끌고 있습니다.


- **증강 데이터 관리**: 요즘 새롭게 등장한 동향 가운데 하나는 바로 '증강 데이터 관리'입니다. 가트너(Gartner)에서는 증강 데이터 관리가 2022년까지 상당한 잠재적 영향력을 지닐 것으로 예상합니다. 증강 데이터 관리는 AI와 머신러닝을 사용해 데이터 관리 프로세스의 자체 구성 및 조정을 가능하게 합니다. 증강 데이터 관리는 데이터 품질 및 마스터 데이터 관리에서 데이터 통합에 이르는 모든 프로세스를 자동화해 전문 기술을 지닌 직원이 더욱 가치 있는 작업에 집중할 수 있도록 합니다.


- **증강 분석**: 가트너(Gartner)에서 또 하나의 기술 동향 중 하나로 제시한  증강 분석은 이미 현실이 되었습니다. 증강 분석은 인공지능, 머신러닝, 자연어 처리(NLP)를 활용해 대부분의 중요 인사이트를 자동으로 찾아낼 뿐 아니라 고급 분석에 대한 액세스를 대중화해 데이터 과학자뿐 아니라 모든 사용자가 데이터에 관해 질문하고 자연스러운 대화의 형태로 답변을 얻을 수 있 합니다.
