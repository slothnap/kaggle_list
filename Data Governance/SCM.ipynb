### SCM
- 공급망 관리에 대한 정말 기본적인 정의는 고객에게 제품을 배송하기 위해 뒤에서 이루어지는 모든 작업입니다. 즉 구매와 주문, 제조, 운송, 재고 관리 등이 포함됩니다. 이 프로세스는 시장의 목소리에 귀를 기울여 고객이 원하는 것이 무엇이고 원하는 시기와 그 정도를 파악하는 데서 시작됩니다. 그런 다음 소싱에서 생산, 물류에 이르는 모든 단계를 진행합니다. 이처럼 매우 복잡한 작업이 최대한 효율적이려면 각 파트너, 즉 망 내의 '연결 고리'를 긴밀하게 조율된 반응형 시스템으로 통합해야 합니다.


### 공급망 관리는 왜 중요할까요?
- 주위를 둘러보세요. 기본적으로 공급망 없이 집이나 직장에 있는 것은 없습니다. 전 세계 수백억 개 일자리가 이러한 활동과 연결되어 있습니다. 저렴한 소비재부터 수술 장비, 중요한 자원에 이르기까지 모두가 공급망을 통해 제공됩니다. 그런데 SCM이 글로벌 경제의 핵심임에도 불구하고 많은 기업이 여전히 50년 동안 사용했던 프로세스와 기계로 공급망을 운영하고 있습니다.

 
- SCM 관행을 개선하면 비즈니스를 혁신할 수 있습니다. 기업은 낭비와 잉여를 최소화하는 한편 비용을 절감하고 효율성을 높임으로써 경쟁력을 강화할 수 있습니다. 개인 취향에 맞는 개인화된 물류를 제공함으로써 고객 충성도를 높일 수 있습니다. 또한 프로세스를 자동화해 더 빠르고 스마트하며 보다 생산적으로 개선할 수 있습니다.  

### 글로벌 공급망 관리 – 과거와 현재
- 글로벌 공급망은 지난 수십 년에 걸쳐 일부 국가에 존재하는 낮은 임금과 저렴한 원자재를 이용하기 위해 구축되었습니다. 이제 세상은 여러 중요한 측면에서 달라졌습니다. 무역 및 관세 정책이 변화하고 예측할 수 없는 경우가 많습니다. 미래는 정치적으로 불확실합니다. 고객은 원자재에서 포장에 이르는 모든 과정, 집까지 오는 배송 트럭에 사용되는 연료에 이르기까지 제품의 원산지와 지속가능성에 대한 정보를 얻고자 합니다. 다행히 기술에도 변화가 있었습니다. 최신 공급망은 공급망의 모든 연결 고리에서 생성되는 빅데이터를 사용래 생산 및 제조를 합리화하는 동시에 자체 개발 공급망 솔루션 개발을 늘리고 있습니다.

## 공급망 관리의 기본 구성요소
### 공급망 계획이란?
- 공급망 계획은 제품 수요를 예측한 후 제품을 제공하기 위해 공급망의 모든 연결 고리를 조정하는 프로세스입니다. 수요 예측 및 계획 외에도 공급 계획, 자재 소요량 계획(MRP), 생산 계획, 판매 및 운영 계획(S&OP)이 포함됩니다. 최신 SCM에서는 일반적으로 이러한 계획이 모두 하나의 통합사업계획(IBP)에 연결됩니다.
 

### 물류 관리란?
- 물류 관리는 원자재 및 제조가 포함된 공급망의 시작부터 매장이나 고객에게 완제품을 배송하는 동안은 물론 제품 서비스 제공, 반품, 재활용과 관련된 제품의 운송과 보관에 해당합니다. 관련 비즈니스 직무에는 인바운드 및 아웃바운드 운송 관리, 차량군 관리, 창고 관리, 재고 관리, 고객 서비스가 있습니다.
 

### 제조 기술이란?
- SCM에서 제조의 가장 큰 우선순위는 품질과 지속가능성, 고객 만족도의 유지관리(및 개선)와 함께 생산 및 프로세스를 최대한 합리화하는 데 있습니다. 공급망의 인공지능(AI) 및 사물인터넷(IoT) 기반 시스템은 빅데이터를 수집, 분석go 제조 공정을 합리화, 자동화할 수 있습니다. 온디맨드 3D 인쇄를 사용하면 부족과 잉여가 발생하지 않고 스마트 머신은 대량 주문 제작을 경제적으로 수행할 수 있습니다. 
 

### 제품 수명주기 관리(PLM)란?
- 제품 수명주기 관리(PLM)는 아이디어, 엔지니어링, 디자인에서 제조, 서비스, 폐기(또는 재활용)에 이르는 전체 수명주기 동안 제품을 관리하는 프로세스입니다.  PLM 소프트웨어 시스템은 이러한 프로세스를 통합하고 전사적 차원의 협업을 촉진하며 수명주기 동안 모든 제품에 대한 제품 정보 백본을 제공합니다. 
 

### 기업자산관리(EAM)란?
- 기업 자산 관리는 공장 로봇에서 배송 차량군에 이르기까지 공급망 전반에 걸쳐 실물 자산을 관리하고 유지보수하는 프로세스입니다. IoT 센서, 사물지능통신(M2M) 연결, 디지털 트윈은 EAM을 혁신해 효율성과 가동 시간, 안전, 예방 및 예지보전 등을 개선하고 있습니다. 일부 연결 자산은 수리 또는 고장을 예상하고 스스로 유지보수를 수행하며 수명주기를 연장하는 데 필요한 부품을 바로 소싱하고 주문할 수 있습니다.

 

### 공급망 조달이란?
- 조달은 비즈니스 요구사항을 충족하기 위해 자재, 상품, 서비스를 확보하고 해당 상품의 품질과 가치를 보장하며 공정한 가격을 협상하는 프로세스입니다. 조달 및 소싱 팀의 주요 당면 과제는 부족과 잉여 모두 비즈니스에 손해를 입힐 수 있기 때문에 정확한 주문량을 예상하는 것입니다. 머신러닝과 예측 분석이 통합된 SCM 시스템을 사용하면 조달 및 구매 과정에서 추측을 배제할 수 있습니다.

### 공급망 관리의 이점
- 생산성 향상: 기업 자산 관리 시스템 과 예지보전(예측정비)은 기계와 시스템의 운영 효율 개선을 돕습니다. 이를 통해 병목 현상을 해결하고 워크플로를 개선하며 생산성을 높일 수 있습니다. 프로세스 자동화와 반응형 데이터 분석도 출하 및 납품 소요시간을 단축합니다.


- 공급망 비용 절감: 예측 분석을 사용하면 비용이 많이 드는 '추측'을 배제함으로써 낭비되는 재고와 위험한 부족 현상을 줄일 수 있습니다. IoT를 적용하면 기존 자산의 대응성이 향상되고 모든 상황에 가장 효율적이고 유용한 워크플로우가 실현됩니다. 또한 더 정확한 예측을 통해 반쯤 채워진 배송 트럭, 조정되지 않은 배송 경로, 비효율적인 차량군 관리를 줄일 수 있습니다.


- 공급망의 민첩성과 회복탄력성 향상: 트렌드와 시장 상황이 급변할 수 있으므로 모든 상황에 적응할 수 있도록 민첩성 있는 회복탄력적인 SCM 시스템의 확보가 중요합니다. 실시간 데이터와 스마트 인사이트를 이용하면 공급망 관리자가 기계와 직원을 향상된 워크플로에 재할당할 수 있습니다. 또한 고객 피드백을 듣고 즉시 조치를 취할 수 있습니다. 가상 재고 및 스마트 창고 프로세스는 수요와 공급을 지속적으로 조율합니다.


- 제품 품질 개선: 고객 피드백을 연구개발(R&D) 팀에 직접 제공함으로써 제품 디자인과 개발에 고객의 요구사항을 완벽하게 반영합니다. R&D 팀과 제조팀은 머신러닝과 분석에서 얻은 인사이트를 활용한 의미 있는 제품 디자인 개선으로 고객의 트렌드와 희망사항에 대응할 수 있습니다.


- 고객 서비스 향상: SCM 선진사례는 고객 중심이며 대응력과 적응력을 강화하도록 설계됩니다. 클릭 한 번으로 경쟁사와의 사업 성패가 갈리는 상황에서 최신 SCM으로 설계 및 제조 단계부터 최종 물류, 배송, 반품에 이르기까지 모든 과정에 고객 피드백과 트렌드를 구현할 수 있습니다.

### 공급망 관리의 발전
- 지난 수십 년 동안 고객은 공급망의 가장 마지막 단계에만 개입했습니다. 제품이 어디에서 왔는지, 무엇으로 만들어졌는지, 매장에 어떻게 도착했는지에 대해서는 크게 관심을 두지 않았습니다. 오늘날에는 공급망의 투명성과 지속가능성이 소비자에게 매우 중요한 요소입니다. 마찬가지로 고객이 쇼핑하는 방식과 위치, 주문 처리 과정에서 제공될 것으로 예상하는 옵션을 전체적으로 제어하는 능력도 중요합니다.

 

- 오늘날의 시장에서 성장하고 경쟁하기 위해서는 최신 SCM 소프트웨어를 통해 전체 공급망에서 생성 및 포착된 모든 데이터를 수집하고 해석할 수 있어야 합니다. 이러한 데이터를 실시간 통찰으로 전환하고 SCM 프로세스와 워크플로를 매우 스마트하고 민첩한 방식으로 자동화하는 데 사용하며 최대한 활용하기 위해서는 새로운 기술이 필요합니다. 우리는 미래를 예측할 수 없지만 경제적인 변화, 예기치 않은 사건, 빠르게 진화하는 고객 수요가 발생할 것이라는 사실은 확신할 수 있습니다. 데이터 및 기술 중심의 SCM 시스템을 사용하면 공급망 관리를 혁신하고 2020년대와 그 이후에도 대응력을 갖춘 새로운 종류의 공급망을 구축할 수 있습니다.

## 공급망 관리 트렌드
### 코로나바이러스가 공급망에 미치는 영향
- 2019년에는 2020년에 겪게 될 혼돈과 혼란의 물결에 대해 아무도 알지 못했습니다. 이러한 상황을 알 수는 없었지만 확실히 더 잘 준비할 수 있었습니다. 전 세계 기업들은 코로나19로 인해 일부 공급망 프로세스가 얼마나 취약하고 구시대적이었는지 깨닫게 되었습니다. 인공 지능, 머신 러닝, 예측 분석을 SCM 시스템에 통합한 기업은 스마트한 의사결정을 내리고 예기치 못한 상황이 발생했을 때 신속하게 적응하며 탄력성을 향상시킬 수 있습니다.  

 

### 아마존 효과 
- 아마존(Amazon)은 광대한 창고와 이행 네트워크 덕분에 익일 및 당일 배송 서비스를 제공할 수 있게 되었으며, 이제 고객은 모든 온라인 소매업체에 동일한 배송 속도를 기대합니다. 아마존 효과는 SCM에 큰 영향을 미치고 있으며, 이러한 기대치를 충족하기 위해서는 대규모 창고 네트워크부터 아웃소싱, 라스트 마일 배송에 이르기까지 광범위한 구조 조정이 필요합니다.

 

### 친환경 공급망
- 공급망 관리에 환경친화적인 관행을 통합하면 제품 디자인 및 자재 조달부터 제조, 물류, 배송 그리고 사용 수명 종료 시 제품과 장비에 수행하는 모든 조치를 혁신할 수 있습니다. 순환 공급망에서는 폐기된 제품의 플라스틱과 전자 부품 등을 재활용하는 데 그치지 않고 용도를 변경하는 방법도 모색합니다.

 

### 공급망 투명성
- 2020년, 고객은 글로벌 경제에 대해 그 어느 때보다 잘 알고 있으며 공급망 관행과 정책이 윤리적이라고 알려진 기업과 거래하려는 의욕으로 가득 차 있습니다. 공급망 투명성과 지속가능성을 구축하기 위해 블록체인 및 추적 센서와 같은 기술이 사용되고 있습니다. 고객은 윤리적인 원자재 제공과 생산에 사용되는 노동부터 배송 트럭의 연료에 이르기까지 최신 공급망에 포함된 모든 연결 고리에 대한 원산지와 출처를 확인할 수 있습니다.

 

### 옴니채널 공급망
- 옴니채널 쇼핑에 대한 기대는 점점 더 복잡해지고 있습니다. 기업은 단순히 온라인 매장을 제공하는 것을 넘어 가상에서 오프라인으로 또는 다시 가상으로 원활하게 이동할 수 있는 쇼핑 경험을 제공해야 합니다. 고객은 지불 방법에서 주문 장소, 상품 수령 방법, 주문 과정에서 정보를 제공하는 방식에 이르기까지 자신의 쇼핑 및 이행 선택사항을 조합하여 개인화할 수 있기를 바랍니다. 최적의 SCM 시스템은 모든 고객 접점에서 고객 서비스와 쇼핑 선택사항을 통합합니다.

 

### 무역 및 정치적 변화 
- 지난 몇 년에 걸쳐 새로운 관세 및 무역 정책에서 예측 불가능성이 극명하게 드러났습니다. 한때 해외 공급업체와 제조업체에 거의 전적으로 의존하던 기업들은 그로 인한 취약성을 깨닫고 자국에서 더 가까운, 보다 안전한 옵션을 모색하고 있습니다. IoT 및 주문 제조 솔루션(예: 3D 인쇄)으로 인해 가상 재고 및 국내 제조 솔루션이 점점 더 현실적인 선택사항이 되고 있습니다.
