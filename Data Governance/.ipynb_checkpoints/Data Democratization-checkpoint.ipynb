### 데이터 민주화란
- 데이터 민주화는 기술적인 노하우와 관계없이 조직의 모든 사람이 편안하게 데이터로 작업 하고 데이터 에 대해 자신있게 이야기할 수 있도록 하여 결과적으로 데이터 정보에 입각한 의사 결정 을 내리고 고객 경험 을 구축 할 수 있도록 하는 지속적인 프로세스 입니다. 


### 데이터 민주화 3대 요소
- 직원들이 편안하게 데이터 관련 질문을 할 수 있도록 지원
- 모든 사람이 데이터 작업을 할 수 있도록 올바른 도구 제공
- 데이터의 민주화를 조직 전체의 문화적 변화가 필요할 수도 있는 진행 중인 프로세스로 인식

## 데이터 민주화 핵심
### 가장 일반적인 데이터 문제
- 필요한 데이터에 액세스할 수 없습니다.
- 데이터를 믿을 수 없어
- 데이터에 접근할 수 있지만 질문에 대한 답변을 찾는 기술이 부족합니다.
- 우리 회사에서 제공하는 분석 도구는 제품 팀을 위해 설계되지 않았습니다.
- 우리 회사의 데이터 전문가는 너무 바빠서 나를 도울 수 없습니다.


### 팀별 데이터 사용 방식
- 마케팅 은 데이터와 협력하여 매력적이고 전환율이 높은 콘텐츠와 캠페인을 만듭니다.
- Growth 는 데이터와 함께 실험을 실행하고 개인화된 경험을 제공합니다.
- 제품 및 엔지니어링 은 데이터를 사용하여 고객 가치를 창출하는 기능을 구축하고 그렇지 않은 기능은 폐기합니다.
- 지원 은 데이터와 함께 작동하여 더 빠른 해결을 제공합니다(사용자가 제품 내에서 수행한 작업과 수행하지 않은 작업을 확인하여).
- Customer Success 는 데이터를 사용하여 더 나은 고객 경험을 제공합니다(사용 패턴을 기반으로 고객에게 올바른 질문을 함으로써).
- Sales 는 데이터를 사용하여 전환할 가능성이 있는 잠재 고객을 식별합니다(무료 평가판 동안 수행한 작업을 확인하여).
- 경영진 은 데이터를 사용하여 비즈니스가 어떻게 수행되고 있는지, 향후 어디에 투자해야 하는지 이해합니다.- 

### 데이터 Tool
- Snowflake, BigQuery 또는 Firebolt와 같은 데이터 웨어하우스는 데이터를 분석 및 활성화에 사용할 수 있도록 합니다.
- 웨어하우스 상단에 위치하며 셀프 서비스 분석을 가능하게 하는 Looker, Mode 또는 Superset과 같은 비즈니스 인텔리전스(BI) 도구
- Airbyte, Fivetran 또는 Meltano와 같은 ELT 도구로 타사 도구(위에서 언급한 도구 등)에서 웨어하우스로 데이터 이동
- Census, Hightouch 또는 Grouparoo와 같은 역 ETL 도구로 모델링된 데이터를 웨어하우스에서 타사 도구로 다시 이동

### 조직의 문화적 변화가 필요할 수 있는 데이터 민주화가 진행 중인 프로세스인 이유는 무엇입니까?
- 먼저 회사의 규모와 성장 궤적이 데이터 민주화가 이루어지는 속도에 큰 영향을 미친다는 점부터 말씀드리고 싶습니다. 말할 필요도 없이 데이터 민주주의를 지원하는 문화를 형성하기가 더 쉽기 때문에 회사 초기에는 데이터 민주주의를 구축하는 것이 훨씬 쉽습니다.

- 규모가 큰 조직은 데이터 민주주의가 데이터를 "데모 미친 " 것처럼 보이게 만드는 수많은 문제에 직면할 가능성이 높습니다 .

- 조직이 클수록 데이터 민주화 프로세스에 더 빨리 투자해야 합니다. 조직에서 제품 분석을 설정하려는 경우 데이터 관리에 대한 이 가이드가 유용할 수 있습니다 .

### 문화적 변화
- 자신의 업무에서 탁월하고 목표를 달성하기 위해 데이터에 의존하는 모든 사람은 데이터 전문가가 되어야 합니다.
- 의 모든 사람은 데이터에 대해 자신 있게 이야기하고 데이터 작업을 수행하고 종속성 없이 질문에 대한 답변을 얻을 수 있는 도구와 지식을 갖추어야 합니다.
- 막으로, 조직의 모든 사람에게 데이터 프로젝트에 의미 있는 기여를 할 수 있는 기회가 주어져야 합니다.
